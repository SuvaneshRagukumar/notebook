```python
user_data = {
    "users":[
        {
            "userid":1,
            "name":"john",
            "email":"john@email.com"
        },
        {
            "userid":2,
            "name":"david",
            "email":"david@email.com"
        }
    ]
}
```

```python
print(type(user_data))
```

```python
import json
```

```python
with open("user_data_org.json","w") as f:
    json.dump(user_data,f)
```

```python
#reading a json file
```

```python
with open("user_data_org.json") as f1:
    json_data = json.load(f1)
```

```python
print(json_data)
```

```python
import pandas as pd
```

```python
json_df=pd.DataFrame(json_data)
```

```python
json_df
```

```python
json_df_org = pd.json_normalize(json_df["users"])
```

```python
json_df_org
```

```python
json_df_org.to_csv("json_data_CSVFILE.csv")
```
