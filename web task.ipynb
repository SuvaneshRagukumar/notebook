```python
f=open('suvanesh.html',"w")
value= """<!DOCTYPE html>
<html class="client-nojs" lang="en" dir="ltr">
<head>
<meta charset="UTF-8"/>
<title>List of prime ministers of India - Wikipedia</title>
<script>document.documentElement.className="client-js";RLCONF={"wgBreakFrames":false,"wgSeparatorTransformTable":["",""],"wgDigitTransformTable":["",""],"wgDefaultDateFormat":"dmy","wgMonthNames":["","January","February","March","April","May","June","July","August","September","October","November","December"],"wgRequestId":"612ab820-c80a-4650-b455-8f5411a7f9a4","wgCSPNonce":false,"wgCanonicalNamespace":"","wgCanonicalSpecialPageName":false,"wgNamespaceNumber":0,"wgPageName":"List_of_prime_ministers_of_India","wgTitle":"List of prime ministers of India","wgCurRevisionId":1078988291,"wgRevisionId":1078988291,"wgArticleId":844745,"wgIsArticle":true,"wgIsRedirect":false,"wgAction":"view","wgUserName":null,"wgUserGroups":["*"],"wgCategories":["Pages using the EasyTimeline extension","Articles with short description","Short description is different from Wikidata","Featured lists","Use dmy dates from September 2020","Use Indian English from May 2013","All Wikipedia articles written in Indian English"
,"Articles with hCards","Prime Ministers of India","Lists of prime ministers by country","Lists of political office-holders in India","Lists relating to prime ministers of India"],"wgPageContentLanguage":"en","wgPageContentModel":"wikitext","wgRelevantPageName":"List_of_prime_ministers_of_India","wgRelevantArticleId":844745,"wgIsProbablyEditable":true,"wgRelevantPageIsProbablyEditable":true,"wgRestrictionEdit":[],"wgRestrictionMove":[],"wgFlaggedRevsParams":{"tags":{"status":{"levels":-1}}},"wgVisualEditor":{"pageLanguageCode":"en","pageLanguageDir":"ltr","pageVariantFallbacks":"en"},"wgMFDisplayWikibaseDescriptions":{"search":true,"nearby":true,"watchlist":true,"tagline":false},"wgWMESchemaEditAttemptStepOversample":false,"wgWMEPageLength":30000,"wgNoticeProject":"wikipedia","wgMediaViewerOnClick":true,"wgMediaViewerEnabledByDefault":true,"wgPopupsFlags":10,"wgULSCurrentAutonym":"English","wgEditSubmitButtonLabelPublish":true,"wgCentralAuthMobileDomain":false,"wgULSPosition":
"interlanguage","wgULSisCompactLinksEnabled":true,"wgWikibaseItemId":"Q492746","wgGENewcomerTasksGuidanceEnabled":true,"wgGEAskQuestionEnabled":false,"wgGELinkRecommendationsFrontendEnabled":false};RLSTATE={"ext.globalCssJs.user.styles":"ready","site.styles":"ready","user.styles":"ready","ext.globalCssJs.user":"ready","user":"ready","user.options":"loading","ext.cite.styles":"ready","ext.timeline.styles":"ready","skins.vector.styles.legacy":"ready","jquery.tablesorter.styles":"ready","jquery.makeCollapsible.styles":"ready","ext.visualEditor.desktopArticleTarget.noscript":"ready","ext.wikimediaBadges":"ready","ext.uls.interlanguage":"ready","wikibase.client.init":"ready"};RLPAGEMODULES=["ext.cite.ux-enhancements","ext.scribunto.logs","site","mediawiki.page.ready","jquery.tablesorter","jquery.makeCollapsible","mediawiki.toc","skins.vector.legacy.js","ext.gadget.ReferenceTooltips","ext.gadget.charinsert","ext.gadget.extra-toolbar-buttons","ext.gadget.refToolbar","ext.gadget.switcher",
"mmv.head","mmv.bootstrap.autostart","ext.visualEditor.desktopArticleTarget.init","ext.visualEditor.targetLoader","ext.eventLogging","ext.wikimediaEvents","ext.navigationTiming","ext.cx.eventlogging.campaigns","ext.centralNotice.geoIP","ext.centralNotice.startUp","ext.centralauth.centralautologin","ext.popups","ext.uls.compactlinks","ext.uls.interface","ext.growthExperiments.SuggestedEditSession"];</script>
<script>(RLQ=window.RLQ||[]).push(function(){mw.loader.implement("user.options@1i9g4",function($,jQuery,require,module){mw.user.tokens.set({"patrolToken":"+\\","watchToken":"+\\","csrfToken":"+\\"});});});</script>
<link rel="stylesheet" href="/w/load.php?lang=en&amp;modules=ext.cite.styles%7Cext.timeline.styles%7Cext.uls.interlanguage%7Cext.visualEditor.desktopArticleTarget.noscript%7Cext.wikimediaBadges%7Cjquery.makeCollapsible.styles%7Cjquery.tablesorter.styles%7Cskins.vector.styles.legacy%7Cwikibase.client.init&amp;only=styles&amp;skin=vector"/>
<script async="" src="/w/load.php?lang=en&amp;modules=startup&amp;only=scripts&amp;raw=1&amp;skin=vector"></script>
<meta name="ResourceLoaderDynamicStyles" content=""/>
<link rel="stylesheet" href="/w/load.php?lang=en&amp;modules=site.styles&amp;only=styles&amp;skin=vector"/>
<meta name="generator" content="MediaWiki 1.39.0-wmf.4"/>
<meta name="referrer" content="origin"/>
<meta name="referrer" content="origin-when-crossorigin"/>
<meta name="referrer" content="origin-when-cross-origin"/>
<meta name="format-detection" content="telephone=no"/>
<meta property="og:title" content="List of prime ministers of India - Wikipedia"/>
<meta property="og:type" content="website"/>
<link rel="preconnect" href="//upload.wikimedia.org"/>
<link rel="alternate" media="only screen and (max-width: 720px)" href="//en.m.wikipedia.org/wiki/List_of_prime_ministers_of_India"/>
<link rel="alternate" type="application/x-wiki" title="Edit this page" href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit"/>
<link rel="apple-touch-icon" href="/static/apple-touch/wikipedia.png"/>
<link rel="shortcut icon" href="/static/favicon/wikipedia.ico"/>
<link rel="search" type="application/opensearchdescription+xml" href="/w/opensearch_desc.php" title="Wikipedia (en)"/>
<link rel="EditURI" type="application/rsd+xml" href="//en.wikipedia.org/w/api.php?action=rsd"/>
<link rel="license" href="https://creativecommons.org/licenses/by-sa/3.0/"/>
<link rel="canonical" href="https://en.wikipedia.org/wiki/List_of_prime_ministers_of_India"/>
<link rel="dns-prefetch" href="//meta.wikimedia.org" />
<link rel="dns-prefetch" href="//login.wikimedia.org"/>
</head>
<body class="mediawiki ltr sitedir-ltr mw-hide-empty-elt ns-0 ns-subject mw-editable page-List_of_prime_ministers_of_India rootpage-List_of_prime_ministers_of_India skin-vector action-view skin-vector-legacy"><div id="mw-page-base" class="noprint"></div>
<div id="mw-head-base" class="noprint"></div>
<div id="content" class="mw-body" role="main">
	<a id="top"></a>
	<div id="siteNotice"><!-- CentralNotice --></div>
	<div class="mw-indicators">
	<div id="mw-indicator-featured-star" class="mw-indicator"><a href="/wiki/Wikipedia:Featured_lists" title="This is a featured list. Click here for more information."><img alt="This is a featured list. Click here for more information." src="//upload.wikimedia.org/wikipedia/en/thumb/e/e7/Cscr-featured.svg/20px-Cscr-featured.svg.png" decoding="async" width="20" height="19" srcset="//upload.wikimedia.org/wikipedia/en/thumb/e/e7/Cscr-featured.svg/30px-Cscr-featured.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/e/e7/Cscr-featured.svg/40px-Cscr-featured.svg.png 2x" data-file-width="462" data-file-height="438" /></a></div>
	</div>
	<h1 id="firstHeading" class="firstHeading mw-first-heading">List of prime ministers of India</h1>
	<div id="bodyContent" class="vector-body">
		<div id="siteSub" class="noprint">From Wikipedia, the free encyclopedia</div>
		<div id="contentSub"></div>
		<div id="contentSub2"></div>
		
		<div id="jump-to-nav"></div>
		<a class="mw-jump-link" href="#mw-head">Jump to navigation</a>
		<a class="mw-jump-link" href="#searchInput">Jump to search</a>
		<div id="mw-content-text" class="mw-body-content mw-content-ltr" lang="en" dir="ltr"><div class="mw-parser-output"><p class="mw-empty-elt">



</p>
<style data-mw-deduplicate="TemplateStyles:r1045330069">.mw-parser-output .sidebar{width:22em;float:right;clear:right;margin:0.5em 0 1em 1em;background:#f8f9fa;border:1px solid #aaa;padding:0.2em;text-align:center;line-height:1.4em;font-size:88%;border-collapse:collapse;display:table}body.skin-minerva .mw-parser-output .sidebar{display:table!important;float:right!important;margin:0.5em 0 1em 1em!important}.mw-parser-output .sidebar-subgroup{width:100%;margin:0;border-spacing:0}.mw-parser-output .sidebar-left{float:left;clear:left;margin:0.5em 1em 1em 0}.mw-parser-output .sidebar-none{float:none;clear:both;margin:0.5em 1em 1em 0}.mw-parser-output .sidebar-outer-title{padding:0 0.4em 0.2em;font-size:125%;line-height:1.2em;font-weight:bold}.mw-parser-output .sidebar-top-image{padding:0.4em}.mw-parser-output .sidebar-top-caption,.mw-parser-output .sidebar-pretitle-with-top-image,.mw-parser-output .sidebar-caption{padding:0.2em 0.4em 0;line-height:1.2em}.mw-parser-output .sidebar-pretitle{padding:0.4em 0.4em 0;line-height:1.2em}.mw-parser-output .sidebar-title,.mw-parser-output .sidebar-title-with-pretitle{padding:0.2em 0.8em;font-size:145%;line-height:1.2em}.mw-parser-output .sidebar-title-with-pretitle{padding:0.1em 0.4em}.mw-parser-output .sidebar-image{padding:0.2em 0.4em 0.4em}.mw-parser-output .sidebar-heading{padding:0.1em 0.4em}.mw-parser-output .sidebar-content{padding:0 0.5em 0.4em}.mw-parser-output .sidebar-content-with-subgroup{padding:0.1em 0.4em 0.2em}.mw-parser-output .sidebar-above,.mw-parser-output .sidebar-below{padding:0.3em 0.8em;font-weight:bold}.mw-parser-output .sidebar-collapse .sidebar-above,.mw-parser-output .sidebar-collapse .sidebar-below{border-top:1px solid #aaa;border-bottom:1px solid #aaa}.mw-parser-output .sidebar-navbar{text-align:right;font-size:115%;padding:0 0.4em 0.4em}.mw-parser-output .sidebar-list-title{padding:0 0.4em;text-align:left;font-weight:bold;line-height:1.6em;font-size:105%}.mw-parser-output .sidebar-list-title-c{padding:0 0.4em;text-align:center;margin:0 3.3em}@media(max-width:720px){body.mediawiki .mw-parser-output .sidebar{width:100%!important;clear:both;float:none!important;margin-left:0!important;margin-right:0!important}}</style><table class="sidebar sidebar-collapse nomobile nowraplinks vcard hlist"><tbody><tr><th class="sidebar-title"><div class="adr"><a href="/wiki/Politics_of_India" title="Politics of India">Politics of India</a></div></th></tr><tr><th><a href="/wiki/File:Emblem_of_India.svg" class="image"><img alt="Emblem of India.svg" src="//upload.wikimedia.org/wikipedia/commons/thumb/5/55/Emblem_of_India.svg/70px-Emblem_of_India.svg.png" decoding="async" width="70" height="114" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/5/55/Emblem_of_India.svg/105px-Emblem_of_India.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/5/55/Emblem_of_India.svg/140px-Emblem_of_India.svg.png 2x" data-file-width="370" data-file-height="600" /></a></th></tr><tr><th style="border-bottom: #aaa 1px solid"></th></tr><tr><td class="sidebar-content">
<div class="sidebar-list mw-collapsible mw-collapsed"><div class="sidebar-list-title" style="text-align:center; background:lavender;"><a href="/wiki/Government_of_India" title="Government of India">Federal government</a></div><div class="sidebar-list-content mw-collapsible-content">
<ul><li><b><a href="/wiki/Constitution_of_India" title="Constitution of India">Constitution of India</a></b></li></ul>
<ul><li><a href="/wiki/Law_of_India" title="Law of India">Law</a></li>
<li><a href="/wiki/Human_rights_in_India" title="Human rights in India">Human rights</a></li></ul>
<hr />
<ul><li><a href="/wiki/Indian_Penal_Code" title="Indian Penal Code">Indian Penal Code</a></li>
<li><a href="/wiki/Law_enforcement_in_India" title="Law enforcement in India">Law enforcement</a></li></ul></div></div></td>
</tr><tr><td class="sidebar-content">
<div class="sidebar-list mw-collapsible mw-collapsed"><div class="sidebar-list-title" style="text-align:center; background:lavender;"><a href="/wiki/Parliament_of_India" title="Parliament of India">Legislature</a></div><div class="sidebar-list-content mw-collapsible-content"><b><a href="/wiki/Parliament_of_India" title="Parliament of India">Parliament of India</a></b>
<p><a href="/wiki/File:Rajya_Sabha_June_2020.Svg" class="image"><img alt="Rajya Sabha June 2020.Svg" src="//upload.wikimedia.org/wikipedia/en/thumb/7/7b/Rajya_Sabha_June_2020.Svg/50px-Rajya_Sabha_June_2020.Svg.png" decoding="async" width="50" height="26" srcset="//upload.wikimedia.org/wikipedia/en/thumb/7/7b/Rajya_Sabha_June_2020.Svg/75px-Rajya_Sabha_June_2020.Svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/7/7b/Rajya_Sabha_June_2020.Svg/100px-Rajya_Sabha_June_2020.Svg.png 2x" data-file-width="360" data-file-height="185" /></a>
</p>
<hr />
<ul><li><b><a href="/wiki/Lok_Sabha" title="Lok Sabha">Lok Sabha</a></b></li></ul>
<ul><li><a href="/wiki/Speaker_of_the_Lok_Sabha" title="Speaker of the Lok Sabha">Speaker</a> <a href="/wiki/Om_Birla" title="Om Birla">Om Birla</a> (<a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">BJP</a>)</li></ul>
<ul><li><a href="/wiki/Leader_of_the_House_(Lok_Sabha)" class="mw-redirect" title="Leader of the House (Lok Sabha)">House Leader</a> <a href="/wiki/Narendra_Modi" title="Narendra Modi">Narendra Modi</a> (<a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">BJP</a>)</li></ul>
<ul><li><a href="/wiki/Leader_of_the_Opposition_(India)" title="Leader of the Opposition (India)">Opposition Leader</a> <i>Vacant</i></li></ul>
<hr />
<ul><li><b><a href="/wiki/Rajya_Sabha" title="Rajya Sabha">Rajya Sabha</a></b></li></ul>
<ul><li><a href="/wiki/Vice_President_of_India" title="Vice President of India">Chairman</a> <a href="/wiki/Venkaiah_Naidu" title="Venkaiah Naidu">Venkaiah Naidu</a> (<a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">BJP</a>)</li></ul>
<ul><li><a href="/wiki/Leader_of_the_House_(Rajya_Sabha)" class="mw-redirect" title="Leader of the House (Rajya Sabha)">House Leader</a> <a href="/wiki/Thawar_Chand_Gehlot" title="Thawar Chand Gehlot">Thawar Chand Gehlot</a> (<a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">BJP</a>)</li></ul>
<ul><li><a href="/wiki/Leader_of_the_Opposition_(India)" title="Leader of the Opposition (India)">Opposition Leader</a> <a href="/wiki/Mallikarjun_Kharge" title="Mallikarjun Kharge">Mallikarjun Kharge</a> (<a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC</a>)</li></ul></div></div></td>
</tr><tr><td class="sidebar-content">
<div class="sidebar-list mw-collapsible mw-collapsed"><div class="sidebar-list-title" style="text-align:center; background:lavender;"><a href="/wiki/Executive_(government)" title="Executive (government)">Executive</a></div><div class="sidebar-list-content mw-collapsible-content">
<ul><li><b><a href="/wiki/President_of_India" title="President of India">President of India</a></b></li></ul>
<ul><li><a href="/wiki/Ram_Nath_Kovind" title="Ram Nath Kovind">Ram Nath Kovind</a> (<a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">BJP</a>)</li></ul>
<ul><li><b><a href="/wiki/Vice_President_of_India" title="Vice President of India">Vice President of India</a></b></li></ul>
<ul><li><a href="/wiki/Venkaiah_Naidu" title="Venkaiah Naidu">Venkaiah Naidu</a> (<a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">BJP</a>)</li></ul>
<hr />
<ul><li><b><a href="/wiki/Prime_Minister_of_India" title="Prime Minister of India">Prime Minister of India</a></b></li></ul>
<ul><li><a href="/wiki/Narendra_Modi" title="Narendra Modi">Narendra Modi</a> (<a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">BJP</a>)</li></ul>
<hr />
<ul><li><a href="/wiki/Union_Council_of_Ministers" title="Union Council of Ministers">Union Council of Ministers</a></li>
<li><a href="/wiki/Cabinet_Secretary_of_India" class="mw-redirect" title="Cabinet Secretary of India">Cabinet secretary</a></li></ul></div></div></td>
</tr><tr><td class="sidebar-content">
<div class="sidebar-list mw-collapsible mw-collapsed"><div class="sidebar-list-title" style="text-align:center; background:lavender;"><a href="/wiki/Judiciary_of_India" title="Judiciary of India">Judiciary</a></div><div class="sidebar-list-content mw-collapsible-content">
<ul><li><b><a href="/wiki/Supreme_Court_of_India" title="Supreme Court of India">Supreme Court of India</a></b></li></ul>
<ul><li><a href="/wiki/Chief_Justice_of_India" title="Chief Justice of India">Chief Justice</a> <a href="/wiki/N._V._Ramana" title="N. V. Ramana">N. V. Ramana</a></li></ul>
<ul><li><a href="/wiki/List_of_sitting_judges_of_the_Supreme_Court_of_India" title="List of sitting judges of the Supreme Court of India">Justices</a> <span style="font-size:85%;"><a href="/wiki/List_of_sitting_judges_of_the_Supreme_Court_of_India" title="List of sitting judges of the Supreme Court of India">(List)</a></span></li></ul>
<hr />
<ul><li><a href="/wiki/District_Courts_of_India" class="mw-redirect" title="District Courts of India">District Courts</a></li>
<li><a href="/wiki/Tribunals_in_India" title="Tribunals in India">Other tribunals</a></li></ul></div></div></td>
</tr><tr><td class="sidebar-content">
<div class="sidebar-list mw-collapsible mw-collapsed"><div class="sidebar-list-title" style="text-align:center; background:lavender;"><a href="/wiki/Elections_in_India" title="Elections in India">Elections</a></div><div class="sidebar-list-content mw-collapsible-content">
<ul><li><b>General elections</b></li></ul>
<ul><li><a href="/wiki/2009_Indian_general_election" title="2009 Indian general election">2009</a></li>
<li><a href="/wiki/2014_Indian_general_election" title="2014 Indian general election">2014</a></li>
<li><a href="/wiki/2019_Indian_general_election" title="2019 Indian general election"><i>2019</i></a></li></ul>
<hr />
<ul><li><b><a href="/wiki/State_Assembly_elections_in_India" class="mw-redirect" title="State Assembly elections in India">State elections</a></b></li></ul>
<ul><li><a href="/wiki/2017_elections_in_India" title="2017 elections in India">2017</a></li>
<li><a href="/wiki/2018_elections_in_India" title="2018 elections in India">2018</a></li>
<li><a href="/wiki/2019_elections_in_India" title="2019 elections in India"><i>2019</i></a></li></ul></div></div></td>
</tr><tr><td class="sidebar-content">
<div class="sidebar-list mw-collapsible mw-collapsed"><div class="sidebar-list-title" style="text-align:center; background:lavender;"><a href="/wiki/List_of_political_parties_in_India" title="List of political parties in India">Political parties</a></div><div class="sidebar-list-content mw-collapsible-content"><b>National coalitions</b>
<ul><li><a href="/wiki/National_Democratic_Alliance" title="National Democratic Alliance">National Democratic Alliance</a> (NDA)</li></ul>
<ul><li><a href="/wiki/United_Progressive_Alliance" title="United Progressive Alliance">United Progressive Alliance</a> (UPA)</li></ul>
<hr />
<ul><li><a href="/wiki/List_of_political_parties_in_India#National_parties" title="List of political parties in India">National parties</a></li>
<li><a href="/wiki/List_of_political_parties_in_India#State_parties" title="List of political parties in India">State parties</a></li></ul></div></div></td>
</tr><tr><td class="sidebar-content">
<div class="sidebar-list mw-collapsible mw-collapsed"><div class="sidebar-list-title" style="text-align:center; background:lavender;"><a href="/wiki/Federalism_in_India" title="Federalism in India">Federalism</a></div><div class="sidebar-list-content mw-collapsible-content">
<ul><li><b><a href="/wiki/State_governments_of_India" title="State governments of India">State Government</a></b></li></ul>
<hr />
<ul><li><a href="/wiki/List_of_current_Indian_governors" title="List of current Indian governors">Governors</a></li></ul>
<ul><li><a href="/wiki/State_governments_of_India#Legislative_Assembly" title="State governments of India">Legislatures</a> (<a href="/wiki/State_governments_of_India#Legislative_Assembly#Membership_and_terms_of_office" title="State governments of India">List</a>)</li></ul>
<ul><li><a href="/wiki/High_courts_of_India" title="High courts of India">High Courts</a></li></ul>
<ul><li><ul><li><a href="/wiki/Local_government_in_India" title="Local government in India">Local government</a></li></ul></li></ul></div></div></td>
</tr><tr><td class="sidebar-below">
<a href="/wiki/File:Flag_of_India.svg" class="image"><img alt="Flag of India.svg" src="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/32px-Flag_of_India.svg.png" decoding="async" width="32" height="21" class="noviewer" srcset="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/48px-Flag_of_India.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/64px-Flag_of_India.svg.png 2x" data-file-width="1350" data-file-height="900" /></a>&#160;<a href="/wiki/Portal:India" title="Portal:India">India&#32;portal</a> <div class="hlist hlist-separated" style="text-align:center;font-weight:normal;">
<ul><li><a href="/wiki/Outline_of_political_science#Politics_by_region" title="Outline of political science">Other countries</a></li></ul>
</div></td></tr><tr><td class="sidebar-navbar"><style data-mw-deduplicate="TemplateStyles:r1063604349">.mw-parser-output .navbar{display:inline;font-size:88%;font-weight:normal}.mw-parser-output .navbar-collapse{float:left;text-align:left}.mw-parser-output .navbar-boxtext{word-spacing:0}.mw-parser-output .navbar ul{display:inline-block;white-space:nowrap;line-height:inherit}.mw-parser-output .navbar-brackets::before{margin-right:-0.125em;content:"[ "}.mw-parser-output .navbar-brackets::after{margin-left:-0.125em;content:" ]"}.mw-parser-output .navbar li{word-spacing:-0.125em}.mw-parser-output .navbar a>span,.mw-parser-output .navbar a>abbr{text-decoration:inherit}.mw-parser-output .navbar-mini abbr{font-variant:small-caps;border-bottom:none;text-decoration:none;cursor:inherit}.mw-parser-output .navbar-ct-full{font-size:114%;margin:0 7em}.mw-parser-output .navbar-ct-mini{font-size:114%;margin:0 4em}</style><div class="navbar plainlinks hlist navbar-mini"><ul><li class="nv-view"><a href="/wiki/Template:Politics_of_India" title="Template:Politics of India"><abbr title="View this template">v</abbr></a></li><li class="nv-talk"><a href="/wiki/Template_talk:Politics_of_India" title="Template talk:Politics of India"><abbr title="Discuss this template">t</abbr></a></li><li class="nv-edit"><a class="external text" href="https://en.wikipedia.org/w/index.php?title=Template:Politics_of_India&amp;action=edit"><abbr title="Edit this template">e</abbr></a></li></ul></div></td></tr></tbody></table>
<p>The <b>prime minister of India</b>  is the legislative <a href="/wiki/Head_of_government" title="Head of government">head of government</a> of the <a href="/wiki/Republic_of_India" class="mw-redirect" title="Republic of India">Republic of India</a> and the leader of the <a href="/wiki/Union_Council_of_Ministers" title="Union Council of Ministers">Union Council of Ministers</a>. They can be a member of either of the two houses of the <a href="/wiki/Parliament_of_India" title="Parliament of India">Indian parliament</a>—the lower, or <a href="/wiki/Lok_Sabha" title="Lok Sabha">Lok Sabha</a> or the upper, the <a href="/wiki/Rajya_Sabha" title="Rajya Sabha">Rajya Sabha</a>. The prime minister ranks third in the <a href="/wiki/Indian_order_of_precedence" class="mw-redirect" title="Indian order of precedence">order of precedence</a>. 
</p><p>The prime minister is appointed by the <a href="/wiki/President_of_India" title="President of India">President of India</a>; however the prime minister has to enjoy the confidence of the majority of <a href="/wiki/Lok_Sabha" title="Lok Sabha">Lok Sabha</a> members, who are <a href="/wiki/Direct_election" title="Direct election">directly elected</a> <a href="/wiki/Elections_in_India#Parliamentary_general_elections_(Lok_Sabha)" title="Elections in India">every five years</a>, unless a prime minister resigns. The prime minister is the presiding member of the <a href="/wiki/Union_Council_of_Ministers" title="Union Council of Ministers">Council of Ministers</a> of the <a href="/wiki/Government_of_India" title="Government of India">Union government</a>. The prime minister unilaterally controls the selection and dismissal of members of the Council; and allocation of posts to members within the government. This Council, which is collectively responsible to the <a href="/wiki/Lok_Sabha" title="Lok Sabha">Lok Sabha</a> as per <a href="https://en.wikisource.org/wiki/Constitution_of_India/Part_V" class="extiw" title="s:Constitution of India/Part V">Article 75(3)</a>, assists the President regarding the operations under the latter's powers; however, by the virtue of <a href="https://en.wikisource.org/wiki/Constitution_of_India/Part_V" class="extiw" title="s:Constitution of India/Part V">Article 74</a> of the <a href="/wiki/Constitution_of_India" title="Constitution of India">Constitution</a>, such '<a href="/wiki/Advice_(constitutional_law)" title="Advice (constitutional law)">aid and advice</a>' tendered by the Council is <a href="https://en.wiktionary.org/wiki/binding#Adjective" class="extiw" title="wiktionary:binding">binding</a>.
</p><p>Since 1947 India has had 15 prime ministers.<sup id="cite_ref-2" class="reference"><a href="#cite_note-2">&#91;a&#93;</a></sup>  <a href="/wiki/Jawaharlal_Nehru" title="Jawaharlal Nehru">Jawaharlal Nehru</a> was the India's first prime minister, serving as prime minister of the <a href="/wiki/Dominion_of_India" title="Dominion of India">Dominion of India</a> from 15 August 1947 until 26 January 1950, and thereafter of the <a href="/wiki/Republic_of_India" class="mw-redirect" title="Republic of India">Republic of India</a> until his death in May 1964. (India conducted its first general elections in 1952. Earlier, Nehru had served as prime minister of the <a href="/wiki/Interim_Government_of_India" title="Interim Government of India">Interim Government of India</a> in the <a href="/wiki/British_Raj" title="British Raj">British Raj</a> from 2 September 1946 until 14 August 1947, his party, the Indian National Congress having won the <a href="/wiki/1946_Indian_provincial_elections" title="1946 Indian provincial elections">1946 Indian provincial elections</a>.) Nehru was succeeded by <a href="/wiki/Lal_Bahadur_Shastri" title="Lal Bahadur Shastri">Lal Bahadur Shastri</a>, whose 1 year 7-month term ended in his death in <a href="/wiki/Tashkent" title="Tashkent">Tashkent</a>, then in the <a href="/wiki/USSR" class="mw-redirect" title="USSR">USSR</a>, where he had signed the <a href="/wiki/Tashkent_Declaration" title="Tashkent Declaration">Tashkent Declaration</a> between India and Pakistan.<sup id="cite_ref-3" class="reference"><a href="#cite_note-3">&#91;2&#93;</a></sup> <a href="/wiki/Indira_Gandhi" title="Indira Gandhi">Indira Gandhi</a>, Nehru's daughter, succeeded Shastri in 1966 to become the country's first female prime minister.<sup id="cite_ref-indira_4-0" class="reference"><a href="#cite_note-indira-4">&#91;3&#93;</a></sup> Eleven years later, her party the <a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress</a> lost the <a href="/wiki/1977_Indian_general_election" title="1977 Indian general election">1977 Indian general election</a> to the <a href="/wiki/Janata_Party" title="Janata Party">Janata Party</a>, whose leader <a href="/wiki/Morarji_Desai" title="Morarji Desai">Morarji Desai</a> became the first non-Congress prime minister.<sup id="cite_ref-5" class="reference"><a href="#cite_note-5">&#91;4&#93;</a></sup> After Desai resigned in 1979, his former associate <a href="/wiki/Charan_Singh" title="Charan Singh">Charan Singh</a> briefly held office until the Congress won the <a href="/wiki/1980_Indian_general_election" title="1980 Indian general election">1980 Indian general election</a> and Indira Gandhi returned as prime minister.<sup id="cite_ref-6" class="reference"><a href="#cite_note-6">&#91;5&#93;</a></sup> Her second term as prime minister ended five years later on 31 October 1984, when she was <a href="/wiki/Assassination_of_Indira_Gandhi" title="Assassination of Indira Gandhi">assassinated</a> by her bodyguards.<sup id="cite_ref-indira_4-1" class="reference"><a href="#cite_note-indira-4">&#91;3&#93;</a></sup> Her son <a href="/wiki/Rajiv_Gandhi" title="Rajiv Gandhi">Rajiv Gandhi</a> was sworn in as India's youngest premier. Members of <a href="/wiki/Nehru%E2%80%93Gandhi_family" title="Nehru–Gandhi family">Nehru–Gandhi family</a> have been prime minister for approximately 38 years.<sup id="cite_ref-7" class="reference"><a href="#cite_note-7">&#91;6&#93;</a></sup>
</p><p>After a general election loss, Rajiv Gandhi's five-year term ended; his former cabinet colleague, <a href="/wiki/Vishwanath_Pratap_Singh" title="Vishwanath Pratap Singh">Vishwanath Pratap Singh</a> of the <a href="/wiki/Janata_Dal" title="Janata Dal">Janata Dal</a>, formed the year-long <a href="/wiki/National_Front_(India)" class="mw-redirect" title="National Front (India)">National Front</a> coalition government in 1989. A seven-month interlude under prime minister <a href="/wiki/Chandra_Shekhar" title="Chandra Shekhar">Chandra Shekhar</a> followed, after which the Congress party returned to power, forming the government under <a href="/wiki/P._V._Narasimha_Rao" title="P. V. Narasimha Rao">P. V. Narasimha Rao</a> in June 1991, Rajiv Gandhi having been assassinated earlier that year.<sup id="cite_ref-former_8-0" class="reference"><a href="#cite_note-former-8">&#91;7&#93;</a></sup> Rao's five-year term was succeeded by four short-lived governments—<a href="/wiki/Atal_Bihari_Vajpayee" title="Atal Bihari Vajpayee">Atal Bihari Vajpayee</a> from the <a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">Bharatiya Janata Party</a> (BJP) for 13 days in 1996, a year each under <a href="/wiki/United_Front_(India)" title="United Front (India)">United Front</a> prime ministers <a href="/wiki/H._D._Deve_Gowda" title="H. D. Deve Gowda">H. D. Deve Gowda</a> and <a href="/wiki/Inder_Kumar_Gujral" title="Inder Kumar Gujral">Inder Kumar Gujral</a>, and Vajpayee again for 19 months in 1998–99.<sup id="cite_ref-former_8-1" class="reference"><a href="#cite_note-former-8">&#91;7&#93;</a></sup> In 1998, Vajpayee's <a href="/wiki/National_Democratic_Alliance" title="National Democratic Alliance">National Democratic Alliance</a> (NDA) won the general election, the first non-Congress alliance to do so, and he served a full five-year term as prime minister.<sup id="cite_ref-9" class="reference"><a href="#cite_note-9">&#91;8&#93;</a></sup> The Congress, and its <a href="/wiki/United_Progressive_Alliance" title="United Progressive Alliance">United Progressive Alliance</a> won the general elections in 2004 and 2009, <a href="/wiki/Manmohan_Singh" title="Manmohan Singh">Manmohan Singh</a> serving as prime minister between 2004 and 2014.<sup id="cite_ref-10" class="reference"><a href="#cite_note-10">&#91;9&#93;</a></sup> The BJP won the <a href="/wiki/2014_Indian_general_election" title="2014 Indian general election">2014 Indian general election</a>, and its parliamentary leader <a href="/wiki/Narendra_Modi" title="Narendra Modi">Narendra Modi</a> formed the first non-Congress single party majority government. Modi has served as prime minister since, his party winning the <a href="/wiki/2014_Indian_general_election" title="2014 Indian general election">2014 Indian general election</a>.<sup id="cite_ref-11" class="reference"><a href="#cite_note-11">&#91;10&#93;</a></sup>
</p>
<div id="toc" class="toc" role="navigation" aria-labelledby="mw-toc-heading"><input type="checkbox" role="button" id="toctogglecheckbox" class="toctogglecheckbox" style="display:none" /><div class="toctitle" lang="en" dir="ltr"><h2 id="mw-toc-heading">Contents</h2><span class="toctogglespan"><label class="toctogglelabel" for="toctogglecheckbox"></label></span></div>
<ul>
<li class="toclevel-1 tocsection-1"><a href="#List"><span class="tocnumber">1</span> <span class="toctext">List</span></a></li>
<li class="toclevel-1 tocsection-2"><a href="#Statistics"><span class="tocnumber">2</span> <span class="toctext">Statistics</span></a></li>
<li class="toclevel-1 tocsection-3"><a href="#See_also"><span class="tocnumber">3</span> <span class="toctext">See also</span></a></li>
<li class="toclevel-1 tocsection-4"><a href="#Footnotes"><span class="tocnumber">4</span> <span class="toctext">Footnotes</span></a></li>
<li class="toclevel-1 tocsection-5"><a href="#Notes"><span class="tocnumber">5</span> <span class="toctext">Notes</span></a></li>
<li class="toclevel-1 tocsection-6"><a href="#References"><span class="tocnumber">6</span> <span class="toctext">References</span></a></li>
<li class="toclevel-1 tocsection-7"><a href="#External_links"><span class="tocnumber">7</span> <span class="toctext">External links</span></a></li>
</ul>
</div>

<h2><span class="mw-headline" id="List">List</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit&amp;section=1" title="Edit section: List">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<div>
<table class="multicol" role="presentation" style="border-collapse: collapse; padding: 0; border: 0; background:transparent; width:100%;">

<tbody><tr>
<td style="width: 50%;text-align: left; vertical-align: top;">
<dl><dt>Key</dt></dl>
<ul><li>No.: Incumbent number</li>
<li><style data-mw-deduplicate="TemplateStyles:r1041539562">.mw-parser-output .citation{word-wrap:break-word}.mw-parser-output .citation:target{background-color:rgba(0,127,255,0.133)}</style><span class="citation wikicite" id="endnote_††"><a href="#ref_††"><b><sup>†</sup></b></a></span>  Assassinated or died in office</li>
<li><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1041539562"/><span class="citation wikicite" id="endnote_§§"><a href="#ref_§§"><b><sup>§</sup></b></a></span>  Returned to office after a previous non-consecutive term</li>
<li><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1041539562"/><span class="citation wikicite" id="endnote_RESRES"><a href="#ref_RESRES"><b><sup>RES</sup></b></a></span>  Resigned</li>
<li><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1041539562"/><span class="citation wikicite" id="endnote_NCNC"><a href="#ref_NCNC"><b><sup>NC</sup></b></a></span>  Resigned following a <a href="/wiki/No-confidence_motion" class="mw-redirect" title="No-confidence motion">no-confidence motion</a></li></ul>
<p>&#32;
</p>
</td></tr></tbody></table></div>
<dl><dt>Legend</dt></dl>
<ul><li><style data-mw-deduplicate="TemplateStyles:r981673959">.mw-parser-output .legend{page-break-inside:avoid;break-inside:avoid-column}.mw-parser-output .legend-color{display:inline-block;min-width:1.25em;height:1.25em;line-height:1.25;margin:1px 0;text-align:center;border:1px solid black;background-color:transparent;color:black}.mw-parser-output .legend-text{}</style><div class="legend"><span class="legend-color" style="border: 1px solid #AAAAAA;background-color:wheat; color:black;">&#160;</span>&#160;<a href="/wiki/Interim_Prime_Minister" class="mw-redirect" title="Interim Prime Minister">Interim Prime Minister</a></div></li></ul>
<table class="wikitable sortable" id="pm_list" style="width:100%; text-align:center">
<tbody><tr>
<th colspan="12"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r981673959"/><span class="legend-color" style="background-color:#FF9933; color:black;border:1px solid black;">&#160;</span> <a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">BJP</a>&#160;&#160; (2)<sup id="cite_ref-12" class="reference"><a href="#cite_note-12">&#91;b&#93;</a></sup><span style="padding-left:1.0em;">&#160;</span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r981673959"/><span class="legend-color" style="background-color:#19AAED; color:black;border:1px solid black;">&#160;</span> <a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC/INC(I)/INC(R)</a>&#160;&#160;<sup id="cite_ref-14" class="reference"><a href="#cite_note-14">&#91;c&#93;</a></sup> (6+1 acting<sup id="cite_ref-15" class="reference"><a href="#cite_note-15">&#91;d&#93;</a></sup>)<span style="padding-left:1.0em;">&#160;</span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r981673959"/><span class="legend-color" style="background-color:#2E8B57; color:black;border:1px solid black;">&#160;</span> <a href="/wiki/Janata_Dal" title="Janata Dal">JD</a>&#160;&#160; (3)<span style="padding-left:1.0em;">&#160;</span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r981673959"/><span class="legend-color" style="background-color:#1F75FE; color:black;border:1px solid black;">&#160;</span> <a href="/wiki/Janata_Party" title="Janata Party">JP</a>&#160;&#160; (1)<span style="padding-left:1.0em;">&#160;</span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r981673959"/><span class="legend-color" style="background-color:#000080; color:white;border:1px solid black;">&#160;</span> <a href="/wiki/Janata_Party_(Secular)" title="Janata Party (Secular)">JP(S)</a>&#160;&#160; (1)<span style="padding-left:1.0em;">&#160;</span><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r981673959"/><span class="legend-color" style="background-color:#74C365; color:black;border:1px solid black;">&#160;</span> <a href="/wiki/Samajwadi_Janata_Party_(Rashtriya)" title="Samajwadi Janata Party (Rashtriya)">SJP(R)</a>&#160;&#160; (1)
</th></tr>
<tr>
<th colspan="2" rowspan="2" width="5px">No
</th>
<th colspan="1" rowspan="2" width="100px">Portrait
</th>
<th rowspan="2">Name
<p><small><span class="nowrap">(Birth–Death)</span></small> 
</p>
</th>
<th rowspan="2">Party
</th>
<th colspan="2">Term of office<sup id="cite_ref-16" class="reference"><a href="#cite_note-16">&#91;12&#93;</a></sup>
</th>
<th rowspan="2">Time in office
</th>
<th rowspan="2">Lok Sabha<sup id="cite_ref-17" class="reference"><a href="#cite_note-17">&#91;e&#93;</a></sup>
</th>
<th rowspan="2">Ministry
</th>
<th rowspan="2" data-sort-type="number">Appointed by
</th></tr>
<tr>
<th>Took office
</th>
<th>Left office
</th></tr>
<tr>
<th rowspan="4">1
</th>
<td rowspan="4" style="background:#19AAED;">
</td>
<td rowspan="4"><a href="/wiki/File:Nehru.jpg" class="image"><img alt="Jawaharlal Nehru" src="//upload.wikimedia.org/wikipedia/commons/thumb/2/2f/Nehru.jpg/109px-Nehru.jpg" decoding="async" width="109" height="157" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/2/2f/Nehru.jpg/164px-Nehru.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/2/2f/Nehru.jpg 2x" data-file-width="187" data-file-height="269" /></a>
</td>
<td rowspan="4"><b><span data-sort-value="Nehru, Jawaharlal"><span class="vcard"><span class="fn"><a href="/wiki/Jawaharlal_Nehru" title="Jawaharlal Nehru">Jawaharlal Nehru</a></span></span></span></b><br /><small>(1889–<a href="/wiki/Death_and_state_funeral_of_Jawaharlal_Nehru" title="Death and state funeral of Jawaharlal Nehru">1964</a>)</small><br />
</td>
<td rowspan="4"><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress</a>
</td>
<td>15 August 1947
</td>
<td>15 April 1952
</td>
<td rowspan="4">16&#160;years, 286&#160;days
</td>
<td data-sort-value="–"><a href="/wiki/Constituent_Assembly_of_India" title="Constituent Assembly of India">Constituent Assembly</a><sup id="cite_ref-18" class="reference"><a href="#cite_note-18">&#91;f&#93;</a></sup>
</td>
<td><a href="/wiki/First_Nehru_ministry" title="First Nehru ministry">Nehru I</a>
</td>
<td><a href="/wiki/Louis_Mountbatten,_1st_Earl_Mountbatten_of_Burma" title="Louis Mountbatten, 1st Earl Mountbatten of Burma">The Earl Mountbatten of Burma</a>
</td></tr>
<tr>
<td>15 April 1952
</td>
<td>17 April 1957
</td>
<td data-sort-value="1"><a href="/wiki/1st_Lok_Sabha" title="1st Lok Sabha">1st</a>
</td>
<td><a href="/wiki/Second_Nehru_ministry" title="Second Nehru ministry">Nehru II</a>
</td>
<td rowspan="3"><span data-sort-value="Prasad, Rajendra"><span class="vcard"><span class="fn"><a href="/wiki/Rajendra_Prasad" title="Rajendra Prasad">Rajendra Prasad</a></span></span></span>
</td></tr>
<tr>
<td>17 April 1957
</td>
<td>2 April 1962
</td>
<td data-sort-value="2"><a href="/wiki/2nd_Lok_Sabha" title="2nd Lok Sabha">2nd</a>
</td>
<td><a href="/wiki/Third_Nehru_ministry" title="Third Nehru ministry">Nehru III</a>
</td></tr>
<tr>
<td>2 April 1962
</td>
<td>27 May 1964<small>†</small>
</td>
<td rowspan="5" data-sort-value="3"><a href="/wiki/3rd_Lok_Sabha" title="3rd Lok Sabha">3rd</a>
</td>
<td><a href="/wiki/Fourth_Nehru_ministry" title="Fourth Nehru ministry">Nehru IV</a>
</td></tr>
<tr>
<th style="background-color:Wheat">-
</th>
<td style="background:#19AAED;">
</td>
<td style="background:wheat;"><a href="/wiki/File:Gulzarilal_Nanda_1.jpg" class="image"><img alt="Gulzarilal Nanda 1.jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/f/f7/Gulzarilal_Nanda_1.jpg/100px-Gulzarilal_Nanda_1.jpg" decoding="async" width="100" height="159" srcset="//upload.wikimedia.org/wikipedia/commons/f/f7/Gulzarilal_Nanda_1.jpg 1.5x" data-file-width="148" data-file-height="235" /></a>
</td>
<td style="background:wheat;"><b><span data-sort-value="Nanda, Gulzarilal"><span class="vcard"><span class="fn"><a href="/wiki/Gulzarilal_Nanda" title="Gulzarilal Nanda">Gulzarilal Nanda</a></span></span></span></b><br /><small><span class="nowrap">(1898–1998)</span></small>
</td>
<td style="background:wheat;"><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress</a>
</td>
<td style="background:wheat;">27 May 1964
</td>
<td style="background:wheat;">9 June 1964
</td>
<td style="background:wheat;">13 days
</td>
<td style="background:wheat;"><a href="/wiki/First_Nanda_ministry" title="First Nanda ministry">Nanda I</a>
</td>
<td rowspan="5"><span data-sort-value="Radhakrishnan, Sarvepalli"><span class="vcard"><span class="fn"><a href="/wiki/Sarvepalli_Radhakrishnan" title="Sarvepalli Radhakrishnan">Sarvepalli Radhakrishnan</a></span></span></span>
</td></tr>
<tr>
<th><b><a href="/wiki/Premiership_of_Lal_Bahadur_Shastri" title="Premiership of Lal Bahadur Shastri">2</a></b>
</th>
<td style="background:#19AAED;">
</td>
<td><a href="/wiki/File:LB_Shastri.jpg" class="image"><img alt="LB Shastri.jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/c/c2/LB_Shastri.jpg/100px-LB_Shastri.jpg" decoding="async" width="100" height="134" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/c/c2/LB_Shastri.jpg/150px-LB_Shastri.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/c/c2/LB_Shastri.jpg/200px-LB_Shastri.jpg 2x" data-file-width="722" data-file-height="964" /></a>
</td>
<td><b><span data-sort-value="Shastri, Lal Bahadur"><span class="vcard"><span class="fn"><a href="/wiki/Lal_Bahadur_Shastri" title="Lal Bahadur Shastri">Lal Bahadur Shastri</a></span></span></span></b><br /><small>(1904–1966)</small><br />
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress</a>
</td>
<td>9 June 1964
</td>
<td>11 January 1966<small>†</small>
</td>
<td>1&#160;year, 216&#160;days
</td>
<td><a href="/wiki/Shastri_ministry" title="Shastri ministry">Shastri</a>
</td></tr>
<tr>
<th style="background-color:Wheat">-
</th>
<td style="background:#19AAED;">
</td>
<td style="background:wheat;"><a href="/wiki/File:Gulzarilal_Nanda_1.jpg" class="image"><img alt="Gulzarilal Nanda 1.jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/f/f7/Gulzarilal_Nanda_1.jpg/100px-Gulzarilal_Nanda_1.jpg" decoding="async" width="100" height="159" srcset="//upload.wikimedia.org/wikipedia/commons/f/f7/Gulzarilal_Nanda_1.jpg 1.5x" data-file-width="148" data-file-height="235" /></a>
</td>
<td style="background:wheat;"><b><span data-sort-value="Nanda, Gulzarilal"><span class="vcard"><span class="fn"><a href="/wiki/Gulzarilal_Nanda" title="Gulzarilal Nanda">Gulzarilal Nanda</a></span></span></span></b><br /><small>(1898–1998)</small><br />
</td>
<td style="background:wheat;"><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress</a>
</td>
<td style="background:wheat;">11 January 1966
</td>
<td style="background:wheat;">24 January 1966
</td>
<td style="background:wheat;">13 days
</td>
<td style="background:wheat;"><a href="/wiki/Second_Nanda_ministry" title="Second Nanda ministry">Nanda II</a>
</td></tr>
<tr>
<th rowspan="3">3
</th>
<td rowspan="3" style="background:#19AAED;">
</td>
<td rowspan="3"><a href="/wiki/File:Indira_Gandhi_1974_(cropped).jpg" class="image"><img alt="Indira Gandhi 1974 (cropped).jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/4/4e/Indira_Gandhi_1974_%28cropped%29.jpg/100px-Indira_Gandhi_1974_%28cropped%29.jpg" decoding="async" width="100" height="129" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/4/4e/Indira_Gandhi_1974_%28cropped%29.jpg/150px-Indira_Gandhi_1974_%28cropped%29.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/4/4e/Indira_Gandhi_1974_%28cropped%29.jpg/199px-Indira_Gandhi_1974_%28cropped%29.jpg 2x" data-file-width="488" data-file-height="631" /></a>
</td>
<td rowspan="3"><b><span data-sort-value="Gandhi, Indira"><span class="vcard"><span class="fn"><a href="/wiki/Indira_Gandhi" title="Indira Gandhi">Indira Gandhi</a></span></span></span></b><br /><small>(1917–<a href="/wiki/Assassination_of_Indira_Gandhi" title="Assassination of Indira Gandhi">1984</a>)</small><br />
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress</a>
</td>
<td>24 January 1966
</td>
<td>4 March 1967
</td>
<td rowspan="3">11&#160;years, 59&#160;days
</td>
<td rowspan="2"><a href="/wiki/First_Indira_Gandhi_ministry" title="First Indira Gandhi ministry">Indira I</a>
</td></tr>
<tr>
<td rowspan="2"><a href="/wiki/Indian_National_Congress_(R)" title="Indian National Congress (R)">Indian National Congress (R)</a>
</td>
<td>4 March 1967
</td>
<td>15 March 1971
</td>
<td data-sort-value="4"><a href="/wiki/4th_Lok_Sabha" title="4th Lok Sabha">4th</a>
</td></tr>
<tr>
<td>15 March 1971
</td>
<td>24 March 1977
</td>
<td data-sort-value="5"><a href="/wiki/5th_Lok_Sabha" title="5th Lok Sabha">5th</a>
</td>
<td><a href="/wiki/Second_Indira_Gandhi_ministry" title="Second Indira Gandhi ministry">Indira II</a>
</td>
<td rowspan="1"><span data-sort-value="Giri, V. V."><span class="vcard"><span class="fn"><a href="/wiki/V._V._Giri" title="V. V. Giri">V. V. Giri</a></span></span></span>
</td></tr>
<tr>
<th><a href="/wiki/Premiership_of_Morarji_Desai" title="Premiership of Morarji Desai">4</a>
</th>
<td style="background:#1F75FE;">
</td>
<td><a href="/wiki/File:Morarji_Desai_visits_the_USA_(cropped).jpg" class="image"><img alt="Morarji Desai visits the USA (cropped).jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/e/e2/Morarji_Desai_visits_the_USA_%28cropped%29.jpg/100px-Morarji_Desai_visits_the_USA_%28cropped%29.jpg" decoding="async" width="100" height="128" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/e/e2/Morarji_Desai_visits_the_USA_%28cropped%29.jpg/150px-Morarji_Desai_visits_the_USA_%28cropped%29.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/e/e2/Morarji_Desai_visits_the_USA_%28cropped%29.jpg/200px-Morarji_Desai_visits_the_USA_%28cropped%29.jpg 2x" data-file-width="1210" data-file-height="1553" /></a>
</td>
<td><b><span data-sort-value="Desai, Morarji"><span class="vcard"><span class="fn"><a href="/wiki/Morarji_Desai" title="Morarji Desai">Morarji Desai</a></span></span></span></b><br /><small>(1896–1995)</small><br />
</td>
<td><a href="/wiki/Janata_Party" title="Janata Party">Janata Party</a>
</td>
<td>24 March 1977
</td>
<td>28 July 1979<sup class="reference" id="ref_RESRES"><a href="#endnote_RESRES">[RES]</a></sup>
</td>
<td>2&#160;years, 126&#160;days
</td>
<td rowspan="2" data-sort-value="6"><a href="/wiki/6th_Lok_Sabha" title="6th Lok Sabha">6th</a>
</td>
<td><a href="/wiki/Morarji_Desai_ministry" class="mw-redirect" title="Morarji Desai ministry">Desai</a>
</td>
<td><span data-sort-value="Jatti, B. D."><span class="vcard"><span class="fn"><a href="/wiki/B._D._Jatti" title="B. D. Jatti">B. D. Jatti</a></span></span></span><br /><small><i>(acting)</i></small>
</td></tr>
<tr>
<th>5
</th>
<td style="background:#000080;">
</td>
<td><a href="/wiki/File:Charan_Singh.jpg" class="image"><img alt="Charan Singh.jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/1/15/Charan_Singh.jpg/100px-Charan_Singh.jpg" decoding="async" width="100" height="136" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/1/15/Charan_Singh.jpg/150px-Charan_Singh.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/1/15/Charan_Singh.jpg/200px-Charan_Singh.jpg 2x" data-file-width="553" data-file-height="753" /></a>
</td>
<td><b><span data-sort-value="Singh, Chaudhary Charan"><span class="vcard"><span class="fn"><a href="/wiki/Chaudhary_Charan_Singh" class="mw-redirect" title="Chaudhary Charan Singh">Chaudhary Charan Singh</a></span></span></span></b><br /><small>(1902–1987)</small><br />
</td>
<td><a href="/wiki/Janata_Party_(Secular)" title="Janata Party (Secular)">Janata Party (Secular)</a>
</td>
<td>28 July 1979
</td>
<td>14 January 1980<sup class="reference" id="ref_RESRES"><a href="#endnote_RESRES">[RES]</a></sup>
</td>
<td>170 days
</td>
<td><a href="/wiki/Charan_Singh_ministry" title="Charan Singh ministry">Charan</a>
</td>
<td rowspan="2"><span data-sort-value="Reddy, Neelam Sanjiva"><span class="vcard"><span class="fn"><a href="/wiki/Neelam_Sanjiva_Reddy" title="Neelam Sanjiva Reddy">Neelam Sanjiva Reddy</a></span></span></span>
</td></tr>
<tr>
<th>(3)
</th>
<td style="background:#19AAED;">
</td>
<td><a href="/wiki/File:Indira_Gandhi_2017_stamp.png" class="image"><img alt="Indira Gandhi 2017 stamp.png" src="//upload.wikimedia.org/wikipedia/commons/thumb/d/dc/Indira_Gandhi_2017_stamp.png/100px-Indira_Gandhi_2017_stamp.png" decoding="async" width="100" height="121" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/d/dc/Indira_Gandhi_2017_stamp.png/150px-Indira_Gandhi_2017_stamp.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/d/dc/Indira_Gandhi_2017_stamp.png/199px-Indira_Gandhi_2017_stamp.png 2x" data-file-width="1177" data-file-height="1430" /></a>
</td>
<td><b><span data-sort-value="Gandhi, Indira"><span class="vcard"><span class="fn"><a href="/wiki/Indira_Gandhi" title="Indira Gandhi">Indira Gandhi</a></span></span></span></b><br /><small>(1917–<a href="/wiki/Assassination_of_Indira_Gandhi" title="Assassination of Indira Gandhi">1984</a>)</small><br />
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress (I)</a>
</td>
<td>14 January 1980<sup class="reference" id="ref_§§"><a href="#endnote_§§">[§]</a></sup>
</td>
<td>31 October 1984<small>†</small>
</td>
<td>4&#160;years, 291&#160;days
</td>
<td rowspan="2" data-sort-value="7"><a href="/wiki/7th_Lok_Sabha" title="7th Lok Sabha">7th</a>
</td>
<td><a href="/wiki/Third_Indira_Gandhi_ministry" title="Third Indira Gandhi ministry">Indira III</a>
</td></tr>
<tr>
<th rowspan="2">6
</th>
<td rowspan="2" style="background:#19AAED;">
</td>
<td rowspan="2"><a href="/wiki/File:Rajiv_Gandhi_(1987).jpg" class="image"><img alt="Rajiv Gandhi (1987).jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/0/02/Rajiv_Gandhi_%281987%29.jpg/100px-Rajiv_Gandhi_%281987%29.jpg" decoding="async" width="100" height="133" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/0/02/Rajiv_Gandhi_%281987%29.jpg/150px-Rajiv_Gandhi_%281987%29.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/0/02/Rajiv_Gandhi_%281987%29.jpg/200px-Rajiv_Gandhi_%281987%29.jpg 2x" data-file-width="755" data-file-height="1006" /></a>
</td>
<td rowspan="2"><b><span data-sort-value="Gandhi, Rajiv"><span class="vcard"><span class="fn"><a href="/wiki/Rajiv_Gandhi" title="Rajiv Gandhi">Rajiv Gandhi</a></span></span></span></b><br /><small>(1944–<a href="/wiki/Assassination_of_Rajiv_Gandhi" title="Assassination of Rajiv Gandhi">1991</a>)</small><br />
</td>
<td rowspan="2"><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress (I)</a>
</td>
<td>31 October 1984
</td>
<td>31 December 1984
</td>
<td rowspan="2">5&#160;years, 32&#160;days
</td>
<td rowspan="2"><a href="/wiki/Rajiv_Gandhi_ministry" title="Rajiv Gandhi ministry">Rajiv</a>
</td>
<td rowspan="2"><span data-sort-value="Singh, Zail"><span class="vcard"><span class="fn"><a href="/wiki/Zail_Singh" title="Zail Singh">Zail Singh</a></span></span></span>
</td></tr>
<tr>
<td>31 December 1984
</td>
<td>2 December 1989
</td>
<td data-sort-value="8"><a href="/wiki/8th_Lok_Sabha" title="8th Lok Sabha">8th</a>
</td></tr>
<tr>
<th><a href="/wiki/Premiership_of_Vishwanath_Pratap_Singh" title="Premiership of Vishwanath Pratap Singh">7</a>
</th>
<td style="background:#2E8B57;">
</td>
<td><a href="/wiki/File:Visit_of_Vishwanath_Pratap_Sing,_Indian_Minister_for_Trade,_to_the_CEC_(cropped).jpg" class="image"><img alt="Visit of Vishwanath Pratap Sing, Indian Minister for Trade, to the CEC (cropped).jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/f/fc/Visit_of_Vishwanath_Pratap_Sing%2C_Indian_Minister_for_Trade%2C_to_the_CEC_%28cropped%29.jpg/100px-Visit_of_Vishwanath_Pratap_Sing%2C_Indian_Minister_for_Trade%2C_to_the_CEC_%28cropped%29.jpg" decoding="async" width="100" height="134" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/f/fc/Visit_of_Vishwanath_Pratap_Sing%2C_Indian_Minister_for_Trade%2C_to_the_CEC_%28cropped%29.jpg/150px-Visit_of_Vishwanath_Pratap_Sing%2C_Indian_Minister_for_Trade%2C_to_the_CEC_%28cropped%29.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/f/fc/Visit_of_Vishwanath_Pratap_Sing%2C_Indian_Minister_for_Trade%2C_to_the_CEC_%28cropped%29.jpg/200px-Visit_of_Vishwanath_Pratap_Sing%2C_Indian_Minister_for_Trade%2C_to_the_CEC_%28cropped%29.jpg 2x" data-file-width="785" data-file-height="1052" /></a>
</td>
<td><b><span data-sort-value="Singh, Vishwanath Pratap"><span class="vcard"><span class="fn"><a href="/wiki/Vishwanath_Pratap_Singh" title="Vishwanath Pratap Singh">Vishwanath Pratap Singh</a></span></span></span></b><br /><small>(1931–2008)</small><br />
</td>
<td><a href="/wiki/Janata_Dal" title="Janata Dal">Janata Dal</a><br /><span style="font-size:85%;">(<i><a href="/wiki/National_Front_(India)" class="mw-redirect" title="National Front (India)">National Front</a></i>)</span>
</td>
<td>2 December 1989
</td>
<td><span class="nowrap">10 November 1990<sup class="reference" id="ref_NCNC"><a href="#endnote_NCNC">[NC]</a></sup></span>
</td>
<td>343 days
</td>
<td rowspan="2" data-sort-value="9"><a href="/wiki/9th_Lok_Sabha" title="9th Lok Sabha">9th</a>
</td>
<td><a href="/wiki/Vishwanath_Pratap_Singh_ministry" title="Vishwanath Pratap Singh ministry">Vishwanath</a>
</td>
<td rowspan="3"><span data-sort-value="Venkataraman, R."><span class="vcard"><span class="fn"><a href="/wiki/R._Venkataraman" class="mw-redirect" title="R. Venkataraman">R. Venkataraman</a></span></span></span>
</td></tr>
<tr>
<th>8
</th>
<td style="background:#74C365;">
</td>
<td><a href="/wiki/File:Chandra_Shekhar_Singh_enhanced.jpg" class="image"><img alt="Chandra Shekhar Singh enhanced.jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/0/06/Chandra_Shekhar_Singh_enhanced.jpg/103px-Chandra_Shekhar_Singh_enhanced.jpg" decoding="async" width="103" height="136" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/0/06/Chandra_Shekhar_Singh_enhanced.jpg/155px-Chandra_Shekhar_Singh_enhanced.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/0/06/Chandra_Shekhar_Singh_enhanced.jpg/206px-Chandra_Shekhar_Singh_enhanced.jpg 2x" data-file-width="2060" data-file-height="2714" /></a>
</td>
<td><b><span data-sort-value="Shekhar, Chandra"><span class="vcard"><span class="fn"><a href="/wiki/Chandra_Shekhar" title="Chandra Shekhar">Chandra Shekhar</a></span></span></span></b><br /><small>(1927–2007)</small><br />
</td>
<td><a href="/wiki/Samajwadi_Janata_Party_(Rashtriya)" title="Samajwadi Janata Party (Rashtriya)">Samajwadi Janata Party (Rashtriya)</a><br />
</td>
<td><span class="nowrap">10 November 1990</span>
</td>
<td>21 June 1991<sup class="reference" id="ref_RESRES"><a href="#endnote_RESRES">[RES]</a></sup>
</td>
<td>223 days
</td>
<td><a href="/wiki/Chandra_Shekhar_ministry" title="Chandra Shekhar ministry">Chandra Shekhar</a>
</td></tr>
<tr>
<th>9
</th>
<td style="background:#19AAED;">
</td>
<td><a href="/wiki/File:Visit_of_Narasimha_Rao,_Indian_Minister_for_Foreign_Affairs,_to_the_CEC_(cropped)(2).jpg" class="image"><img alt="Visit of Narasimha Rao, Indian Minister for Foreign Affairs, to the CEC (cropped)(2).jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/a/a2/Visit_of_Narasimha_Rao%2C_Indian_Minister_for_Foreign_Affairs%2C_to_the_CEC_%28cropped%29%282%29.jpg/100px-Visit_of_Narasimha_Rao%2C_Indian_Minister_for_Foreign_Affairs%2C_to_the_CEC_%28cropped%29%282%29.jpg" decoding="async" width="100" height="117" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/a/a2/Visit_of_Narasimha_Rao%2C_Indian_Minister_for_Foreign_Affairs%2C_to_the_CEC_%28cropped%29%282%29.jpg/150px-Visit_of_Narasimha_Rao%2C_Indian_Minister_for_Foreign_Affairs%2C_to_the_CEC_%28cropped%29%282%29.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/a/a2/Visit_of_Narasimha_Rao%2C_Indian_Minister_for_Foreign_Affairs%2C_to_the_CEC_%28cropped%29%282%29.jpg/200px-Visit_of_Narasimha_Rao%2C_Indian_Minister_for_Foreign_Affairs%2C_to_the_CEC_%28cropped%29%282%29.jpg 2x" data-file-width="425" data-file-height="496" /></a>
</td>
<td><b><span data-sort-value="Narasimha Rao, Pamulaparthi Venkata"><span class="vcard"><span class="fn"><a href="/wiki/Pamulaparthi_Venkata_Narasimha_Rao" class="mw-redirect" title="Pamulaparthi Venkata Narasimha Rao">Pamulaparthi Venkata  Narasimha Rao</a></span></span></span></b><br /><small>(1921–2004)</small><br />
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress (I)</a>
</td>
<td>21 June 1991
</td>
<td>16 May 1996
</td>
<td>4&#160;years, 330&#160;days
</td>
<td data-sort-value="10"><a href="/wiki/10th_Lok_Sabha" title="10th Lok Sabha">10th</a>
</td>
<td><a href="/wiki/Rao_ministry" title="Rao ministry">Rao</a>
</td></tr>
<tr>
<th><a href="/wiki/Premiership_of_Atal_Bihari_Vajpayee" title="Premiership of Atal Bihari Vajpayee">10</a>
</th>
<td style="background:#FF9933;">
</td>
<td><a href="/wiki/File:Atal_Bihari_Vajpayee_(12_June_2002).jpg" class="image"><img alt="Atal Bihari Vajpayee (12 June 2002).jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/1/10/Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg/100px-Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg" decoding="async" width="100" height="125" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/1/10/Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg/150px-Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/1/10/Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg/200px-Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg 2x" data-file-width="478" data-file-height="597" /></a>
</td>
<td><b><span data-sort-value="Vajpayee, Atal Bihari"><span class="vcard"><span class="fn"><a href="/wiki/Atal_Bihari_Vajpayee" title="Atal Bihari Vajpayee">Atal Bihari Vajpayee</a></span></span></span></b><br /><small>(1924–2018)</small><br />
</td>
<td><a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">Bharatiya Janata Party</a>
</td>
<td>16 May 1996
</td>
<td>1 June 1996<sup class="reference" id="ref_RESRES"><a href="#endnote_RESRES">[RES]</a></sup>
</td>
<td>16 days
</td>
<td rowspan="3" data-sort-value="11"><a href="/wiki/11th_Lok_Sabha" title="11th Lok Sabha">11th</a>
</td>
<td><a href="/wiki/First_Vajpayee_ministry" title="First Vajpayee ministry">Vajpayee I</a>
</td>
<td rowspan="3"><span data-sort-value="Sharma, Shankar Dayal"><span class="vcard"><span class="fn"><a href="/wiki/Shankar_Dayal_Sharma" title="Shankar Dayal Sharma">Shankar Dayal Sharma</a></span></span></span>
</td></tr>
<tr>
<th>11
</th>
<td style="background:#2E8B57;">
</td>
<td><a href="/wiki/File:H._D._Deve_Gowda_BNC.jpg" class="image"><img alt="H. D. Deve Gowda BNC.jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/7/71/H._D._Deve_Gowda_BNC.jpg/100px-H._D._Deve_Gowda_BNC.jpg" decoding="async" width="100" height="150" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/7/71/H._D._Deve_Gowda_BNC.jpg/150px-H._D._Deve_Gowda_BNC.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/7/71/H._D._Deve_Gowda_BNC.jpg/200px-H._D._Deve_Gowda_BNC.jpg 2x" data-file-width="1600" data-file-height="2400" /></a>
</td>
<td><b><span data-sort-value="Deve Gowda, Haradanahalli Doddegowda"><span class="vcard"><span class="fn"><a href="/wiki/Haradanahalli_Doddegowda_Deve_Gowda" class="mw-redirect" title="Haradanahalli Doddegowda Deve Gowda">Haradanahalli Doddegowda Deve Gowda</a></span></span></span></b><br /><small>(1933–)</small><br />
</td>
<td><a href="/wiki/Janata_Dal" title="Janata Dal">Janata Dal</a><br /><span style="font-size:85%;">(<i><a href="/wiki/United_Front_(India)" title="United Front (India)">United Front</a></i>)</span>
</td>
<td>1 June 1996
</td>
<td>21 April 1997<sup class="reference" id="ref_RESRES"><a href="#endnote_RESRES">[RES]</a></sup>
</td>
<td>324 days
</td>
<td><a href="/wiki/Deve_Gowda_ministry" title="Deve Gowda ministry">Deve Gowda</a>
</td></tr>
<tr>
<th>12
</th>
<td style="background:#2E8B57">
</td>
<td><a href="/wiki/File:Inder_Kumar_Gujral_071.jpg" class="image"><img alt="I.K. Gujral" src="//upload.wikimedia.org/wikipedia/commons/thumb/1/1b/Inder_Kumar_Gujral_071.jpg/99px-Inder_Kumar_Gujral_071.jpg" decoding="async" width="99" height="146" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/1/1b/Inder_Kumar_Gujral_071.jpg/149px-Inder_Kumar_Gujral_071.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/1/1b/Inder_Kumar_Gujral_071.jpg/199px-Inder_Kumar_Gujral_071.jpg 2x" data-file-width="725" data-file-height="1065" /></a>
</td>
<td><b><span data-sort-value="Gujral, Inder Kumar"><span class="vcard"><span class="fn"><a href="/wiki/Inder_Kumar_Gujral" title="Inder Kumar Gujral">Inder Kumar Gujral</a></span></span></span></b><br /><small>(1919–2012)</small><br />
</td>
<td><a href="/wiki/Janata_Dal" title="Janata Dal">Janata Dal</a><br /><span style="font-size:85%;">(<i><a href="/wiki/United_Front_(India)" title="United Front (India)">United Front</a></i>)</span>
</td>
<td>21 April 1997
</td>
<td>19 March 1998<sup class="reference" id="ref_RESRES"><a href="#endnote_RESRES">[RES]</a></sup>
</td>
<td>332 days
</td>
<td><a href="/wiki/Gujral_ministry" title="Gujral ministry">Gujral</a>
</td></tr>
<tr>
<th rowspan="2">(<a href="/wiki/Premiership_of_Atal_Bihari_Vajpayee" title="Premiership of Atal Bihari Vajpayee">10</a>)
</th>
<td rowspan="2" style="background:#FF9933;">
</td>
<td rowspan="2"><a href="/wiki/File:Atal_Bihari_Vajpayee_(12_June_2002).jpg" class="image"><img alt="Atal Bihari Vajpayee (12 June 2002).jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/1/10/Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg/100px-Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg" decoding="async" width="100" height="125" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/1/10/Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg/150px-Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/1/10/Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg/200px-Atal_Bihari_Vajpayee_%2812_June_2002%29.jpg 2x" data-file-width="478" data-file-height="597" /></a>
</td>
<td rowspan="2"><b><span data-sort-value="Vajpayee, Atal Bihari"><span class="vcard"><span class="fn"><a href="/wiki/Atal_Bihari_Vajpayee" title="Atal Bihari Vajpayee">Atal Bihari Vajpayee</a></span></span></span></b><br /><small>(1924–2018)</small><br />
</td>
<td rowspan="2"><a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">Bharatiya Janata Party</a><br /><span style="font-size:85%;">(<i><a href="/wiki/National_Democratic_Alliance_(India)" class="mw-redirect" title="National Democratic Alliance (India)">NDA</a></i>)</span>
</td>
<td>19 March 1998<sup class="reference" id="ref_§§"><a href="#endnote_§§">[§]</a></sup>
</td>
<td><span class="nowrap">10 October 1999<sup class="reference" id="ref_NCNC"><a href="#endnote_NCNC">[NC]</a></sup></span>
</td>
<td rowspan="2">6&#160;years, 64&#160;days
</td>
<td data-sort-value="12"><a href="/wiki/12th_Lok_Sabha" title="12th Lok Sabha">12th</a>
</td>
<td><a href="/wiki/Second_Vajpayee_ministry" title="Second Vajpayee ministry">Vajpayee II</a>
</td>
<td rowspan="2"><span data-sort-value="Narayanan, K. R."><span class="vcard"><span class="fn"><a href="/wiki/K._R._Narayanan" title="K. R. Narayanan">K. R. Narayanan</a></span></span></span>
</td></tr>
<tr>
<td>10 October 1999
</td>
<td>22 May 2004
</td>
<td data-sort-value="13"><a href="/wiki/13th_Lok_Sabha" title="13th Lok Sabha">13th</a>
</td>
<td><a href="/wiki/Third_Vajpayee_ministry" title="Third Vajpayee ministry">Vajpayee III</a>
</td></tr>
<tr>
<th rowspan="2">13
</th>
<td rowspan="2" style="background:#19AAED;">
</td>
<td rowspan="2"><a href="/wiki/File:Prime_Minister_Dr._Manmohan_Singh_in_March_2014.jpg" class="image"><img alt="Manmohan Singh" src="//upload.wikimedia.org/wikipedia/commons/thumb/2/2c/Prime_Minister_Dr._Manmohan_Singh_in_March_2014.jpg/100px-Prime_Minister_Dr._Manmohan_Singh_in_March_2014.jpg" decoding="async" width="100" height="126" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/2/2c/Prime_Minister_Dr._Manmohan_Singh_in_March_2014.jpg/150px-Prime_Minister_Dr._Manmohan_Singh_in_March_2014.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/2/2c/Prime_Minister_Dr._Manmohan_Singh_in_March_2014.jpg/200px-Prime_Minister_Dr._Manmohan_Singh_in_March_2014.jpg 2x" data-file-width="879" data-file-height="1106" /></a>
</td>
<td rowspan="2"><b><span data-sort-value="Singh, Manmohan"><span class="vcard"><span class="fn"><a href="/wiki/Manmohan_Singh" title="Manmohan Singh">Manmohan Singh</a></span></span></span></b><br /><small>(1932–)</small><br />
</td>
<td rowspan="2"><a href="/wiki/Indian_National_Congress" title="Indian National Congress">Indian National Congress</a><br /><span style="font-size:85%;">(<i><a href="/wiki/United_Progressive_Alliance" title="United Progressive Alliance">UPA</a></i>)</span>
</td>
<td>22 May 2004
</td>
<td>22 May 2009
</td>
<td rowspan="2">10&#160;years, 4&#160;days
</td>
<td data-sort-value="14"><a href="/wiki/14th_Lok_Sabha" title="14th Lok Sabha">14th</a>
</td>
<td><a href="/wiki/First_Manmohan_Singh_ministry" title="First Manmohan Singh ministry">Manmohan I</a>
</td>
<td><span data-sort-value="Kalam, A. P. J. Abdul"><span class="vcard"><span class="fn"><a href="/wiki/A._P._J._Abdul_Kalam" title="A. P. J. Abdul Kalam">A. P. J. Abdul Kalam</a></span></span></span>
</td></tr>
<tr>
<td>22 May 2009
</td>
<td>26 May 2014
</td>
<td data-sort-value="15"><a href="/wiki/15th_Lok_Sabha" title="15th Lok Sabha">15th</a>
</td>
<td><a href="/wiki/Second_Manmohan_Singh_ministry" title="Second Manmohan Singh ministry">Manmohan II</a>
</td>
<td><span data-sort-value="Patil, Pratibha"><span class="vcard"><span class="fn"><a href="/wiki/Pratibha_Patil" title="Pratibha Patil">Pratibha Patil</a></span></span></span>
</td></tr>
<tr>
<th rowspan="2"><a href="/wiki/Premiership_of_Narendra_Modi" title="Premiership of Narendra Modi">14</a>
</th>
<td rowspan="2" style="background:#FF9933;">
</td>
<td rowspan="2"><a href="/wiki/File:Prime_Minister_Narendra_Modi_welcomed_at_world_summit_.jpg" class="image"><img alt="Narendra Modi" src="//upload.wikimedia.org/wikipedia/commons/thumb/7/7a/Prime_Minister_Narendra_Modi_welcomed_at_world_summit_.jpg/93px-Prime_Minister_Narendra_Modi_welcomed_at_world_summit_.jpg" decoding="async" width="93" height="100" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/7/7a/Prime_Minister_Narendra_Modi_welcomed_at_world_summit_.jpg/139px-Prime_Minister_Narendra_Modi_welcomed_at_world_summit_.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/7/7a/Prime_Minister_Narendra_Modi_welcomed_at_world_summit_.jpg/185px-Prime_Minister_Narendra_Modi_welcomed_at_world_summit_.jpg 2x" data-file-width="819" data-file-height="885" /></a>
</td>
<td rowspan="2"><b><a href="/wiki/Narendra_Modi" title="Narendra Modi">Narendra Damodardas Modi</a></b><br /><small>(1950–)</small><br />
</td>
<td rowspan="2"><a href="/wiki/Bharatiya_Janata_Party" title="Bharatiya Janata Party">Bharatiya Janata Party</a><br /><span style="font-size:85%;">(<i><a href="/wiki/National_Democratic_Alliance_(India)" class="mw-redirect" title="National Democratic Alliance (India)">NDA</a></i>)</span>
</td>
<td>26 May 2014
</td>
<td>30 May 2019
</td>
<td rowspan="2">7&#160;years, 305&#160;days
</td>
<td data-sort-value="16"><a href="/wiki/16th_Lok_Sabha" title="16th Lok Sabha">16th</a>
</td>
<td><a href="/wiki/First_Modi_ministry" title="First Modi ministry">Modi I</a>
</td>
<td><span data-sort-value="Mukherjee, Pranab"><span class="vcard"><span class="fn"><a href="/wiki/Pranab_Mukherjee" title="Pranab Mukherjee">Pranab Mukherjee</a></span></span></span>
</td></tr>
<tr>
<td>30 May 2019
</td>
<td><i>Incumbent</i>
</td>
<td><a href="/wiki/17th_Lok_Sabha" title="17th Lok Sabha">17th</a>
</td>
<td><a href="/wiki/Second_Modi_ministry" title="Second Modi ministry">Modi II</a>
</td>
<td><span data-sort-value="Kovind, Ram Nath"><span class="vcard"><span class="fn"><a href="/wiki/Ram_Nath_Kovind" title="Ram Nath Kovind">Ram Nath Kovind</a></span></span></span>
</td></tr></tbody></table>
<h2><span class="mw-headline" id="Statistics">Statistics</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit&amp;section=2" title="Edit section: Statistics">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<dl><dt>List of prime ministers by length of term</dt></dl>
<table class="wikitable sortable" style="width:70%; text-align:center">

<tbody><tr>
<th data-sort-type="number" rowspan="2"><style data-mw-deduplicate="TemplateStyles:r1038841319">.mw-parser-output .tooltip-dotted{border-bottom:1px dotted;cursor:help}</style><span title="" class="rt-commentedText tooltip tooltip-dotted">No.</span>
</th>
<th rowspan="2">Name
</th>
<th rowspan="2">Party
</th>
<th colspan="2">Length of term
</th></tr>
<tr>
<th>Longest continuous term
</th>
<th>Total years of premiership
</th></tr>
<tr>
<th data-sort-value="1">1
</th>
<td><span data-sort-value="Nehru, Jawaharlal"><span class="vcard"><span class="fn"><a href="/wiki/Jawaharlal_Nehru" title="Jawaharlal Nehru">Jawaharlal Nehru</a></span></span></span>
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC</a>
</td>
<td>16 years, 286 days
</td>
<td>16 years, 286 days
</td></tr>
<tr>
<th data-sort-value="2">2
</th>
<td><span data-sort-value="Gandhi, Indira"><span class="vcard"><span class="fn"><a href="/wiki/Indira_Gandhi" title="Indira Gandhi">Indira Gandhi</a></span></span></span>
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC/INC(I)</a>/<a href="/wiki/Indian_National_Congress_(R)" title="Indian National Congress (R)">INC(R)</a>
</td>
<td>11 years, 59 days
</td>
<td>15 years, 350 days
</td></tr>
<tr>
<th data-sort-value="3">3
</th>
<td><span data-sort-value="Singh, Manmohan"><span class="vcard"><span class="fn"><a href="/wiki/Manmohan_Singh" title="Manmohan Singh">Manmohan Singh</a></span></span></span>
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC</a>
</td>
<td>10 years, 4 days
</td>
<td>10 years, 4 days
</td></tr>
<tr>
<th data-sort-value="4">4
</th>
<td><b><span data-sort-value="Modi, Narendra"><span class="vcard"><span class="fn"><a href="/wiki/Narendra_Modi" title="Narendra Modi">Narendra Modi</a></span></span></span></b>
</td>
<td><b><a href="/wiki/BJP" class="mw-redirect" title="BJP">BJP</a></b>
</td>
<td><b>7&#160;years, 305&#160;days</b>
</td>
<td><b>7&#160;years, 305&#160;days</b>
</td></tr>
<tr>
<th data-sort-value="5">5
</th>
<td><span data-sort-value="Vajpayee, Atal Bihari"><span class="vcard"><span class="fn"><a href="/wiki/Atal_Bihari_Vajpayee" title="Atal Bihari Vajpayee">Atal Bihari Vajpayee</a></span></span></span>
</td>
<td><a href="/wiki/BJP" class="mw-redirect" title="BJP">BJP</a>
</td>
<td>6 years, 64 days
</td>
<td>6 years, 80 days
</td></tr>
<tr>
<th data-sort-value="6">6
</th>
<td><span data-sort-value="Gandhi, Rajiv"><span class="vcard"><span class="fn"><a href="/wiki/Rajiv_Gandhi" title="Rajiv Gandhi">Rajiv Gandhi</a></span></span></span>
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC(I)</a>
</td>
<td>5 years, 32 days
</td>
<td>5 years, 32 days
</td></tr>
<tr>
<th data-sort-value="7">7
</th>
<td><span data-sort-value="Rao, P. V. Narasimha"><span class="vcard"><span class="fn"><a href="/wiki/P._V._Narasimha_Rao" title="P. V. Narasimha Rao">P. V. Narasimha Rao</a></span></span></span>
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC(I)</a>
</td>
<td>4 years, 330 days
</td>
<td>4 years, 330 days
</td></tr>
<tr>
<th data-sort-value="8">8
</th>
<td><span data-sort-value="Desai, Morarji"><span class="vcard"><span class="fn"><a href="/wiki/Morarji_Desai" title="Morarji Desai">Morarji Desai</a></span></span></span>
</td>
<td><a href="/wiki/Janata_Party" title="Janata Party">JP</a>
</td>
<td>2 years, 126 days
</td>
<td>2 years, 126 days
</td></tr>
<tr>
<th data-sort-value="9">9
</th>
<td><span data-sort-value="Shastri, Lal Bahadur"><span class="vcard"><span class="fn"><a href="/wiki/Lal_Bahadur_Shastri" title="Lal Bahadur Shastri">Lal Bahadur Shastri</a></span></span></span>
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC</a>
</td>
<td>1 year, 216 days
</td>
<td>1 year, 216 days
</td></tr>
<tr>
<th data-sort-value="10">10
</th>
<td><span data-sort-value="Singh, Vishwanath Pratap"><span class="vcard"><span class="fn"><a href="/wiki/Vishwanath_Pratap_Singh" title="Vishwanath Pratap Singh">Vishwanath Pratap Singh</a></span></span></span>
</td>
<td><a href="/wiki/Janata_Dal" title="Janata Dal">JD</a>
</td>
<td>343 days
</td>
<td>343 days
</td></tr>
<tr>
<th data-sort-value="11">11
</th>
<td><span data-sort-value="Gujral, Inder Kumar"><span class="vcard"><span class="fn"><a href="/wiki/Inder_Kumar_Gujral" title="Inder Kumar Gujral">Inder Kumar Gujral</a></span></span></span>
</td>
<td><a href="/wiki/Janata_Dal" title="Janata Dal">JD</a>
</td>
<td>332 days
</td>
<td>332 days
</td></tr>
<tr>
<th data-sort-value="12">12
</th>
<td><span data-sort-value="Deve Gowda, H. D."><span class="vcard"><span class="fn"><a href="/wiki/H._D._Deve_Gowda" title="H. D. Deve Gowda">H. D. Deve Gowda</a></span></span></span>
</td>
<td><a href="/wiki/Janata_Dal" title="Janata Dal">JD</a>
</td>
<td>324 days
</td>
<td>324 days
</td></tr>
<tr>
<th data-sort-value="13">13
</th>
<td><span data-sort-value="Shekhar, Chandra"><span class="vcard"><span class="fn"><a href="/wiki/Chandra_Shekhar" title="Chandra Shekhar">Chandra Shekhar</a></span></span></span>
</td>
<td><a href="/wiki/Samajwadi_Janata_Party_(Rashtriya)" title="Samajwadi Janata Party (Rashtriya)">SJP(R)</a>
</td>
<td>223 days
</td>
<td>223 days
</td></tr>
<tr>
<th data-sort-value="14">14
</th>
<td><span data-sort-value="Singh, Charan"><span class="vcard"><span class="fn"><a href="/wiki/Charan_Singh" title="Charan Singh">Charan Singh</a></span></span></span>
</td>
<td><a href="/wiki/Janata_Party_(Secular)" title="Janata Party (Secular)">JP(S)</a>
</td>
<td>170 days
</td>
<td>170 days
</td></tr>
<tr>
<th data-sort-value="15">Acting
</th>
<td><span data-sort-value="Nanda, Gulzarilal"><span class="vcard"><span class="fn"><a href="/wiki/Gulzarilal_Nanda" title="Gulzarilal Nanda">Gulzarilal Nanda</a></span></span></span>
</td>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC</a>
</td>
<td>13 days
</td>
<td>26 days
</td></tr></tbody></table>
<dl><dt>Timeline</dt></dl>
<div class="timeline-wrapper"><map name="timeline_o3w4r3n5ddo2fydrphzck0ot0icbyhe"><area shape="rect" href="/wiki/Narendra_Modi" coords="950,321,1015,345" title="Narendra Modi" alt="Narendra Modi" /><area shape="rect" href="/wiki/Manmohan_Singh" coords="853,303,945,327" title="Manmohan Singh" alt="Manmohan Singh" /><area shape="rect" href="/wiki/Atal_Bihari_Vajpayee" coords="732,284,824,308" title="Atal Bihari Vajpayee" alt="Atal Bihari Vajpayee" /><area shape="rect" href="/wiki/Inder_Kumar_Gujral" coords="657,265,735,289" title="Inder Kumar Gujral" alt="Inder Kumar Gujral" /><area shape="rect" href="/wiki/H._D._Deve_Gowda" coords="646,246,751,270" title="H. D. Deve Gowda" alt="H. D. Deve Gowda" /><area shape="rect" href="/wiki/Atal_Bihari_Vajpayee" coords="635,227,727,251" title="Atal Bihari Vajpayee" alt="Atal Bihari Vajpayee" /><area shape="rect" href="/wiki/P._V._Narasimha_Rao" coords="635,208,693,232" title="P. V. Narasimha Rao" alt="P. V. Narasimha Rao" /><area shape="rect" href="/wiki/Chandra_Shekhar" coords="575,189,713,213" title="Chandra Shekhar" alt="Chandra Shekhar" /><area shape="rect" href="/wiki/Vishwanath_Pratap_Singh" coords="568,170,673,194" title="Vishwanath Pratap Singh" alt="Vishwanath Pratap Singh" /><area shape="rect" href="/wiki/Rajiv_Gandhi" coords="556,151,628,175" title="Rajiv Gandhi" alt="Rajiv Gandhi" /><area shape="rect" href="/wiki/Indira_Gandhi" coords="494,133,573,157" title="Indira Gandhi" alt="Indira Gandhi" /><area shape="rect" href="/wiki/Charan_Singh" coords="436,114,514,138" title="Charan Singh" alt="Charan Singh" /><area shape="rect" href="/wiki/Morarji_Desai" coords="430,95,502,119" title="Morarji Desai" alt="Morarji Desai" /><area shape="rect" href="/wiki/Indira_Gandhi" coords="402,76,480,100" title="Indira Gandhi" alt="Indira Gandhi" /><area shape="rect" href="/wiki/Gulzarilal_Nanda" coords="266,57,338,81" title="Gulzarilal Nanda" alt="Gulzarilal Nanda" /><area shape="rect" href="/wiki/Lal_Bahadur_Shastri" coords="265,38,350,62" title="Lal Bahadur Shastri" alt="Lal Bahadur Shastri" /><area shape="rect" href="/wiki/Gulzarilal_Nanda" coords="246,19,318,43" title="Gulzarilal Nanda" alt="Gulzarilal Nanda" /><area shape="rect" href="/wiki/Jawaharlal_Nehru" coords="246,0,317,24" title="Jawaharlal Nehru" alt="Jawaharlal Nehru" /></map><img usemap="#timeline_o3w4r3n5ddo2fydrphzck0ot0icbyhe" src="//upload.wikimedia.org/wikipedia/en/timeline/o3w4r3n5ddo2fydrphzck0ot0icbyhe.png" /></div>
<dl><dt>List by party</dt></dl>
<table class="wikitable sortable" style="width:100%; text-align:center">
<caption>Political parties by total timespan of their member holding <a href="/wiki/Prime_Minister%27s_Office_(India)" title="Prime Minister&#39;s Office (India)">PMO</a> (1 October 2021)
</caption>
<tbody><tr>
<th data-sort-type="number"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1038841319"/><span title="" class="rt-commentedText tooltip tooltip-dotted">No.</span>
</th>
<th>Political party
</th>
<th data-sort-type="number">Number of Prime ministers
</th>
<th>Total years of holding PMO
</th></tr>
<tr>
<th data-sort-value="1">1
</th>
<td><a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC/INC(I)</a>/<a href="/wiki/Indian_National_Congress_(R)" title="Indian National Congress (R)">INC(R)</a>
</td>
<td data-sort-type="6">6 (+1 acting)
</td>
<td>54 years, 123 days
</td></tr>
<tr>
<th data-sort-value="2">2
</th>
<td><b><a href="/wiki/BJP" class="mw-redirect" title="BJP">BJP</a></b>
</td>
<td data-sort-type="2"><b>2</b>
</td>
<td><b>14&#160;years, 17&#160;days</b>
</td></tr>
<tr>
<th data-sort-value="3">3
</th>
<td><a href="/wiki/Janata_Dal" title="Janata Dal">JD</a>
</td>
<td data-sort-type="3">3
</td>
<td>2 years, 269 days
</td></tr>
<tr>
<th data-sort-value="4">4
</th>
<td><a href="/wiki/Janata_Party" title="Janata Party">JP</a>
</td>
<td data-sort-type="1">1
</td>
<td>2 years, 126 days
</td></tr>
<tr>
<th data-sort-value="5">5
</th>
<td><a href="/wiki/Samajwadi_Janata_Party_(Rashtriya)" title="Samajwadi Janata Party (Rashtriya)">SJP(R)</a>
</td>
<td data-sort-type="1">1
</td>
<td>223 days
</td></tr>
<tr>
<th data-sort-value="6">6
</th>
<td><a href="/wiki/Janata_Party_(Secular)" title="Janata Party (Secular)">JP(S)</a>
</td>
<td data-sort-type="1">1
</td>
<td>170 days
</td></tr></tbody></table>
<dl><dt>Parties by total duration (in years) of holding Prime Minister's Office</dt></dl>
<div style="margin-top:1em;max-width:800px;" class="chart noresize">
<div style="position:relative;min-height:300px;min-width:800px;max-width:800px;">
<div style="float:right;position:relative;min-height:220px;min-width:700px;max-width:700px;border-left:1px black solid;border-bottom:1px black solid;">
<div style="position:absolute;left:39px;top:21px;height:198px;min-width:36px;max-width:36px;background-color:#00BFFF;-webkit-print-color-adjust:exact;border:1px solid #00BFFF;border-bottom:none;overflow:hidden;" title="&#91;&#91;Indian National Congress&#124;INC/INC(I)&#93;&#93;/&#91;&#91;Indian National Congress (R)&#124;INC(R)&#93;&#93;: 54.34"></div>
<div style="position:absolute;left:155px;top:170px;height:49px;min-width:36px;max-width:36px;background-color:#FF9933;-webkit-print-color-adjust:exact;border:1px solid #FF9933;border-bottom:none;overflow:hidden;" title="&#91;&#91;BJP&#93;&#93;: 13.57"></div>
<div style="position:absolute;left:271px;top:210px;height:9px;min-width:36px;max-width:36px;background-color:#2E8B57;-webkit-print-color-adjust:exact;border:1px solid #2E8B57;border-bottom:none;overflow:hidden;" title="&#91;&#91;Janata Dal&#124;JD&#93;&#93;: 2.74"></div>
<div style="position:absolute;left:387px;top:211px;height:8px;min-width:36px;max-width:36px;background-color:#1F75FE;-webkit-print-color-adjust:exact;border:1px solid #1F75FE;border-bottom:none;overflow:hidden;" title="&#91;&#91;Janata Party&#124;JP&#93;&#93;: 2.35"></div>
<div style="position:absolute;left:503px;top:218px;height:1px;min-width:36px;max-width:36px;background-color:#00008B;-webkit-print-color-adjust:exact;border:1px solid #00008B;border-bottom:none;overflow:hidden;" title="&#91;&#91;Janata Party (Secular)&#124;JP(S)&#93;&#93;: 0.61"></div>
<div style="position:absolute;left:619px;top:218px;height:1px;min-width:36px;max-width:36px;background-color:#74C365;-webkit-print-color-adjust:exact;border:1px solid #74C365;border-bottom:none;overflow:hidden;" title="&#91;&#91;Samajwadi Janata Party (Rashtriya)&#124;SJP(R)&#93;&#93;: 0.47"></div>
</div>
<div style="position:absolute;height:220px;min-width:100px;max-width:100px;">
<div style="position:absolute;height=20px;text-align:right;vertical-align:middle;width:90px;top:173px;padding:0 2px">10</div>
<div style="position:absolute;height=1px;min-width:5px;top:183px;left:96px;border:1px solid black;"></div>
<div style="position:absolute;height=20px;text-align:right;vertical-align:middle;width:90px;top:137px;padding:0 2px">20</div>
<div style="position:absolute;height=1px;min-width:5px;top:147px;left:96px;border:1px solid black;"></div>
<div style="position:absolute;height=20px;text-align:right;vertical-align:middle;width:90px;top:100px;padding:0 2px">30</div>
<div style="position:absolute;height=1px;min-width:5px;top:110px;left:96px;border:1px solid black;"></div>
<div style="position:absolute;height=20px;text-align:right;vertical-align:middle;width:90px;top:63px;padding:0 2px">40</div>
<div style="position:absolute;height=1px;min-width:5px;top:73px;left:96px;border:1px solid black;"></div>
<div style="position:absolute;height=20px;text-align:right;vertical-align:middle;width:90px;top:27px;padding:0 2px">50</div>
<div style="position:absolute;height=1px;min-width:5px;top:37px;left:96px;border:1px solid black;"></div>
<div style="position:absolute;height=20px;text-align:right;vertical-align:middle;width:90px;top:-10px;padding:0 2px">60</div>
<div style="position:absolute;height=1px;min-width:5px;top:0px;left:96px;border:1px solid black;"></div>
</div>
<div style="position:absolute;top:220px;left:100px;width:700px;">
<div style="position:absolute;left:5px;top:10px;min-width:106px;max-width:106px;text-align:center;vertical-align:top;">INC</div>
<div style="position:absolute;left:58px;height:10px;width:1px;border-left:1px solid black;"></div>
<div style="position:absolute;left:121px;top:10px;min-width:106px;max-width:106px;text-align:center;vertical-align:top;">BJP</div>
<div style="position:absolute;left:174px;height:10px;width:1px;border-left:1px solid black;"></div>
<div style="position:absolute;left:237px;top:10px;min-width:106px;max-width:106px;text-align:center;vertical-align:top;">JD</div>
<div style="position:absolute;left:290px;height:10px;width:1px;border-left:1px solid black;"></div>
<div style="position:absolute;left:353px;top:10px;min-width:106px;max-width:106px;text-align:center;vertical-align:top;">JP</div>
<div style="position:absolute;left:406px;height:10px;width:1px;border-left:1px solid black;"></div>
<div style="position:absolute;left:469px;top:10px;min-width:106px;max-width:106px;text-align:center;vertical-align:top;">JP(S)</div>
<div style="position:absolute;left:522px;height:10px;width:1px;border-left:1px solid black;"></div>
<div style="position:absolute;left:585px;top:10px;min-width:106px;max-width:106px;text-align:center;vertical-align:top;">SJP(R)</div>
<div style="position:absolute;left:638px;height:10px;width:1px;border-left:1px solid black;"></div>
</div>
</div>
<div>
<ul style="width:100%;list-style:none;column-width:12em;"><li><span style="padding:0 1em;background-color:#00BFFF;border:1px solid #00BFFF;margin-right:1em;-webkit-print-color-adjust:exact;">&#160;</span> <a href="/wiki/Indian_National_Congress" title="Indian National Congress">INC/INC(I)</a>/<a href="/wiki/Indian_National_Congress_(R)" title="Indian National Congress (R)">INC(R)</a></li>
<li><span style="padding:0 1em;background-color:#FF9933;border:1px solid #FF9933;margin-right:1em;-webkit-print-color-adjust:exact;">&#160;</span> <a href="/wiki/BJP" class="mw-redirect" title="BJP">BJP</a></li>
<li><span style="padding:0 1em;background-color:#2E8B57;border:1px solid #2E8B57;margin-right:1em;-webkit-print-color-adjust:exact;">&#160;</span> <a href="/wiki/Janata_Dal" title="Janata Dal">JD</a></li>
<li><span style="padding:0 1em;background-color:#1F75FE;border:1px solid #1F75FE;margin-right:1em;-webkit-print-color-adjust:exact;">&#160;</span> <a href="/wiki/Janata_Party" title="Janata Party">JP</a></li>
<li><span style="padding:0 1em;background-color:#00008B;border:1px solid #00008B;margin-right:1em;-webkit-print-color-adjust:exact;">&#160;</span> <a href="/wiki/Janata_Party_(Secular)" title="Janata Party (Secular)">JP(S)</a></li>
<li><span style="padding:0 1em;background-color:#74C365;border:1px solid #74C365;margin-right:1em;-webkit-print-color-adjust:exact;">&#160;</span> <a href="/wiki/Samajwadi_Janata_Party_(Rashtriya)" title="Samajwadi Janata Party (Rashtriya)">SJP(R)</a></li></ul>
</div>
</div>
<h2><span class="mw-headline" id="See_also">See also</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit&amp;section=3" title="Edit section: See also">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<style data-mw-deduplicate="TemplateStyles:r1072126029">.mw-parser-output .portalbox{float:right;padding:0}.mw-parser-output .portalborder{border:solid #aaa 1px}.mw-parser-output .portalbox.tleft{margin:0.5em 1em 0.5em 0}.mw-parser-output .portalbox.tright{margin:0.5em 0 0.5em 1em}.mw-parser-output .portalbox>ul{display:table;box-sizing:border-box;max-width:175px;font-size:85%;line-height:110%;font-style:italic;font-weight:bold}.mw-parser-output .portalborder>ul{padding:0.1em;background:#f9f9f9}.mw-parser-output .portalbox>ul>li{display:table-row}.mw-parser-output .portalbox>ul>li>span:first-child{display:table-cell;padding:0.2em;vertical-align:middle;text-align:center}.mw-parser-output .portalbox>ul>li>span:last-child{display:table-cell;padding:0.2em 0.2em 0.2em 0.3em;vertical-align:middle}</style><div role="navigation" aria-label="Portals" class="noprint plainlist portalbox portalborder tright">
<ul>
<li><span><img alt="flag" src="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/32px-Flag_of_India.svg.png" decoding="async" width="32" height="21" class="noviewer thumbborder" srcset="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/48px-Flag_of_India.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/64px-Flag_of_India.svg.png 2x" data-file-width="1350" data-file-height="900" /></span><span><a href="/wiki/Portal:India" title="Portal:India">India portal</a></span></li></ul></div>
<ul><li><a href="/wiki/President_of_India" title="President of India">President of India</a></li>
<li><a href="/wiki/Vice_President_of_India" title="Vice President of India">Vice President of India</a></li>
<li><a href="/wiki/Deputy_Prime_Minister_of_India" title="Deputy Prime Minister of India">Deputy Prime Minister of India</a></li>
<li><a href="/wiki/List_of_presidents_of_India" title="List of presidents of India">List of presidents of India</a></li>
<li><a href="/wiki/List_of_vice_presidents_of_India" title="List of vice presidents of India">List of vice presidents of India</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_India_by_education" title="List of prime ministers of India by education">List of prime ministers of India by education</a></li>
<li><a href="/wiki/List_of_children_of_the_prime_ministers_of_India" title="List of children of the prime ministers of India">List of children of the prime ministers of India</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_India_by_previous_experience" title="List of prime ministers of India by previous experience">List of prime ministers of India by previous experience</a></li>
<li><a href="/wiki/List_of_heads_of_state_and_government_of_Indian_origin" title="List of heads of state and government of Indian origin">List of heads of state and government of Indian origin</a></li>
<li><a href="/wiki/List_of_nicknames_of_the_prime_ministers_of_India" title="List of nicknames of the prime ministers of India">List of nicknames of the prime ministers of India</a></li></ul>
<h2><span class="mw-headline" id="Footnotes">Footnotes</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit&amp;section=4" title="Edit section: Footnotes">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<ul><li><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1041539562"/><span class="citation wikicite" id="endnote_††"><a href="#ref_††"><b><sup>†</sup></b></a></span>  Assassinated or died in office</li>
<li><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1041539562"/><span class="citation wikicite" id="endnote_§§"><a href="#ref_§§"><b><sup>§</sup></b></a></span>  Returned to office after a previous non-consecutive term</li>
<li><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1041539562"/><span class="citation wikicite" id="endnote_RESRES"><a href="#ref_RESRES"><b><sup>RES</sup></b></a></span>  Resigned</li>
<li><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1041539562"/><span class="citation wikicite" id="endnote_NCNC"><a href="#ref_NCNC"><b><sup>NC</sup></b></a></span>  Resigned following a <a href="/wiki/No-confidence_motion" class="mw-redirect" title="No-confidence motion">no-confidence motion</a></li></ul>
<h2><span class="mw-headline" id="Notes">Notes</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit&amp;section=5" title="Edit section: Notes">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<style data-mw-deduplicate="TemplateStyles:r1011085734">.mw-parser-output .reflist{font-size:90%;margin-bottom:0.5em;list-style-type:decimal}.mw-parser-output .reflist .references{font-size:100%;margin-bottom:0;list-style-type:inherit}.mw-parser-output .reflist-columns-2{column-width:30em}.mw-parser-output .reflist-columns-3{column-width:25em}.mw-parser-output .reflist-columns{margin-top:0.3em}.mw-parser-output .reflist-columns ol{margin-top:0}.mw-parser-output .reflist-columns li{page-break-inside:avoid;break-inside:avoid-column}.mw-parser-output .reflist-upper-alpha{list-style-type:upper-alpha}.mw-parser-output .reflist-upper-roman{list-style-type:upper-roman}.mw-parser-output .reflist-lower-alpha{list-style-type:lower-alpha}.mw-parser-output .reflist-lower-greek{list-style-type:lower-greek}.mw-parser-output .reflist-lower-roman{list-style-type:lower-roman}</style><div class="reflist reflist-lower-alpha">
<div class="mw-references-wrap"><ol class="references">
<li id="cite_note-2"><span class="mw-cite-backlink"><b><a href="#cite_ref-2">^</a></b></span> <span class="reference-text">15 including <a href="/wiki/Gulzarilal_Nanda" title="Gulzarilal Nanda">Gulzarilal Nanda</a> who twice acted in the role, of which 6 having at least one full term, ruling country for about 60 years.<sup id="cite_ref-1" class="reference"><a href="#cite_note-1">&#91;1&#93;</a></sup></span>
</li>
<li id="cite_note-12"><span class="mw-cite-backlink"><b><a href="#cite_ref-12">^</a></b></span> <span class="reference-text">In office</span>
</li>
<li id="cite_note-14"><span class="mw-cite-backlink"><b><a href="#cite_ref-14">^</a></b></span> <span class="reference-text">Known as <a href="/wiki/Indian_National_Congress_(R)" title="Indian National Congress (R)">Indian National Congress (R)</a> between 1969–1978 Indian National Congress (I) between 1978–96.
<sup id="cite_ref-13" class="reference"><a href="#cite_note-13">&#91;11&#93;</a></sup></span>
</li>
<li id="cite_note-15"><span class="mw-cite-backlink"><b><a href="#cite_ref-15">^</a></b></span> <span class="reference-text"><a href="/wiki/Gulzarilal_Nanda" title="Gulzarilal Nanda">Gulzarilal Nanda</a> twice appointed as acting Prime minister of India following deaths of two prime ministers.</span>
</li>
<li id="cite_note-17"><span class="mw-cite-backlink"><b><a href="#cite_ref-17">^</a></b></span> <span class="reference-text">Although the prime minister can be a member of either house of the <a href="/wiki/Parliament_of_India" title="Parliament of India">Parliament</a>, they have to command the confidence of the <a href="/wiki/Lok_Sabha" title="Lok Sabha">Lok Sabha</a>. Upon dissolution of the Lok Sabha, the outgoing PM remains in office until their successor is sworn in.</span>
</li>
<li id="cite_note-18"><span class="mw-cite-backlink"><b><a href="#cite_ref-18">^</a></b></span> <span class="reference-text">The <a href="/wiki/Constituent_Assembly_of_India" title="Constituent Assembly of India">Constituent Assembly of India</a> consisted of 389 members elected in 1946 by the provincial assemblies by a <a href="/wiki/Single_transferable_vote" title="Single transferable vote">single, transferable-vote</a> system of proportional representation. The Assembly was replaced by the Provisional Parliament of India after adoption of the <a href="/wiki/Constitution_of_India" title="Constitution of India">Constitution</a> on 26 January 1950 until the <a href="/wiki/Indian_general_election,_1951%E2%80%9352" class="mw-redirect" title="Indian general election, 1951–52">first general elections</a>.</span>
</li>
</ol></div></div>
<h2><span class="mw-headline" id="References">References</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit&amp;section=6" title="Edit section: References">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1011085734"/><div class="reflist">
<div class="mw-references-wrap mw-references-columns"><ol class="references">
<li id="cite_note-1"><span class="mw-cite-backlink"><b><a href="#cite_ref-1">^</a></b></span> <span class="reference-text"><style data-mw-deduplicate="TemplateStyles:r1067248974">.mw-parser-output cite.citation{font-style:inherit;word-wrap:break-word}.mw-parser-output .citation q{quotes:"\"""\"""'""'"}.mw-parser-output .citation:target{background-color:rgba(0,127,255,0.133)}.mw-parser-output .id-lock-free a,.mw-parser-output .citation .cs1-lock-free a{background:linear-gradient(transparent,transparent),url("//upload.wikimedia.org/wikipedia/commons/6/65/Lock-green.svg")right 0.1em center/9px no-repeat}.mw-parser-output .id-lock-limited a,.mw-parser-output .id-lock-registration a,.mw-parser-output .citation .cs1-lock-limited a,.mw-parser-output .citation .cs1-lock-registration a{background:linear-gradient(transparent,transparent),url("//upload.wikimedia.org/wikipedia/commons/d/d6/Lock-gray-alt-2.svg")right 0.1em center/9px no-repeat}.mw-parser-output .id-lock-subscription a,.mw-parser-output .citation .cs1-lock-subscription a{background:linear-gradient(transparent,transparent),url("//upload.wikimedia.org/wikipedia/commons/a/aa/Lock-red-alt-2.svg")right 0.1em center/9px no-repeat}.mw-parser-output .cs1-ws-icon a{background:linear-gradient(transparent,transparent),url("//upload.wikimedia.org/wikipedia/commons/4/4c/Wikisource-logo.svg")right 0.1em center/12px no-repeat}.mw-parser-output .cs1-code{color:inherit;background:inherit;border:none;padding:inherit}.mw-parser-output .cs1-hidden-error{display:none;color:#d33}.mw-parser-output .cs1-visible-error{color:#d33}.mw-parser-output .cs1-maint{display:none;color:#3a3;margin-left:0.3em}.mw-parser-output .cs1-format{font-size:95%}.mw-parser-output .cs1-kern-left{padding-left:0.2em}.mw-parser-output .cs1-kern-right{padding-right:0.2em}.mw-parser-output .citation .mw-selflink{font-weight:inherit}</style><cite id="CITEREFMahurkar1996" class="citation news cs1">Mahurkar, Uday (15 May 1996). <a rel="nofollow" class="external text" href="https://www.indiatoday.in/magazine/profile/story/19960515-at-98-two-time-interim-pm-gulzarilal-nanda-is-the-epitome-of-gandhian-ideals-834853-1996-05-15">"At 98, two-time interim PM Gulzarilal Nanda is the epitome of Gandhian ideals"</a>. <i><a href="/wiki/India_Today" title="India Today">India Today</a></i><span class="reference-accessdate">. Retrieved <span class="nowrap">4 February</span> 2019</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=India+Today&amp;rft.atitle=At+98%2C+two-time+interim+PM+Gulzarilal+Nanda+is+the+epitome+of+Gandhian+ideals&amp;rft.date=1996-05-15&amp;rft.aulast=Mahurkar&amp;rft.aufirst=Uday&amp;rft_id=https%3A%2F%2Fwww.indiatoday.in%2Fmagazine%2Fprofile%2Fstory%2F19960515-at-98-two-time-interim-pm-gulzarilal-nanda-is-the-epitome-of-gandhian-ideals-834853-1996-05-15&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-3"><span class="mw-cite-backlink"><b><a href="#cite_ref-3">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite id="CITEREFMalhotra1995" class="citation news cs1">Malhotra, Inder (15 January 1995). <a rel="nofollow" class="external text" href="https://www.indiatoday.in/magazine/society-the-arts/story/19950115-book-review-lal-bahadur-shastri-prime-minister-of-india-1964-66-a-life-of-truth-in-politics-806851-1995-01-15">"Book review: Lal Bahadur Shastri Prime Minister of India 1964-66: A Life of Truth in Politics"</a>. <i>India Today</i><span class="reference-accessdate">. Retrieved <span class="nowrap">4 February</span> 2019</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=India+Today&amp;rft.atitle=Book+review%3A+Lal+Bahadur+Shastri+Prime+Minister+of+India+1964-66%3A+A+Life+of+Truth+in+Politics&amp;rft.date=1995-01-15&amp;rft.aulast=Malhotra&amp;rft.aufirst=Inder&amp;rft_id=https%3A%2F%2Fwww.indiatoday.in%2Fmagazine%2Fsociety-the-arts%2Fstory%2F19950115-book-review-lal-bahadur-shastri-prime-minister-of-india-1964-66-a-life-of-truth-in-politics-806851-1995-01-15&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-indira-4"><span class="mw-cite-backlink">^ <a href="#cite_ref-indira_4-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-indira_4-1"><sup><i><b>b</b></i></sup></a></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite id="CITEREFVijaykumar2017" class="citation news cs1">Vijaykumar, Neeti (19 January 2017). <a rel="nofollow" class="external text" href="https://www.theweek.in/content/archival/news/india/1966-indira-gandhi.html">"Today in 1966: Indira Gandhi becomes Prime Minister"</a>. <i><a href="/wiki/The_Week_(Indian_magazine)" title="The Week (Indian magazine)">The Week</a></i>. <a rel="nofollow" class="external text" href="https://web.archive.org/web/20180216132208/http://www.theweek.in/content/archival/news/india/1966-indira-gandhi.html">Archived</a> from the original on 16 February 2018<span class="reference-accessdate">. Retrieved <span class="nowrap">4 February</span> 2019</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=The+Week&amp;rft.atitle=Today+in+1966%3A+Indira+Gandhi+becomes+Prime+Minister&amp;rft.date=2017-01-19&amp;rft.aulast=Vijaykumar&amp;rft.aufirst=Neeti&amp;rft_id=https%3A%2F%2Fwww.theweek.in%2Fcontent%2Farchival%2Fnews%2Findia%2F1966-indira-gandhi.html&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-5"><span class="mw-cite-backlink"><b><a href="#cite_ref-5">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite class="citation news cs1"><a rel="nofollow" class="external text" href="https://www.rediff.com/news/report/slide-show-1-ls-election-before-modi-there-was-morarjibhai/20140407.htm">"Before Modi, there was Morarjibhai"</a>. <i><a href="/wiki/Rediff.com" title="Rediff.com">Rediff.com</a></i>. 7 April 2014. <a rel="nofollow" class="external text" href="https://web.archive.org/web/20180330222007/http://www.rediff.com/news/report/slide-show-1-ls-election-before-modi-there-was-morarjibhai/20140407.htm">Archived</a> from the original on 30 March 2018<span class="reference-accessdate">. Retrieved <span class="nowrap">4 February</span> 2019</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Rediff.com&amp;rft.atitle=Before+Modi%2C+there+was+Morarjibhai&amp;rft.date=2014-04-07&amp;rft_id=https%3A%2F%2Fwww.rediff.com%2Fnews%2Freport%2Fslide-show-1-ls-election-before-modi-there-was-morarjibhai%2F20140407.htm&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-6"><span class="mw-cite-backlink"><b><a href="#cite_ref-6">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite class="citation news cs1"><a rel="nofollow" class="external text" href="https://economictimes.indiatimes.com/news/politics-and-nation/jd-u-demands-bharat-ratna-to-former-pm-charan-singh/articleshow/50272488.cms">"JD-U demands Bharat Ratna to former PM Charan Singh"</a>. <i><a href="/wiki/The_Economic_Times" title="The Economic Times">The Economic Times</a></i>. 21 December 2015<span class="reference-accessdate">. Retrieved <span class="nowrap">4 February</span> 2018</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=The+Economic+Times&amp;rft.atitle=JD-U+demands+Bharat+Ratna+to+former+PM+Charan+Singh&amp;rft.date=2015-12-21&amp;rft_id=https%3A%2F%2Feconomictimes.indiatimes.com%2Fnews%2Fpolitics-and-nation%2Fjd-u-demands-bharat-ratna-to-former-pm-charan-singh%2Farticleshow%2F50272488.cms&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-7"><span class="mw-cite-backlink"><b><a href="#cite_ref-7">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite id="CITEREFDenyer2011" class="citation news cs1">Denyer, Simon (2 December 2011). <a rel="nofollow" class="external text" href="https://www.washingtonpost.com/world/asia_pacific/in-india-next-generation-of-gandhi-dynasty/2011/11/17/gIQA28SdMO_story.html">"In India, next generation of Gandhi dynasty"</a>. <i>The Washington Post</i>. <a rel="nofollow" class="external text" href="https://web.archive.org/web/20161228195140/https://www.washingtonpost.com/world/asia_pacific/in-india-next-generation-of-gandhi-dynasty/2011/11/17/gIQA28SdMO_story.html">Archived</a> from the original on 28 December 2016<span class="reference-accessdate">. Retrieved <span class="nowrap">27 December</span> 2016</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=The+Washington+Post&amp;rft.atitle=In+India%2C+next+generation+of+Gandhi+dynasty&amp;rft.date=2011-12-02&amp;rft.aulast=Denyer&amp;rft.aufirst=Simon&amp;rft_id=https%3A%2F%2Fwww.washingtonpost.com%2Fworld%2Fasia_pacific%2Fin-india-next-generation-of-gandhi-dynasty%2F2011%2F11%2F17%2FgIQA28SdMO_story.html&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-former-8"><span class="mw-cite-backlink">^ <a href="#cite_ref-former_8-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-former_8-1"><sup><i><b>b</b></i></sup></a></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite id="CITEREFIype2004" class="citation web cs1">Iype, George (3 May 2004). <a rel="nofollow" class="external text" href="https://www.rediff.com/election/2004/may/03espec1.htm">"What the former PMs are doing"</a>. <i>Rediff.com</i>. <a rel="nofollow" class="external text" href="https://web.archive.org/web/20100325091259/http://www.rediff.com/election/2004/may/03espec1.htm">Archived</a> from the original on 25 March 2010<span class="reference-accessdate">. Retrieved <span class="nowrap">4 February</span> 2019</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=unknown&amp;rft.jtitle=Rediff.com&amp;rft.atitle=What+the+former+PMs+are+doing&amp;rft.date=2004-05-03&amp;rft.aulast=Iype&amp;rft.aufirst=George&amp;rft_id=https%3A%2F%2Fwww.rediff.com%2Felection%2F2004%2Fmay%2F03espec1.htm&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-9"><span class="mw-cite-backlink"><b><a href="#cite_ref-9">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite id="CITEREFGhosh2018" class="citation news cs1">Ghosh, Deepshikha (16 August 2018). <a rel="nofollow" class="external text" href="https://www.ndtv.com/india-news/atal-bihari-vajpayee-dies-the-3-time-pm-who-captivated-india-with-his-oratory-1901509">"Atal Bihari Vajpayee: The 3-Time PM Who Captivated India With His Oratory"</a>. <i><a href="/wiki/NDTV" title="NDTV">NDTV</a></i>. <a rel="nofollow" class="external text" href="https://web.archive.org/web/20181223151524/https://www.ndtv.com/india-news/atal-bihari-vajpayee-dies-the-3-time-pm-who-captivated-india-with-his-oratory-1901509">Archived</a> from the original on 23 December 2018<span class="reference-accessdate">. Retrieved <span class="nowrap">4 February</span> 2019</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=NDTV&amp;rft.atitle=Atal+Bihari+Vajpayee%3A+The+3-Time+PM+Who+Captivated+India+With+His+Oratory&amp;rft.date=2018-08-16&amp;rft.aulast=Ghosh&amp;rft.aufirst=Deepshikha&amp;rft_id=https%3A%2F%2Fwww.ndtv.com%2Findia-news%2Fatal-bihari-vajpayee-dies-the-3-time-pm-who-captivated-india-with-his-oratory-1901509&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-10"><span class="mw-cite-backlink"><b><a href="#cite_ref-10">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite class="citation news cs1"><a rel="nofollow" class="external text" href="https://www.outlookindia.com/website/story/pm-modi-rahul-gandhi-greet-manmohan-singh-on-his-86th-birthday/317198">"PM Modi, Rahul Gandhi Greet Manmohan Singh On His 86th Birthday"</a>. <i><a href="/wiki/Outlook_(Indian_magazine)" title="Outlook (Indian magazine)">Outlook</a></i>. 26 September 2018. <a rel="nofollow" class="external text" href="https://web.archive.org/web/20180928170607/https://www.outlookindia.com/website/story/pm-modi-rahul-gandhi-greet-manmohan-singh-on-his-86th-birthday/317198">Archived</a> from the original on 28 September 2018<span class="reference-accessdate">. Retrieved <span class="nowrap">4 February</span> 2019</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Outlook&amp;rft.atitle=PM+Modi%2C+Rahul+Gandhi+Greet+Manmohan+Singh+On+His+86th+Birthday&amp;rft.date=2018-09-26&amp;rft_id=https%3A%2F%2Fwww.outlookindia.com%2Fwebsite%2Fstory%2Fpm-modi-rahul-gandhi-greet-manmohan-singh-on-his-86th-birthday%2F317198&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-11"><span class="mw-cite-backlink"><b><a href="#cite_ref-11">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite id="CITEREFPanda2014" class="citation news cs1">Panda, Ankit (16 May 2014). <a rel="nofollow" class="external text" href="https://thediplomat.com/2014/05/bjp-modi-win-landslide-victory-in-indian-elections/">"BJP, Modi Win Landslide Victory in Indian Elections"</a>. <a href="/wiki/The_Diplomat" title="The Diplomat">The Diplomat</a>. <a rel="nofollow" class="external text" href="https://web.archive.org/web/20161221054315/https://thediplomat.com/2014/05/bjp-modi-win-landslide-victory-in-indian-elections//">Archived</a> from the original on 21 December 2016<span class="reference-accessdate">. Retrieved <span class="nowrap">27 December</span> 2016</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.atitle=BJP%2C+Modi+Win+Landslide+Victory+in+Indian+Elections&amp;rft.date=2014-05-16&amp;rft.aulast=Panda&amp;rft.aufirst=Ankit&amp;rft_id=https%3A%2F%2Fthediplomat.com%2F2014%2F05%2Fbjp-modi-win-landslide-victory-in-indian-elections%2F&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-13"><span class="mw-cite-backlink"><b><a href="#cite_ref-13">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite class="citation book cs1"><a rel="nofollow" class="external text" href="https://web.archive.org/web/20140718175926/http://eci.nic.in/eci_main/StatisticalReports/LS_1980/Vol_I_LS_80.pdf"><i>Statistical Report on General Elections, 1980 to the Seventh Lok Sabha</i></a> <span class="cs1-format">(PDF)</span>. New Delhi: <a href="/wiki/Election_Commission_of_India" title="Election Commission of India">Election Commission of India</a>. p.&#160;1. Archived from <a rel="nofollow" class="external text" href="http://eci.nic.in/eci_main/StatisticalReports/LS_1980/Vol_I_LS_80.pdf">the original</a> <span class="cs1-format">(PDF)</span> on 18 July 2014<span class="reference-accessdate">. Retrieved <span class="nowrap">20 May</span> 2020</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=book&amp;rft.btitle=Statistical+Report+on+General+Elections%2C+1980+to+the+Seventh+Lok+Sabha&amp;rft.place=New+Delhi&amp;rft.pages=1&amp;rft.pub=Election+Commission+of+India&amp;rft_id=http%3A%2F%2Feci.nic.in%2Feci_main%2FStatisticalReports%2FLS_1980%2FVol_I_LS_80.pdf&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
<li id="cite_note-16"><span class="mw-cite-backlink"><b><a href="#cite_ref-16">^</a></b></span> <span class="reference-text"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1067248974"/><cite class="citation web cs1"><a rel="nofollow" class="external text" href="https://web.archive.org/web/20141009232119/http://pmindia.gov.in/en/former-prime-ministers/">"Former Prime Ministers"</a>. PM India. Archived from <a rel="nofollow" class="external text" href="http://pmindia.gov.in/en/former-prime-ministers/">the original</a> on 9 October 2014<span class="reference-accessdate">. Retrieved <span class="nowrap">2 January</span> 2015</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=Former+Prime+Ministers&amp;rft.pub=PM+India&amp;rft_id=http%3A%2F%2Fpmindia.gov.in%2Fen%2Fformer-prime-ministers%2F&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AList+of+prime+ministers+of+India" class="Z3988"></span></span>
</li>
</ol></div></div>
<h2><span class="mw-headline" id="External_links">External links</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit&amp;section=7" title="Edit section: External links">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<ul><li><a rel="nofollow" class="external text" href="https://www.pmindia.gov.in/en/">Prime Minister of India</a></li></ul>
<div class="navbox-styles nomobile"><style data-mw-deduplicate="TemplateStyles:r1061467846">.mw-parser-output .navbox{box-sizing:border-box;border:1px solid #a2a9b1;width:100%;clear:both;font-size:88%;text-align:center;padding:1px;margin:1em auto 0}.mw-parser-output .navbox .navbox{margin-top:0}.mw-parser-output .navbox+.navbox,.mw-parser-output .navbox+.navbox-styles+.navbox{margin-top:-1px}.mw-parser-output .navbox-inner,.mw-parser-output .navbox-subgroup{width:100%}.mw-parser-output .navbox-group,.mw-parser-output .navbox-title,.mw-parser-output .navbox-abovebelow{padding:0.25em 1em;line-height:1.5em;text-align:center}.mw-parser-output .navbox-group{white-space:nowrap;text-align:right}.mw-parser-output .navbox,.mw-parser-output .navbox-subgroup{background-color:#fdfdfd}.mw-parser-output .navbox-list{line-height:1.5em;border-color:#fdfdfd}.mw-parser-output .navbox-list-with-group{text-align:left;border-left-width:2px;border-left-style:solid}.mw-parser-output tr+tr>.navbox-abovebelow,.mw-parser-output tr+tr>.navbox-group,.mw-parser-output tr+tr>.navbox-image,.mw-parser-output tr+tr>.navbox-list{border-top:2px solid #fdfdfd}.mw-parser-output .navbox-title{background-color:#ccf}.mw-parser-output .navbox-abovebelow,.mw-parser-output .navbox-group,.mw-parser-output .navbox-subgroup .navbox-title{background-color:#ddf}.mw-parser-output .navbox-subgroup .navbox-group,.mw-parser-output .navbox-subgroup .navbox-abovebelow{background-color:#e6e6ff}.mw-parser-output .navbox-even{background-color:#f7f7f7}.mw-parser-output .navbox-odd{background-color:transparent}.mw-parser-output .navbox .hlist td dl,.mw-parser-output .navbox .hlist td ol,.mw-parser-output .navbox .hlist td ul,.mw-parser-output .navbox td.hlist dl,.mw-parser-output .navbox td.hlist ol,.mw-parser-output .navbox td.hlist ul{padding:0.125em 0}.mw-parser-output .navbox .navbar{display:block;font-size:100%}.mw-parser-output .navbox-title .navbar{float:left;text-align:left;margin-right:0.5em}</style></div><div role="navigation" class="navbox" aria-labelledby="23x15px&amp;#124;border_&amp;#124;alt=India&amp;#124;link=India_Prime_Minister_of_India" style="padding:3px"><table class="nowraplinks hlist mw-collapsible mw-collapsed navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="3"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1063604349"/><div class="navbar plainlinks hlist navbar-mini"><ul><li class="nv-view"><a href="/wiki/Template:Prime_ministers_of_India" title="Template:Prime ministers of India"><abbr title="View this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">v</abbr></a></li><li class="nv-talk"><a href="/wiki/Template_talk:Prime_ministers_of_India" title="Template talk:Prime ministers of India"><abbr title="Discuss this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">t</abbr></a></li><li class="nv-edit"><a class="external text" href="https://en.wikipedia.org/w/index.php?title=Template:Prime_ministers_of_India&amp;action=edit"><abbr title="Edit this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">e</abbr></a></li></ul></div><div id="23x15px&amp;#124;border_&amp;#124;alt=India&amp;#124;link=India_Prime_Minister_of_India" style="font-size:114%;margin:0 4em"><span class="flagicon"><a href="/wiki/India" title="India"><img alt="India" src="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/23px-Flag_of_India.svg.png" decoding="async" width="23" height="15" class="thumbborder" srcset="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/35px-Flag_of_India.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/45px-Flag_of_India.svg.png 2x" data-file-width="1350" data-file-height="900" /></a></span> <a href="/wiki/Prime_Minister_of_India" title="Prime Minister of India">Prime Minister of India</a></div></th></tr><tr><td class="navbox-abovebelow" colspan="3"><div id="*_List_*_By_education_*_By_previous_experience">
<ul><li><a class="mw-selflink selflink">List</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_India_by_education" title="List of prime ministers of India by education">By education</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_India_by_previous_experience" title="List of prime ministers of India by previous experience">By previous experience</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Prime<br />ministers</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Jawaharlal_Nehru" title="Jawaharlal Nehru">Jawaharlal Nehru</a></li>
<li><a href="/wiki/Lal_Bahadur_Shastri" title="Lal Bahadur Shastri">Lal Bahadur Shastri</a></li>
<li><a href="/wiki/Indira_Gandhi" title="Indira Gandhi">Indira Gandhi</a></li>
<li><a href="/wiki/Morarji_Desai" title="Morarji Desai">Morarji Desai</a></li>
<li><a href="/wiki/Charan_Singh" title="Charan Singh">Charan Singh</a></li>
<li><a href="/wiki/Rajiv_Gandhi" title="Rajiv Gandhi">Rajiv Gandhi</a></li>
<li><a href="/wiki/Vishwanath_Pratap_Singh" title="Vishwanath Pratap Singh">Vishwanath Pratap Singh</a></li>
<li><a href="/wiki/Chandra_Shekhar" title="Chandra Shekhar">Chandra Shekhar</a></li>
<li><a href="/wiki/P._V._Narasimha_Rao" title="P. V. Narasimha Rao">P. V. Narasimha Rao</a></li>
<li><a href="/wiki/Atal_Bihari_Vajpayee" title="Atal Bihari Vajpayee">Atal Bihari Vajpayee</a></li>
<li><a href="/wiki/H._D._Deve_Gowda" title="H. D. Deve Gowda">H. D. Deve Gowda</a></li>
<li><a href="/wiki/Inder_Kumar_Gujral" title="Inder Kumar Gujral">Inder Kumar Gujral</a></li>
<li><a href="/wiki/Manmohan_Singh" title="Manmohan Singh">Manmohan Singh</a></li>
<li><a href="/wiki/Narendra_Modi" title="Narendra Modi">Narendra Modi</a> <span style="font-size:85%;">(incumbent)</span></li></ul>
</div></td><td class="noviewer navbox-image" rowspan="2" style="width:1px;padding:0 0 0 2px"><div><a href="/wiki/File:Emblem_of_India.svg" class="image" title="Emblem of India"><img alt="Emblem of India" src="//upload.wikimedia.org/wikipedia/commons/thumb/5/55/Emblem_of_India.svg/31px-Emblem_of_India.svg.png" decoding="async" width="31" height="50" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/5/55/Emblem_of_India.svg/46px-Emblem_of_India.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/5/55/Emblem_of_India.svg/61px-Emblem_of_India.svg.png 2x" data-file-width="370" data-file-height="600" /></a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Acting</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Gulzarilal_Nanda" title="Gulzarilal Nanda">Gulzarilal Nanda</a></li></ul>
</div></td></tr><tr><td class="navbox-abovebelow" colspan="3"><div>
<ul><li><a href="/wiki/List_of_assassination_attempts_on_the_prime_ministers_of_India" title="List of assassination attempts on the prime ministers of India">Assassination attempts</a></li>
<li><a href="/wiki/List_of_children_of_the_prime_ministers_of_India" title="List of children of the prime ministers of India">Children</a></li>
<li><a href="/wiki/List_of_things_named_after_the_prime_ministers_of_India" title="List of things named after the prime ministers of India">Eponyms</a>
<ul><li><a href="/wiki/List_of_things_named_after_Jawaharlal_Nehru" title="List of things named after Jawaharlal Nehru">Nehru</a></li>
<li><a href="/wiki/List_of_things_named_after_Indira_Gandhi" title="List of things named after Indira Gandhi">Indira</a></li>
<li><a href="/wiki/List_of_things_named_after_Rajiv_Gandhi" title="List of things named after Rajiv Gandhi">Rajiv</a></li></ul></li>
<li><a href="/wiki/List_of_international_trips_made_by_the_prime_ministers_of_India" title="List of international trips made by the prime ministers of India">International trips</a></li>
<li><a href="/wiki/List_of_nicknames_of_the_prime_ministers_of_India" title="List of nicknames of the prime ministers of India">Nicknames</a></li>
<li><a href="/wiki/Prime_Minister%27s_Office_(India)" title="Prime Minister&#39;s Office (India)">Office</a></li>
<li><a href="/wiki/Air_India_One" title="Air India One">Official aircraft</a></li>
<li><a href="/wiki/7,_Lok_Kalyan_Marg" title="7, Lok Kalyan Marg">Official residence</a></li>
<li><a href="/wiki/Spouse_of_the_prime_minister_of_India" title="Spouse of the prime minister of India">Spouse</a></li></ul>
</div></td></tr></tbody></table></div>
<div class="navbox-styles nomobile"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1061467846"/></div><div role="navigation" class="navbox" aria-labelledby="20x20px_Indian_union_ministries" style="padding:3px"><table class="nowraplinks mw-collapsible autocollapse navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1063604349"/><div class="navbar plainlinks hlist navbar-mini"><ul><li class="nv-view"><a href="/wiki/Template:Indian_union_ministries" title="Template:Indian union ministries"><abbr title="View this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">v</abbr></a></li><li class="nv-talk"><a href="/wiki/Template_talk:Indian_union_ministries" title="Template talk:Indian union ministries"><abbr title="Discuss this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">t</abbr></a></li><li class="nv-edit"><a class="external text" href="https://en.wikipedia.org/w/index.php?title=Template:Indian_union_ministries&amp;action=edit"><abbr title="Edit this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">e</abbr></a></li></ul></div><div id="20x20px_Indian_union_ministries" style="font-size:114%;margin:0 4em"><span style="vertical-align:1px;"><a href="/wiki/File:Flag_of_India.svg" class="image"><img alt="Flag of India.svg" src="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/20px-Flag_of_India.svg.png" decoding="async" width="20" height="13" srcset="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/30px-Flag_of_India.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/40px-Flag_of_India.svg.png 2x" data-file-width="1350" data-file-height="900" /></a></span> <a href="/wiki/List_of_Indian_union_ministries" title="List of Indian union ministries">Indian union ministries</a></div></th></tr><tr><td colspan="2" class="navbox-list navbox-odd hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><i><a href="/wiki/Interim_Government_of_India#Cabinet_of_the_Interim_Government_of_India" title="Interim Government of India">Interim</a></i></li>
<li><a href="/wiki/First_Nehru_ministry" title="First Nehru ministry">Nehru I</a></li>
<li><a href="/wiki/Second_Nehru_ministry" title="Second Nehru ministry">Nehru II</a></li>
<li><a href="/wiki/Third_Nehru_ministry" title="Third Nehru ministry">Nehru III</a></li>
<li><a href="/wiki/Fourth_Nehru_ministry" title="Fourth Nehru ministry">Nehru IV</a></li>
<li><i><a href="/wiki/First_Nanda_ministry" title="First Nanda ministry">Nanda I</a></i></li>
<li><a href="/wiki/Shastri_ministry" title="Shastri ministry">Shastri</a></li>
<li><i><a href="/wiki/Second_Nanda_ministry" title="Second Nanda ministry">Nanda II</a></i></li>
<li><a href="/wiki/First_Indira_Gandhi_ministry" title="First Indira Gandhi ministry">Indira I</a></li>
<li><a href="/wiki/Second_Indira_Gandhi_ministry" title="Second Indira Gandhi ministry">Indira II</a></li>
<li><a href="/wiki/Premiership_of_Morarji_Desai#Morarji_Desai_ministry" title="Premiership of Morarji Desai">Desai</a></li>
<li><a href="/wiki/Charan_Singh_ministry" title="Charan Singh ministry">Charan</a></li>
<li><a href="/wiki/Third_Indira_Gandhi_ministry" title="Third Indira Gandhi ministry">Indira III</a></li>
<li><a href="/wiki/Rajiv_Gandhi_ministry" title="Rajiv Gandhi ministry">Rajiv</a></li>
<li><a href="/wiki/Vishwanath_Pratap_Singh_ministry" title="Vishwanath Pratap Singh ministry">Vishwanath</a></li>
<li><a href="/wiki/Chandra_Shekhar_ministry" title="Chandra Shekhar ministry">Chandra Shekhar</a></li>
<li><a href="/wiki/Rao_ministry" title="Rao ministry">Rao</a></li>
<li><a href="/wiki/First_Vajpayee_ministry" title="First Vajpayee ministry">Vajpayee I</a></li>
<li><a href="/wiki/Deve_Gowda_ministry" title="Deve Gowda ministry">Deve Gowda</a></li>
<li><a href="/wiki/Gujral_ministry" title="Gujral ministry">Gujral</a></li>
<li><a href="/wiki/Second_Vajpayee_ministry" title="Second Vajpayee ministry">Vajpayee II</a></li>
<li><a href="/wiki/Third_Vajpayee_ministry" title="Third Vajpayee ministry">Vajpayee III</a></li>
<li><a href="/wiki/First_Manmohan_Singh_ministry" title="First Manmohan Singh ministry">Manmohan I</a></li>
<li><a href="/wiki/Second_Manmohan_Singh_ministry" title="Second Manmohan Singh ministry">Manmohan II</a></li>
<li><a href="/wiki/First_Modi_ministry" title="First Modi ministry">Modi I</a></li>
<li><a href="/wiki/Second_Modi_ministry" title="Second Modi ministry">Modi II</a></li></ul>
</div></td></tr></tbody></table></div>
<div class="navbox-styles nomobile"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1061467846"/></div><div role="navigation" class="navbox" aria-labelledby="Lists_of_prime_ministers" style="padding:3px"><table class="nowraplinks hlist mw-collapsible autocollapse navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1063604349"/><div class="navbar plainlinks hlist navbar-mini"><ul><li class="nv-view"><a href="/wiki/Template:Lists_of_prime_ministers" title="Template:Lists of prime ministers"><abbr title="View this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">v</abbr></a></li><li class="nv-talk"><a href="/wiki/Template_talk:Lists_of_prime_ministers" title="Template talk:Lists of prime ministers"><abbr title="Discuss this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">t</abbr></a></li><li class="nv-edit"><a class="external text" href="https://en.wikipedia.org/w/index.php?title=Template:Lists_of_prime_ministers&amp;action=edit"><abbr title="Edit this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">e</abbr></a></li></ul></div><div id="Lists_of_prime_ministers" style="font-size:114%;margin:0 4em"><a href="/wiki/Prime_minister" title="Prime minister">Lists of prime ministers</a></div></th></tr><tr><th scope="row" class="navbox-group" style="width:1%">Lists of prime ministers by country</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/List_of_Prime_Ministers_of_Abkhazia" class="mw-redirect" title="List of Prime Ministers of Abkhazia">Abkhazia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Albania" class="mw-redirect" title="List of Prime Ministers of Albania">Albania</a></li>
<li><a href="/wiki/List_of_heads_of_government_of_Algeria" title="List of heads of government of Algeria">Algeria</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Antigua_and_Barbuda" class="mw-redirect" title="List of Prime Ministers of Antigua and Barbuda">Antigua and Barbuda</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Australia" class="mw-redirect" title="List of Prime Ministers of Australia">Australia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_Bahamas" class="mw-redirect" title="List of Prime Ministers of the Bahamas">Bahamas</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Bangladesh" class="mw-redirect" title="List of Prime Ministers of Bangladesh">Bangladesh</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Barbados" class="mw-redirect" title="List of Prime Ministers of Barbados">Barbados</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Belarus" class="mw-redirect" title="List of Prime Ministers of Belarus">Belarus</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Belgium" class="mw-redirect" title="List of Prime Ministers of Belgium">Belgium</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Belize" class="mw-redirect" title="List of Prime Ministers of Belize">Belize</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Bhutan" class="mw-redirect" title="List of Prime Ministers of Bhutan">Bhutan</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Bosnia_and_Herzegovina" class="mw-redirect" title="List of Prime Ministers of Bosnia and Herzegovina">Bosnia and Herzegovina</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Bulgaria" class="mw-redirect" title="List of Prime Ministers of Bulgaria">Bulgaria</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Burkina_Faso" class="mw-redirect" title="List of Prime Ministers of Burkina Faso">Burkina Faso</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Cambodia" class="mw-redirect" title="List of Prime Ministers of Cambodia">Cambodia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Canada" class="mw-redirect" title="List of Prime Ministers of Canada">Canada</a></li>
<li><a href="/wiki/List_of_Prime_ministers_of_the_Democratic_Republic_of_the_Congo" class="mw-redirect" title="List of Prime ministers of the Democratic Republic of the Congo">Congo</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Croatia" class="mw-redirect" title="List of Prime Ministers of Croatia">Croatia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Cuba" class="mw-redirect" title="List of Prime Ministers of Cuba">Cuba</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_Czech_Republic" class="mw-redirect" title="List of Prime Ministers of the Czech Republic">Czech Republic</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Czechoslovakia" class="mw-redirect" title="List of Prime Ministers of Czechoslovakia">Czechoslovakia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Denmark" class="mw-redirect" title="List of Prime Ministers of Denmark">Denmark</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Djibouti" class="mw-redirect" title="List of Prime Ministers of Djibouti">Djibouti</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Dominica" class="mw-redirect" title="List of Prime Ministers of Dominica">Dominica</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Egypt" class="mw-redirect" title="List of Prime Ministers of Egypt">Egypt</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Equatorial_Guinea" class="mw-redirect" title="List of Prime Ministers of Equatorial Guinea">Equatorial Guinea</a></li>
<li><a href="/wiki/Prime_Minister_of_Fiji" title="Prime Minister of Fiji">Fiji</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Finland" class="mw-redirect" title="List of Prime Ministers of Finland">Finland</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_France" class="mw-redirect" title="List of Prime Ministers of France">France</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Ghana" class="mw-redirect" title="List of Prime Ministers of Ghana">Ghana</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Great_Britain" class="mw-redirect" title="List of Prime Ministers of Great Britain">Great Britain</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_Greece" title="List of prime ministers of Greece">Greece</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Greenland" class="mw-redirect" title="List of Prime Ministers of Greenland">Greenland</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Grenada" class="mw-redirect" title="List of Prime Ministers of Grenada">Grenada</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Guinea" class="mw-redirect" title="List of Prime Ministers of Guinea">Guinea</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Guyana" class="mw-redirect" title="List of Prime Ministers of Guyana">Guyana</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Haiti" class="mw-redirect" title="List of Prime Ministers of Haiti">Haiti</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Hungary" class="mw-redirect" title="List of Prime Ministers of Hungary">Hungary</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Iceland" class="mw-redirect" title="List of Prime Ministers of Iceland">Iceland</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_India" class="mw-redirect" title="List of Prime Ministers of India">India</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Indonesia" class="mw-redirect" title="List of Prime Ministers of Indonesia">Indonesia</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_Iran" title="List of prime ministers of Iran">Iran</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_Iraq" title="List of prime ministers of Iraq">Iraq</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Ireland" class="mw-redirect" title="List of Prime Ministers of Ireland">Ireland</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Israel" class="mw-redirect" title="List of Prime Ministers of Israel">Israel</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Italy" class="mw-redirect" title="List of Prime Ministers of Italy">Italy</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Jamaica" class="mw-redirect" title="List of Prime Ministers of Jamaica">Jamaica</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Jammu_and_Kashmir" class="mw-redirect" title="List of Prime Ministers of Jammu and Kashmir">Jammu and Kashmir</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_Japan" title="List of prime ministers of Japan">Japan</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_Jordan" title="List of prime ministers of Jordan">Jordan</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Kampuchea" class="mw-redirect" title="List of Prime Ministers of Kampuchea">Kampuchea</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Kazakhstan" class="mw-redirect" title="List of Prime Ministers of Kazakhstan">Kazakhstan</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Korea" class="mw-redirect" title="List of Prime Ministers of Korea">Korea</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Kuwait" class="mw-redirect" title="List of Prime Ministers of Kuwait">Kuwait</a></li>
<li><a href="/wiki/List_of_prime_ministers_of_Lebanon" title="List of prime ministers of Lebanon">Lebanon</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Libya" class="mw-redirect" title="List of Prime Ministers of Libya">Libya</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Luxembourg" class="mw-redirect" title="List of Prime Ministers of Luxembourg">Luxembourg</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Madagascar" class="mw-redirect" title="List of Prime Ministers of Madagascar">Madagascar</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Malaysia" class="mw-redirect" title="List of Prime Ministers of Malaysia">Malaysia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Malta" class="mw-redirect" title="List of Prime Ministers of Malta">Malta</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Mauritania" class="mw-redirect" title="List of Prime Ministers of Mauritania">Mauritania</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Mauritius" class="mw-redirect" title="List of Prime Ministers of Mauritius">Mauritius</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Moldova" class="mw-redirect" title="List of Prime Ministers of Moldova">Moldova</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Mongolia" class="mw-redirect" title="List of Prime Ministers of Mongolia">Mongolia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Montenegro" class="mw-redirect" title="List of Prime Ministers of Montenegro">Montenegro</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Morocco" class="mw-redirect" title="List of Prime Ministers of Morocco">Morocco</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Mozambique" class="mw-redirect" title="List of Prime Ministers of Mozambique">Mozambique</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Nagorno-Karabakh" class="mw-redirect" title="List of Prime Ministers of Nagorno-Karabakh">Nagorno-Karabakh</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Nepal" class="mw-redirect" title="List of Prime Ministers of Nepal">Nepal</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_Netherlands" class="mw-redirect" title="List of Prime Ministers of the Netherlands">Netherlands</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_Netherlands_Antilles" class="mw-redirect" title="List of Prime Ministers of the Netherlands Antilles">Netherlands Antilles</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Newfoundland" class="mw-redirect" title="List of Prime Ministers of Newfoundland">Newfoundland</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_New_Zealand" class="mw-redirect" title="List of Prime Ministers of New Zealand">New Zealand</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Niger" class="mw-redirect" title="List of Prime Ministers of Niger">Niger</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_North_Korea" class="mw-redirect" title="List of Prime Ministers of North Korea">North Korea</a></li>
<li><a href="/wiki/Prime_Minister_of_North_Macedonia#List_of_prime_ministers_of_North_Macedonia" title="Prime Minister of North Macedonia">North Macedonia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Northern_Cyprus" class="mw-redirect" title="List of Prime Ministers of Northern Cyprus">Northern Cyprus</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Northern_Ireland" class="mw-redirect" title="List of Prime Ministers of Northern Ireland">Northern Ireland</a></li>
<li><a href="/wiki/List_of_Prime_ministers_of_North_Rhine-Westphalia" class="mw-redirect" title="List of Prime ministers of North Rhine-Westphalia">North Rhine-Westphalia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Norway" class="mw-redirect" title="List of Prime Ministers of Norway">Norway</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Pakistan" class="mw-redirect" title="List of Prime Ministers of Pakistan">Pakistan</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Palestine" class="mw-redirect" title="List of Prime Ministers of Palestine">Palestine</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Papua_New_Guinea" class="mw-redirect" title="List of Prime Ministers of Papua New Guinea">Papua New Guinea</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Peru" class="mw-redirect" title="List of Prime Ministers of Peru">Peru</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_Philippines" class="mw-redirect" title="List of Prime Ministers of the Philippines">Philippines</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Poland" class="mw-redirect" title="List of Prime Ministers of Poland">Poland</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Portugal" class="mw-redirect" title="List of Prime Ministers of Portugal">Portugal</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Prussia" class="mw-redirect" title="List of Prime Ministers of Prussia">Prussia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Qatar" class="mw-redirect" title="List of Prime Ministers of Qatar">Qatar</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Queen_Elizabeth_II" class="mw-redirect" title="List of Prime Ministers of Queen Elizabeth II">Queen Elizabeth II</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Queen_Victoria" class="mw-redirect" title="List of Prime Ministers of Queen Victoria">Queen Victoria</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Republika_Srpska" class="mw-redirect" title="List of Prime Ministers of Republika Srpska">Republika Srpska</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Rhodesia" class="mw-redirect" title="List of Prime Ministers of Rhodesia">Rhodesia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Romania" class="mw-redirect" title="List of Prime Ministers of Romania">Romania</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Russia" class="mw-redirect" title="List of Prime Ministers of Russia">Russia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Rwanda" class="mw-redirect" title="List of Prime Ministers of Rwanda">Rwanda</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Saint_Kitts_and_Nevis" class="mw-redirect" title="List of Prime Ministers of Saint Kitts and Nevis">Saint Kitts and Nevis</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Saint_Lucia" class="mw-redirect" title="List of Prime Ministers of Saint Lucia">Saint Lucia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Saint_Vincent_and_the_Grenadines" class="mw-redirect" title="List of Prime Ministers of Saint Vincent and the Grenadines">Saint Vincent and the Grenadines</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Samoa" class="mw-redirect" title="List of Prime Ministers of Samoa">Samoa</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_S%C3%A3o_Tom%C3%A9_and_Pr%C3%ADncipe" class="mw-redirect" title="List of Prime Ministers of São Tomé and Príncipe">São Tomé and Príncipe</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Serbia" class="mw-redirect" title="List of Prime Ministers of Serbia">Serbia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_Republic_of_Serbian_Krajina" class="mw-redirect" title="List of Prime Ministers of the Republic of Serbian Krajina">Serbian Krajina</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Sierra_Leone" class="mw-redirect" title="List of Prime Ministers of Sierra Leone">Sierra Leone</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Singapore" class="mw-redirect" title="List of Prime Ministers of Singapore">Singapore</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Slovakia" class="mw-redirect" title="List of Prime Ministers of Slovakia">Slovakia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_Solomon_Islands" class="mw-redirect" title="List of Prime Ministers of the Solomon Islands">Solomon Islands</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Somalia" class="mw-redirect" title="List of Prime Ministers of Somalia">Somalia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_South_Africa" class="mw-redirect" title="List of Prime Ministers of South Africa">South Africa</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_South_Korea" class="mw-redirect" title="List of Prime Ministers of South Korea">South Korea</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_South_Vietnam" class="mw-redirect" title="List of Prime Ministers of South Vietnam">South Vietnam</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_Soviet_Union" class="mw-redirect" title="List of Prime Ministers of the Soviet Union">Soviet Union</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Spain" class="mw-redirect" title="List of Prime Ministers of Spain">Spain</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Sri_Lanka" class="mw-redirect" title="List of Prime Ministers of Sri Lanka">Sri Lanka</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Sudan" class="mw-redirect" title="List of Prime Ministers of Sudan">Sudan</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Suriname" class="mw-redirect" title="List of Prime Ministers of Suriname">Suriname</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Swaziland" class="mw-redirect" title="List of Prime Ministers of Swaziland">Swaziland</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Sweden" class="mw-redirect" title="List of Prime Ministers of Sweden">Sweden</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Syria" class="mw-redirect" title="List of Prime Ministers of Syria">Syria</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Tajikistan" class="mw-redirect" title="List of Prime Ministers of Tajikistan">Tajikistan</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Tanganyika" class="mw-redirect" title="List of Prime Ministers of Tanganyika">Tanganyika</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Tanzania" class="mw-redirect" title="List of Prime Ministers of Tanzania">Tanzania</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Tatarstan" class="mw-redirect" title="List of Prime Ministers of Tatarstan">Tatarstan</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Thailand" class="mw-redirect" title="List of Prime Ministers of Thailand">Thailand</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Trinidad_and_Tobago" class="mw-redirect" title="List of Prime Ministers of Trinidad and Tobago">Trinidad and Tobago</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Tunisia" class="mw-redirect" title="List of Prime Ministers of Tunisia">Tunisia</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Turkey" class="mw-redirect" title="List of Prime Ministers of Turkey">Turkey</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Uganda" class="mw-redirect" title="List of Prime Ministers of Uganda">Uganda</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Ukraine" class="mw-redirect" title="List of Prime Ministers of Ukraine">Ukraine</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_United_Arab_Emirates" class="mw-redirect" title="List of Prime Ministers of the United Arab Emirates">United Arab Emirates</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_the_United_Kingdom" class="mw-redirect" title="List of Prime Ministers of the United Kingdom">United Kingdom</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Vietnam" class="mw-redirect" title="List of Prime Ministers of Vietnam">Vietnam</a></li>
<li><a href="/wiki/List_of_Prime_Ministers_of_Yugoslavia" class="mw-redirect" title="List of Prime Ministers of Yugoslavia">Yugoslavia</a></li></ul>
</div></td></tr></tbody></table></div>
<div class="navbox-styles nomobile"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1061467846"/></div><div role="navigation" class="navbox" aria-labelledby="23x15px&amp;#124;border_&amp;#124;alt=India&amp;#124;link=India_Government_ministries_of_India" style="padding:3px"><table class="nowraplinks mw-collapsible autocollapse navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1063604349"/><div class="navbar plainlinks hlist navbar-mini"><ul><li class="nv-view"><a href="/wiki/Template:Union_ministries_of_India" title="Template:Union ministries of India"><abbr title="View this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">v</abbr></a></li><li class="nv-talk"><a href="/wiki/Template_talk:Union_ministries_of_India" title="Template talk:Union ministries of India"><abbr title="Discuss this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">t</abbr></a></li><li class="nv-edit"><a class="external text" href="https://en.wikipedia.org/w/index.php?title=Template:Union_ministries_of_India&amp;action=edit"><abbr title="Edit this template" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">e</abbr></a></li></ul></div><div id="23x15px&amp;#124;border_&amp;#124;alt=India&amp;#124;link=India_Government_ministries_of_India" style="font-size:114%;margin:0 4em"><span class="flagicon"><a href="/wiki/India" title="India"><img alt="India" src="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/23px-Flag_of_India.svg.png" decoding="async" width="23" height="15" class="thumbborder" srcset="//upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/35px-Flag_of_India.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/45px-Flag_of_India.svg.png 2x" data-file-width="1350" data-file-height="900" /></a></span> <a href="/wiki/Union_government_ministries_of_India" title="Union government ministries of India">Government ministries of India</a></div></th></tr><tr><td class="navbox-abovebelow" colspan="2"><div id="President_(List)_•_Vice_President_(List)_•_Government_of_India_•_Prime_Minister_(List_•_Office)_•_Deputy_Prime_Minister_•_Council_of_Ministers"><a href="/wiki/President_of_India" title="President of India">President</a> (<a href="/wiki/List_of_presidents_of_India" title="List of presidents of India">List</a>) • <a href="/wiki/Vice_President_of_India" title="Vice President of India">Vice President</a> (<a href="/wiki/List_of_vice_presidents_of_India" title="List of vice presidents of India">List</a>) • <a href="/wiki/Government_of_India" title="Government of India">Government of India</a> • <a href="/wiki/Prime_Minister_of_India" title="Prime Minister of India">Prime Minister</a> (<a href="/wiki/List_of_Prime_Ministers_of_India" class="mw-redirect" title="List of Prime Ministers of India">List</a> • <a href="/wiki/Prime_Minister%27s_Office_(India)" title="Prime Minister&#39;s Office (India)">Office</a>) • <a href="/wiki/Deputy_Prime_Minister_of_India" title="Deputy Prime Minister of India">Deputy Prime Minister</a> • <a href="/wiki/Union_Council_of_Ministers" title="Union Council of Ministers">Council of Ministers</a></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Current</th><td class="navbox-list-with-group navbox-list navbox-odd hlist" style="width:100%;padding:0"><div style="padding:0 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%">Key</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Ministry_of_Defence_(India)" title="Ministry of Defence (India)">Defence</a></li>
<li><a href="/wiki/Ministry_of_External_Affairs_(India)" title="Ministry of External Affairs (India)">External Affairs</a></li>
<li><a href="/wiki/Ministry_of_Finance_(India)" title="Ministry of Finance (India)">Finance</a></li>
<li><a href="/wiki/Ministry_of_Home_Affairs_(India)" title="Ministry of Home Affairs (India)">Home Affairs</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Economy &amp; Industry</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Ministry_of_Chemicals_and_Fertilizers" title="Ministry of Chemicals and Fertilizers">Chemicals and Fertilizers</a></li>
<li><a href="/wiki/Ministry_of_Civil_Aviation_(India)" title="Ministry of Civil Aviation (India)">Civil Aviation</a></li>
<li><a href="/wiki/Ministry_of_Commerce_and_Industry_(India)" title="Ministry of Commerce and Industry (India)">Commerce and Industry</a></li>
<li><a href="/wiki/Ministry_of_Communications_(India)" title="Ministry of Communications (India)">Communications</a></li>
<li><a href="/wiki/Ministry_of_Co-operation" title="Ministry of Co-operation">Co-operation</a></li>
<li><a href="/wiki/Ministry_of_Corporate_Affairs" title="Ministry of Corporate Affairs">Corporate Affairs</a></li>
<li><a href="/wiki/Ministry_of_Electronics_and_Information_Technology" title="Ministry of Electronics and Information Technology">Electronics and Information Technology</a></li>
<li><a href="/wiki/Ministry_of_Heavy_Industries_and_Public_Enterprises" class="mw-redirect" title="Ministry of Heavy Industries and Public Enterprises">Heavy Industries and Public Enterprises</a></li>
<li><a href="/wiki/Ministry_of_Information_and_Broadcasting_(India)" title="Ministry of Information and Broadcasting (India)">Information and Broadcasting</a></li>
<li><a href="/wiki/Ministry_of_Labour_and_Employment_(India)" title="Ministry of Labour and Employment (India)">Labour and Employment</a></li>
<li><a href="/wiki/Ministry_of_Micro,_Small_and_Medium_Enterprises" title="Ministry of Micro, Small and Medium Enterprises">Micro, Small and Medium Enterprises</a></li>
<li><a href="/wiki/Ministry_of_Mines_(India)" title="Ministry of Mines (India)">Mines</a></li>
<li><a href="/wiki/Ministry_of_Steel" title="Ministry of Steel">Steel</a></li>
<li><a href="/wiki/Ministry_of_Textiles" title="Ministry of Textiles">Textiles</a></li>
<li><a href="/wiki/Ministry_of_Tourism_(India)" title="Ministry of Tourism (India)">Tourism</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Infrastructure</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Ministry_of_Development_of_North_Eastern_Region" title="Ministry of Development of North Eastern Region">Development of North Eastern Region</a></li>
<li><a href="/wiki/Ministry_of_Jal_Shakti" title="Ministry of Jal Shakti">Jal Shakti</a></li>
<li><a href="/wiki/Ministry_of_Ports,_Shipping_and_Waterways" title="Ministry of Ports, Shipping and Waterways">Ports, Shipping and Waterways</a></li>
<li><a href="/wiki/Ministry_of_Railways_(India)" title="Ministry of Railways (India)">Railways</a></li>
<li><a href="/wiki/Ministry_of_Road_Transport_and_Highways" title="Ministry of Road Transport and Highways">Road Transport and Highways</a></li>
<li><a href="/wiki/Ministry_of_Rural_Development_(India)" title="Ministry of Rural Development (India)">Rural Development</a></li>
<li><a href="/wiki/Ministry_of_Housing_and_Urban_Affairs" title="Ministry of Housing and Urban Affairs">Housing and Urban Affairs</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Food</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Ministry_of_Agriculture_and_Farmers%27_Welfare" title="Ministry of Agriculture and Farmers&#39; Welfare">Agriculture and Farmers' Welfare</a></li>
<li><a href="/wiki/Ministry_of_Consumer_Affairs,_Food_and_Public_Distribution" title="Ministry of Consumer Affairs, Food and Public Distribution">Consumer Affairs, Food and Public Distribution</a></li>
<li><a href="/wiki/Ministry_of_Fisheries,_Animal_Husbandry_and_Dairying" title="Ministry of Fisheries, Animal Husbandry and Dairying">Fisheries, Animal Husbandry and Dairying</a></li>
<li><a href="/wiki/Ministry_of_Food_Processing_Industries" title="Ministry of Food Processing Industries">Food Processing Industries</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Energy &amp; Climate Change</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Ministry_of_Coal" title="Ministry of Coal">Coal</a></li>
<li><a href="/wiki/Ministry_of_Environment,_Forest_and_Climate_Change" title="Ministry of Environment, Forest and Climate Change">Environment, Forest and Climate Change</a></li>
<li><a href="/wiki/Ministry_of_New_and_Renewable_Energy" title="Ministry of New and Renewable Energy">New and Renewable Energy</a></li>
<li><a href="/wiki/Ministry_of_Petroleum_and_Natural_Gas" title="Ministry of Petroleum and Natural Gas">Petroleum and Natural Gas</a></li>
<li><a href="/wiki/Ministry_of_Power_(India)" title="Ministry of Power (India)">Power</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Science, Education &amp; Health</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Ministry_of_AYUSH" title="Ministry of AYUSH">AYUSH</a></li>
<li><a href="/wiki/Ministry_of_Earth_Sciences" title="Ministry of Earth Sciences">Earth Sciences</a></li>
<li><a href="/wiki/Ministry_of_Education_(India)" title="Ministry of Education (India)">Education</a></li>
<li><a href="/wiki/Ministry_of_Health_and_Family_Welfare" title="Ministry of Health and Family Welfare">Health and Family Welfare</a></li>
<li><a href="/wiki/Ministry_of_Science_and_Technology_(India)" title="Ministry of Science and Technology (India)">Science and Technology</a></li>
<li><a href="/wiki/Ministry_of_Skill_Development_and_Entrepreneurship" title="Ministry of Skill Development and Entrepreneurship">Skill Development and Entrepreneurship</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">General Administration</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Ministry_of_Culture_(India)" title="Ministry of Culture (India)">Culture</a></li>
<li><a href="/wiki/Ministry_of_Law_and_Justice_(India)" title="Ministry of Law and Justice (India)">Law and Justice</a></li>
<li><a href="/wiki/Ministry_of_Minority_Affairs" title="Ministry of Minority Affairs">Minority Affairs</a></li>
<li><a href="/wiki/Ministry_of_Panchayati_Raj" title="Ministry of Panchayati Raj">Panchayati Raj</a></li>
<li><a href="/wiki/Ministry_of_Parliamentary_Affairs_(India)" title="Ministry of Parliamentary Affairs (India)">Parliamentary Affairs</a></li>
<li><a href="/wiki/Ministry_of_Personnel,_Public_Grievances_and_Pensions" title="Ministry of Personnel, Public Grievances and Pensions">Personnel, Public Grievances and Pensions</a></li>
<li><a href="/wiki/Ministry_of_Planning_(India)" title="Ministry of Planning (India)">Planning</a></li>
<li><a href="/wiki/Ministry_of_Social_Justice_and_Empowerment" title="Ministry of Social Justice and Empowerment">Social Justice and Empowerment</a></li>
<li><a href="/wiki/Ministry_of_Statistics_and_Programme_Implementation" title="Ministry of Statistics and Programme Implementation">Statistics and Programme Implementation</a></li>
<li><a href="/wiki/Ministry_of_Tribal_Affairs" title="Ministry of Tribal Affairs"> Tribal Affairs</a></li>
<li><a href="/wiki/Ministry_of_Women_and_Child_Development" title="Ministry of Women and Child Development">Women and Child Development</a></li>
<li><a href="/wiki/Ministry_of_Youth_Affairs_and_Sports" title="Ministry of Youth Affairs and Sports">Youth Affairs and Sports</a></li></ul>
</div></td></tr></tbody></table><div></div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Defunct</th><td class="navbox-list-with-group navbox-list navbox-even hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Ministry_of_Communications_and_Information_Technology_(India)" title="Ministry of Communications and Information Technology (India)">Communications and Information Technology</a></li>
<li><a href="/wiki/Ministry_of_Overseas_Indian_Affairs" title="Ministry of Overseas Indian Affairs">Overseas Indian Affairs</a></li>
<li><a href="/wiki/Ministry_of_Surface_Transport" title="Ministry of Surface Transport">Surface Transport</a></li>
<li><a href="/wiki/Ministry_of_Drinking_Water_and_Sanitation" title="Ministry of Drinking Water and Sanitation">Drinking Water and Sanitation</a></li>
<li><a href="/wiki/Ministry_of_Water_Resources,_River_Development_and_Ganga_Rejuvenation" title="Ministry of Water Resources, River Development and Ganga Rejuvenation">Water Resources, River Development and Ganga Rejuvenation</a></li>
<li><a href="/wiki/Ministry_of_Housing_and_Urban_Poverty_Alleviation" title="Ministry of Housing and Urban Poverty Alleviation">Housing and Urban Poverty Alleviation</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Ministers</th><td class="navbox-list-with-group navbox-list navbox-odd hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Minister_of_Agriculture_and_Farmers%27_Welfare" class="mw-redirect" title="Minister of Agriculture and Farmers&#39; Welfare">Agriculture and Farmers' Welfare</a></li>
<li><a href="/wiki/Minister_of_Defence_(India)" title="Minister of Defence (India)">Defence</a></li>
<li><a href="/wiki/Minister_of_Education_(India)" title="Minister of Education (India)">Education</a></li>
<li><a href="/wiki/Minister_of_External_Affairs_(India)" title="Minister of External Affairs (India)">External Affairs</a></li>
<li><a href="/wiki/Minister_of_Finance_(India)" title="Minister of Finance (India)">Finance</a></li>
<li><a href="/wiki/Minister_of_Home_Affairs_(India)" title="Minister of Home Affairs (India)">Home Affairs</a></li>
<li><a href="/wiki/Minister_of_Law_and_Justice" title="Minister of Law and Justice">Law and Justice</a></li>
<li><a href="/wiki/Minister_of_Personnel,_Public_Grievances_and_Pensions" title="Minister of Personnel, Public Grievances and Pensions">Personnel, Public Grievances and Pensions</a></li>
<li><a href="/wiki/Minister_of_Petroleum_and_Natural_Gas" title="Minister of Petroleum and Natural Gas">Petroleum and Natural Gas</a></li>
<li><a href="/wiki/Minister_of_Railways_(India)" title="Minister of Railways (India)">Railways</a></li>
<li><a href="/wiki/Minister_of_Steel" title="Minister of Steel">Steel</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Secretary_to_the_Government_of_India" title="Secretary to the Government of India">Secretaries</a></th><td class="navbox-list-with-group navbox-list navbox-even hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Cabinet_Secretary_(India)" title="Cabinet Secretary (India)">Cabinet</a></li>
<li><a href="/wiki/Defence_Secretary_(India)" title="Defence Secretary (India)">Defence</a></li>
<li><a href="/wiki/Finance_Secretary_(India)" title="Finance Secretary (India)">Finance</a></li>
<li><a href="/wiki/Foreign_Secretary_(India)" title="Foreign Secretary (India)">Foreign</a></li>
<li><a href="/wiki/Home_Secretary_(India)" title="Home Secretary (India)">Home</a></li>
<li><a href="/wiki/Personnel_Secretary_(India)" title="Personnel Secretary (India)">Personnel</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">Departments</th><td class="navbox-list-with-group navbox-list navbox-odd hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/Department_of_Atomic_Energy" title="Department of Atomic Energy">Atomic Energy</a></li>
<li><a href="/wiki/Department_of_Biotechnology" title="Department of Biotechnology">Biotechnology</a></li>
<li><a href="/wiki/Department_of_Higher_Education_(India)" title="Department of Higher Education (India)">Higher Education</a></li>
<li><a href="/wiki/Department_of_Official_Language" title="Department of Official Language">Official Language</a></li>
<li><a href="/wiki/Department_of_Science_and_Technology_(India)" title="Department of Science and Technology (India)">Science and Technology (DST)</a></li>
<li><a href="/wiki/Department_of_Military_Affairs_(India)" class="mw-redirect" title="Department of Military Affairs (India)">Military Affairs</a></li>
<li><a href="/wiki/Department_for_Promotion_of_Industry_and_Internal_Trade" title="Department for Promotion of Industry and Internal Trade">Promotion of Industry and Internal Trade</a></li>
<li><a href="/wiki/Department_of_Space" title="Department of Space">Space</a></li>
<li><a href="/wiki/Ministry_of_Communications_and_Information_Technology_(India)" title="Ministry of Communications and Information Technology (India)">Telecommunication</a></li></ul>
</div></td></tr></tbody></table></div>
<!-- 
NewPP limit report
Parsed by mw1420
Cached time: 20220327190844
Cache expiry: 1814400
Reduced expiry: false
Complications: [vary‐revision‐sha1]
CPU time usage: 0.819 seconds
Real time usage: 1.011 seconds
Preprocessor visited node count: 6145/1000000
Post‐expand include size: 161661/2097152 bytes
Template argument size: 8157/2097152 bytes
Highest expansion depth: 13/100
Expensive parser function count: 3/500
Unstrip recursion depth: 1/20
Unstrip post‐expand size: 65474/5000000 bytes
Lua time usage: 0.374/10.000 seconds
Lua memory usage: 8649562/52428800 bytes
Number of Wikibase entities loaded: 0/400
-->
<!--
Transclusion expansion time report (%,ms,calls,template)
100.00%  777.018      1 -total
 23.80%  184.905      2 Template:Reflist
 16.89%  131.275      9 Template:Cite_news
  9.98%   77.521      1 Template:Politics_of_India
  9.70%   75.350      1 Template:Sidebar_with_collapsible_lists
  7.42%   57.627      1 Template:Short_description
  7.23%   56.174      6 Template:Party_index_link
  7.06%   54.850     45 Template:Sortname
  6.24%   48.505      6 Template:Color_box
  5.62%   43.685      9 Template:Main_other
-->

<!-- Saved in parser cache with key enwiki:pcache:idhash:844745-0!canonical and timestamp 20220327190843 and revision id 1078988291. Serialized with JSON.
 -->
</div><noscript><img src="//en.wikipedia.org/wiki/Special:CentralAutoLogin/start?type=1x1" alt="" title="" width="1" height="1" style="border: none; position: absolute;" /></noscript>
<div class="printfooter">Retrieved from "<a dir="ltr" href="https://en.wikipedia.org/w/index.php?title=List_of_prime_ministers_of_India&amp;oldid=1078988291">https://en.wikipedia.org/w/index.php?title=List_of_prime_ministers_of_India&amp;oldid=1078988291</a>"</div></div>
		<div id="catlinks" class="catlinks" data-mw="interface"><div id="mw-normal-catlinks" class="mw-normal-catlinks"><a href="/wiki/Help:Category" title="Help:Category">Categories</a>: <ul><li><a href="/wiki/Category:Prime_Ministers_of_India" title="Category:Prime Ministers of India">Prime Ministers of India</a></li><li><a href="/wiki/Category:Lists_of_prime_ministers_by_country" title="Category:Lists of prime ministers by country">Lists of prime ministers by country</a></li><li><a href="/wiki/Category:Lists_of_political_office-holders_in_India" title="Category:Lists of political office-holders in India">Lists of political office-holders in India</a></li><li><a href="/wiki/Category:Lists_relating_to_prime_ministers_of_India" title="Category:Lists relating to prime ministers of India">Lists relating to prime ministers of India</a></li></ul></div><div id="mw-hidden-catlinks" class="mw-hidden-catlinks mw-hidden-cats-hidden">Hidden categories: <ul><li><a href="/wiki/Category:Pages_using_the_EasyTimeline_extension" title="Category:Pages using the EasyTimeline extension">Pages using the EasyTimeline extension</a></li><li><a href="/wiki/Category:Articles_with_short_description" title="Category:Articles with short description">Articles with short description</a></li><li><a href="/wiki/Category:Short_description_is_different_from_Wikidata" title="Category:Short description is different from Wikidata">Short description is different from Wikidata</a></li><li><a href="/wiki/Category:Featured_lists" title="Category:Featured lists">Featured lists</a></li><li><a href="/wiki/Category:Use_dmy_dates_from_September_2020" title="Category:Use dmy dates from September 2020">Use dmy dates from September 2020</a></li><li><a href="/wiki/Category:Use_Indian_English_from_May_2013" title="Category:Use Indian English from May 2013">Use Indian English from May 2013</a></li><li><a href="/wiki/Category:All_Wikipedia_articles_written_in_Indian_English" title="Category:All Wikipedia articles written in Indian English">All Wikipedia articles written in Indian English</a></li><li><a href="/wiki/Category:Articles_with_hCards" title="Category:Articles with hCards">Articles with hCards</a></li></ul></div></div>
	</div>
</div>
<div id='mw-data-after-content'>
	<div class="read-more-container"></div>
</div>

<div id="mw-navigation">
	<h2>Navigation menu</h2>
	<div id="mw-head">
		
<nav id="p-personal" class="mw-portlet mw-portlet-personal vector-user-menu-legacy vector-menu" aria-labelledby="p-personal-label" role="navigation" 
	 >
	<label id="p-personal-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">Personal tools</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="pt-anonuserpage" class="mw-list-item"><span>Not logged in</span></li><li id="pt-anontalk" class="mw-list-item"><a href="/wiki/Special:MyTalk" title="Discussion about edits from this IP address [n]" accesskey="n"><span>Talk</span></a></li><li id="pt-anoncontribs" class="mw-list-item"><a href="/wiki/Special:MyContributions" title="A list of edits made from this IP address [y]" accesskey="y"><span>Contributions</span></a></li><li id="pt-createaccount" class="mw-list-item"><a href="/w/index.php?title=Special:CreateAccount&amp;returnto=List+of+prime+ministers+of+India" title="You are encouraged to create an account and log in; however, it is not mandatory"><span>Create account</span></a></li><li id="pt-login" class="mw-list-item"><a href="/w/index.php?title=Special:UserLogin&amp;returnto=List+of+prime+ministers+of+India" title="You&#039;re encouraged to log in; however, it&#039;s not mandatory. [o]" accesskey="o"><span>Log in</span></a></li></ul>
		
	</div>
</nav>

		<div id="left-navigation">
			
<nav id="p-namespaces" class="mw-portlet mw-portlet-namespaces vector-menu vector-menu-tabs" aria-labelledby="p-namespaces-label" role="navigation" 
	 >
	<label id="p-namespaces-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">Namespaces</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="ca-nstab-main" class="selected mw-list-item"><a href="/wiki/List_of_prime_ministers_of_India" title="View the content page [c]" accesskey="c"><span>Article</span></a></li><li id="ca-talk" class="mw-list-item"><a href="/wiki/Talk:List_of_prime_ministers_of_India" rel="discussion" title="Discuss improvements to the content page [t]" accesskey="t"><span>Talk</span></a></li></ul>
		
	</div>
</nav>

			
<nav id="p-variants" class="mw-portlet mw-portlet-variants emptyPortlet vector-menu-dropdown-noicon vector-menu vector-menu-dropdown" aria-labelledby="p-variants-label" role="navigation" 
	 >
	<input type="checkbox"
		id="p-variants-checkbox"
		role="button"
		aria-haspopup="true"
		data-event-name="ui.dropdown-p-variants"
		class="vector-menu-checkbox" aria-labelledby="p-variants-label" />
	<label id="p-variants-label" aria-label="Change language variant" class="vector-menu-heading">
		<span class="vector-menu-heading-label">English</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"></ul>
		
	</div>
</nav>

		</div>
		<div id="right-navigation">
			
<nav id="p-views" class="mw-portlet mw-portlet-views vector-menu vector-menu-tabs" aria-labelledby="p-views-label" role="navigation" 
	 >
	<label id="p-views-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">Views</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="ca-view" class="selected mw-list-item"><a href="/wiki/List_of_prime_ministers_of_India"><span>Read</span></a></li><li id="ca-edit" class="mw-list-item"><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=edit" title="Edit this page [e]" accesskey="e"><span>Edit</span></a></li><li id="ca-history" class="mw-list-item"><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=history" title="Past revisions of this page [h]" accesskey="h"><span>View history</span></a></li></ul>
		
	</div>
</nav>

			
<nav id="p-cactions" class="mw-portlet mw-portlet-cactions emptyPortlet vector-menu-dropdown-noicon vector-menu vector-menu-dropdown" aria-labelledby="p-cactions-label" role="navigation"  title="More options"
	 >
	<input type="checkbox"
		id="p-cactions-checkbox"
		role="button"
		aria-haspopup="true"
		data-event-name="ui.dropdown-p-cactions"
		class="vector-menu-checkbox" aria-labelledby="p-cactions-label" />
	<label id="p-cactions-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">More</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"></ul>
		
	</div>
</nav>

			
<div id="p-search" role="search" class="vector-search-box-vue  vector-search-box-show-thumbnail vector-search-box-auto-expand-width vector-search-box">
	<div>
			<h3 >
				<label for="searchInput">Search</label>
			</h3>
		<form action="/w/index.php" id="searchform"
			class="vector-search-box-form">
			<div id="simpleSearch"
				class="vector-search-box-inner"
				 data-search-loc="header-navigation">
				<input class="vector-search-box-input"
					 type="search" name="search" placeholder="Search Wikipedia" aria-label="Search Wikipedia" autocapitalize="sentences" title="Search Wikipedia [f]" accesskey="f" id="searchInput"
				/>
				<input type="hidden" name="title" value="Special:Search"/>
				<input id="mw-searchButton"
					 class="searchButton mw-fallbackSearchButton" type="submit" name="fulltext" title="Search Wikipedia for this text" value="Search" />
				<input id="searchButton"
					 class="searchButton" type="submit" name="go" title="Go to a page with this exact name if it exists" value="Go" />
			</div>
		</form>
	</div>
</div>

		</div>
	</div>
	

<div id="mw-panel">
	<div id="p-logo" role="banner">
		<a class="mw-wiki-logo" href="/wiki/Main_Page"
			title="Visit the main page"></a>
	</div>
	
<nav id="p-navigation" class="mw-portlet mw-portlet-navigation vector-menu vector-menu-portal portal" aria-labelledby="p-navigation-label" role="navigation" 
	 >
	<label id="p-navigation-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">Navigation</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="n-mainpage-description" class="mw-list-item"><a href="/wiki/Main_Page" icon="home" title="Visit the main page [z]" accesskey="z"><span>Main page</span></a></li><li id="n-contents" class="mw-list-item"><a href="/wiki/Wikipedia:Contents" title="Guides to browsing Wikipedia"><span>Contents</span></a></li><li id="n-currentevents" class="mw-list-item"><a href="/wiki/Portal:Current_events" title="Articles related to current events"><span>Current events</span></a></li><li id="n-randompage" class="mw-list-item"><a href="/wiki/Special:Random" icon="die" title="Visit a randomly selected article [x]" accesskey="x"><span>Random article</span></a></li><li id="n-aboutsite" class="mw-list-item"><a href="/wiki/Wikipedia:About" title="Learn about Wikipedia and how it works"><span>About Wikipedia</span></a></li><li id="n-contactpage" class="mw-list-item"><a href="//en.wikipedia.org/wiki/Wikipedia:Contact_us" title="How to contact Wikipedia"><span>Contact us</span></a></li><li id="n-sitesupport" class="mw-list-item"><a href="https://donate.wikimedia.org/wiki/Special:FundraiserRedirector?utm_source=donate&amp;utm_medium=sidebar&amp;utm_campaign=C13_en.wikipedia.org&amp;uselang=en" title="Support us by donating to the Wikimedia Foundation"><span>Donate</span></a></li></ul>
		
	</div>
</nav>

	
<nav id="p-interaction" class="mw-portlet mw-portlet-interaction vector-menu vector-menu-portal portal" aria-labelledby="p-interaction-label" role="navigation" 
	 >
	<label id="p-interaction-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">Contribute</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="n-help" class="mw-list-item"><a href="/wiki/Help:Contents" icon="help" title="Guidance on how to use and edit Wikipedia"><span>Help</span></a></li><li id="n-introduction" class="mw-list-item"><a href="/wiki/Help:Introduction" title="Learn how to edit Wikipedia"><span>Learn to edit</span></a></li><li id="n-portal" class="mw-list-item"><a href="/wiki/Wikipedia:Community_portal" title="The hub for editors"><span>Community portal</span></a></li><li id="n-recentchanges" class="mw-list-item"><a href="/wiki/Special:RecentChanges" icon="recentChanges" title="A list of recent changes to Wikipedia [r]" accesskey="r"><span>Recent changes</span></a></li><li id="n-upload" class="mw-list-item"><a href="/wiki/Wikipedia:File_Upload_Wizard" title="Add images or other media for use on Wikipedia"><span>Upload file</span></a></li></ul>
		
	</div>
</nav>

<nav id="p-tb" class="mw-portlet mw-portlet-tb vector-menu vector-menu-portal portal" aria-labelledby="p-tb-label" role="navigation" 
	 >
	<label id="p-tb-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">Tools</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="t-whatlinkshere" class="mw-list-item"><a href="/wiki/Special:WhatLinksHere/List_of_prime_ministers_of_India" title="List of all English Wikipedia pages containing links to this page [j]" accesskey="j"><span>What links here</span></a></li><li id="t-recentchangeslinked" class="mw-list-item"><a href="/wiki/Special:RecentChangesLinked/List_of_prime_ministers_of_India" rel="nofollow" title="Recent changes in pages linked from this page [k]" accesskey="k"><span>Related changes</span></a></li><li id="t-upload" class="mw-list-item"><a href="/wiki/Wikipedia:File_Upload_Wizard" title="Upload files [u]" accesskey="u"><span>Upload file</span></a></li><li id="t-specialpages" class="mw-list-item"><a href="/wiki/Special:SpecialPages" title="A list of all special pages [q]" accesskey="q"><span>Special pages</span></a></li><li id="t-permalink" class="mw-list-item"><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;oldid=1078988291" title="Permanent link to this revision of this page"><span>Permanent link</span></a></li><li id="t-info" class="mw-list-item"><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;action=info" title="More information about this page"><span>Page information</span></a></li><li id="t-cite" class="mw-list-item"><a href="/w/index.php?title=Special:CiteThisPage&amp;page=List_of_prime_ministers_of_India&amp;id=1078988291&amp;wpFormIdentifier=titleform" title="Information on how to cite this page"><span>Cite this page</span></a></li><li id="t-wikibase" class="mw-list-item"><a href="https://www.wikidata.org/wiki/Special:EntityPage/Q492746" title="Structured data on this page hosted by Wikidata [g]" accesskey="g"><span>Wikidata item</span></a></li></ul>
		
	</div>
</nav>

<nav id="p-coll-print_export" class="mw-portlet mw-portlet-coll-print_export vector-menu vector-menu-portal portal" aria-labelledby="p-coll-print_export-label" role="navigation" 
	 >
	<label id="p-coll-print_export-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">Print/export</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li id="coll-download-as-rl" class="mw-list-item"><a href="/w/index.php?title=Special:DownloadAsPdf&amp;page=List_of_prime_ministers_of_India&amp;action=show-download-screen" title="Download this page as a PDF file"><span>Download as PDF</span></a></li><li id="t-print" class="mw-list-item"><a href="/w/index.php?title=List_of_prime_ministers_of_India&amp;printable=yes" title="Printable version of this page [p]" accesskey="p"><span>Printable version</span></a></li></ul>
		
	</div>
</nav>

<nav id="p-wikibase-otherprojects" class="mw-portlet mw-portlet-wikibase-otherprojects vector-menu vector-menu-portal portal" aria-labelledby="p-wikibase-otherprojects-label" role="navigation" 
	 >
	<label id="p-wikibase-otherprojects-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">In other projects</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li class="wb-otherproject-link wb-otherproject-commons mw-list-item"><a href="https://commons.wikimedia.org/wiki/List_of_Prime_Ministers_of_India" hreflang="en"><span>Wikimedia Commons</span></a></li></ul>
		
	</div>
</nav>

	
<nav id="p-lang" class="mw-portlet mw-portlet-lang vector-menu vector-menu-portal portal" aria-labelledby="p-lang-label" role="navigation" 
	 >
	<label id="p-lang-label" aria-label="" class="vector-menu-heading">
		<span class="vector-menu-heading-label">Languages</span>
	</label>
	<div class="vector-menu-content">
		
		<ul class="vector-menu-content-list"><li class="interlanguage-link interwiki-ar mw-list-item"><a href="https://ar.wikipedia.org/wiki/%D9%82%D8%A7%D8%A6%D9%85%D8%A9_%D8%B1%D8%A4%D8%B3%D8%A7%D8%A1_%D9%88%D8%B2%D8%B1%D8%A7%D8%A1_%D8%A7%D9%84%D9%87%D9%86%D8%AF" title="قائمة رؤساء وزراء الهند – Arabic" lang="ar" hreflang="ar" class="interlanguage-link-target"><span>العربية</span></a></li><li class="interlanguage-link interwiki-as mw-list-item"><a href="https://as.wikipedia.org/wiki/%E0%A6%AD%E0%A6%BE%E0%A7%B0%E0%A6%A4%E0%A7%B0_%E0%A6%AA%E0%A7%8D%E0%A7%B0%E0%A6%A7%E0%A6%BE%E0%A6%A8%E0%A6%AE%E0%A6%A8%E0%A7%8D%E0%A6%A4%E0%A7%8D%E0%A7%B0%E0%A7%80%E0%A6%B8%E0%A6%95%E0%A6%B2%E0%A7%B0_%E0%A6%A4%E0%A6%BE%E0%A6%B2%E0%A6%BF%E0%A6%95%E0%A6%BE" title="ভাৰতৰ প্ৰধানমন্ত্ৰীসকলৰ তালিকা – Assamese" lang="as" hreflang="as" class="interlanguage-link-target"><span>অসমীয়া</span></a></li><li class="interlanguage-link interwiki-bn mw-list-item"><a href="https://bn.wikipedia.org/wiki/%E0%A6%AD%E0%A6%BE%E0%A6%B0%E0%A6%A4%E0%A7%87%E0%A6%B0_%E0%A6%AA%E0%A7%8D%E0%A6%B0%E0%A6%A7%E0%A6%BE%E0%A6%A8%E0%A6%AE%E0%A6%A8%E0%A7%8D%E0%A6%A4%E0%A7%8D%E0%A6%B0%E0%A7%80%E0%A6%A6%E0%A7%87%E0%A6%B0_%E0%A6%A4%E0%A6%BE%E0%A6%B2%E0%A6%BF%E0%A6%95%E0%A6%BE" title="ভারতের প্রধানমন্ত্রীদের তালিকা – Bangla" lang="bn" hreflang="bn" class="interlanguage-link-target"><span>বাংলা</span></a></li><li class="interlanguage-link interwiki-bh mw-list-item"><a href="https://bh.wikipedia.org/wiki/%E0%A4%AD%E0%A4%BE%E0%A4%B0%E0%A4%A4_%E0%A4%95%E0%A5%87_%E0%A4%AA%E0%A4%B0%E0%A4%A7%E0%A4%BE%E0%A4%A8%E0%A4%AE%E0%A4%82%E0%A4%A4%E0%A5%8D%E0%A4%B0%E0%A5%80_%E0%A4%B2%E0%A5%8B%E0%A4%97_%E0%A4%95%E0%A5%87_%E0%A4%B2%E0%A4%BF%E0%A4%B8%E0%A5%8D%E0%A4%9F" title="भारत के परधानमंत्री लोग के लिस्ट – Bhojpuri" lang="bh" hreflang="bh" class="interlanguage-link-target"><span>भोजपुरी</span></a></li><li class="interlanguage-link interwiki-de mw-list-item"><a href="https://de.wikipedia.org/wiki/Liste_der_Premierminister_Indiens" title="Liste der Premierminister Indiens – German" lang="de" hreflang="de" class="interlanguage-link-target"><span>Deutsch</span></a></li><li class="interlanguage-link interwiki-el mw-list-item"><a href="https://el.wikipedia.org/wiki/%CE%9A%CE%B1%CF%84%CE%AC%CE%BB%CE%BF%CE%B3%CE%BF%CF%82_%CF%80%CF%81%CF%89%CE%B8%CF%85%CF%80%CE%BF%CF%85%CF%81%CE%B3%CF%8E%CE%BD_%CF%84%CE%B7%CF%82_%CE%99%CE%BD%CE%B4%CE%AF%CE%B1%CF%82" title="Κατάλογος πρωθυπουργών της Ινδίας – Greek" lang="el" hreflang="el" class="interlanguage-link-target"><span>Ελληνικά</span></a></li><li class="interlanguage-link interwiki-eo mw-list-item"><a href="https://eo.wikipedia.org/wiki/Listo_de_%C4%89efministroj_de_Barato" title="Listo de ĉefministroj de Barato – Esperanto" lang="eo" hreflang="eo" class="interlanguage-link-target"><span>Esperanto</span></a></li><li class="interlanguage-link interwiki-fa mw-list-item"><a href="https://fa.wikipedia.org/wiki/%D9%81%D9%87%D8%B1%D8%B3%D8%AA_%D9%86%D8%AE%D8%B3%D8%AA%E2%80%8C%D9%88%D8%B2%DB%8C%D8%B1%D8%A7%D9%86_%D9%87%D9%86%D8%AF" title="فهرست نخست‌وزیران هند – Persian" lang="fa" hreflang="fa" class="interlanguage-link-target"><span>فارسی</span></a></li><li class="interlanguage-link interwiki-fr mw-list-item"><a href="https://fr.wikipedia.org/wiki/Liste_des_Premiers_ministres_de_l%27Inde" title="Liste des Premiers ministres de l&#039;Inde – French" lang="fr" hreflang="fr" class="interlanguage-link-target"><span>Français</span></a></li><li class="interlanguage-link interwiki-gu mw-list-item"><a href="https://gu.wikipedia.org/wiki/%E0%AA%AD%E0%AA%BE%E0%AA%B0%E0%AA%A4%E0%AA%A8%E0%AA%BE_%E0%AA%B5%E0%AA%A1%E0%AA%BE%E0%AA%AA%E0%AB%8D%E0%AA%B0%E0%AA%A7%E0%AA%BE%E0%AA%A8" title="ભારતના વડાપ્રધાન – Gujarati" lang="gu" hreflang="gu" class="interlanguage-link-target"><span>ગુજરાતી</span></a></li><li class="interlanguage-link interwiki-hi mw-list-item"><a href="https://hi.wikipedia.org/wiki/%E0%A4%AD%E0%A4%BE%E0%A4%B0%E0%A4%A4_%E0%A4%95%E0%A5%87_%E0%A4%AA%E0%A5%8D%E0%A4%B0%E0%A4%A7%E0%A4%BE%E0%A4%A8_%E0%A4%AE%E0%A4%82%E0%A4%A4%E0%A5%8D%E0%A4%B0%E0%A4%BF%E0%A4%AF%E0%A5%8B%E0%A4%82_%E0%A4%95%E0%A5%80_%E0%A4%B8%E0%A5%82%E0%A4%9A%E0%A5%80" title="भारत के प्रधान मंत्रियों की सूची – Hindi" lang="hi" hreflang="hi" class="interlanguage-link-target"><span>हिन्दी</span></a></li><li class="interlanguage-link interwiki-io mw-list-item"><a href="https://io.wikipedia.org/wiki/Listo_pri_chefministri_di_India" title="Listo pri chefministri di India – Ido" lang="io" hreflang="io" class="interlanguage-link-target"><span>Ido</span></a></li><li class="interlanguage-link interwiki-id mw-list-item"><a href="https://id.wikipedia.org/wiki/Daftar_Perdana_Menteri_India" title="Daftar Perdana Menteri India – Indonesian" lang="id" hreflang="id" class="interlanguage-link-target"><span>Bahasa Indonesia</span></a></li><li class="interlanguage-link interwiki-it mw-list-item"><a href="https://it.wikipedia.org/wiki/Primi_ministri_dell%27India" title="Primi ministri dell&#039;India – Italian" lang="it" hreflang="it" class="interlanguage-link-target"><span>Italiano</span></a></li><li class="interlanguage-link interwiki-kn mw-list-item"><a href="https://kn.wikipedia.org/wiki/%E0%B2%AD%E0%B2%BE%E0%B2%B0%E0%B2%A4%E0%B2%A6_%E0%B2%AA%E0%B3%8D%E0%B2%B0%E0%B2%A7%E0%B2%BE%E0%B2%A8_%E0%B2%AE%E0%B2%82%E0%B2%A4%E0%B3%8D%E0%B2%B0%E0%B2%BF%E0%B2%97%E0%B2%B3_%E0%B2%AA%E0%B2%9F%E0%B3%8D%E0%B2%9F%E0%B2%BF" title="ಭಾರತದ ಪ್ರಧಾನ ಮಂತ್ರಿಗಳ ಪಟ್ಟಿ – Kannada" lang="kn" hreflang="kn" class="interlanguage-link-target"><span>ಕನ್ನಡ</span></a></li><li class="interlanguage-link interwiki-lb mw-list-item"><a href="https://lb.wikipedia.org/wiki/L%C3%ABscht_vun_de_Premierministere_vun_Indien" title="Lëscht vun de Premierministere vun Indien – Luxembourgish" lang="lb" hreflang="lb" class="interlanguage-link-target"><span>Lëtzebuergesch</span></a></li><li class="interlanguage-link interwiki-ml mw-list-item"><a href="https://ml.wikipedia.org/wiki/%E0%B4%87%E0%B4%A8%E0%B5%8D%E0%B4%A4%E0%B5%8D%E0%B4%AF%E0%B4%AF%E0%B5%81%E0%B4%9F%E0%B5%86_%E0%B4%AA%E0%B5%8D%E0%B4%B0%E0%B4%A7%E0%B4%BE%E0%B4%A8%E0%B4%AE%E0%B4%A8%E0%B5%8D%E0%B4%A4%E0%B5%8D%E0%B4%B0%E0%B4%BF%E0%B4%AE%E0%B4%BE%E0%B4%B0%E0%B5%81%E0%B4%9F%E0%B5%86_%E0%B4%AA%E0%B4%9F%E0%B5%8D%E0%B4%9F%E0%B4%BF%E0%B4%95" title="ഇന്ത്യയുടെ പ്രധാനമന്ത്രിമാരുടെ പട്ടിക – Malayalam" lang="ml" hreflang="ml" class="interlanguage-link-target"><span>മലയാളം</span></a></li><li class="interlanguage-link interwiki-mr mw-list-item"><a href="https://mr.wikipedia.org/wiki/%E0%A4%AD%E0%A4%BE%E0%A4%B0%E0%A4%A4%E0%A4%BE%E0%A4%9A%E0%A5%8D%E0%A4%AF%E0%A4%BE_%E0%A4%AA%E0%A4%82%E0%A4%A4%E0%A4%AA%E0%A5%8D%E0%A4%B0%E0%A4%A7%E0%A4%BE%E0%A4%A8%E0%A4%BE%E0%A4%82%E0%A4%9A%E0%A5%80_%E0%A4%AF%E0%A4%BE%E0%A4%A6%E0%A5%80" title="भारताच्या पंतप्रधानांची यादी – Marathi" lang="mr" hreflang="mr" class="interlanguage-link-target"><span>मराठी</span></a></li><li class="interlanguage-link interwiki-nl mw-list-item"><a href="https://nl.wikipedia.org/wiki/Lijst_van_premiers_van_India" title="Lijst van premiers van India – Dutch" lang="nl" hreflang="nl" class="interlanguage-link-target"><span>Nederlands</span></a></li><li class="interlanguage-link interwiki-ne mw-list-item"><a href="https://ne.wikipedia.org/wiki/%E0%A4%AD%E0%A4%BE%E0%A4%B0%E0%A4%A4%E0%A4%95%E0%A4%BE_%E0%A4%AA%E0%A5%8D%E0%A4%B0%E0%A4%A7%E0%A4%BE%E0%A4%A8%E0%A4%AE%E0%A4%A8%E0%A5%8D%E0%A4%A4%E0%A5%8D%E0%A4%B0%E0%A5%80%E0%A4%B9%E0%A4%B0%E0%A5%82%E0%A4%95%E0%A5%8B_%E0%A4%B8%E0%A5%82%E0%A4%9A%E0%A5%80" title="भारतका प्रधानमन्त्रीहरूको सूची – Nepali" lang="ne" hreflang="ne" class="interlanguage-link-target"><span>नेपाली</span></a></li><li class="interlanguage-link interwiki-ja mw-list-item"><a href="https://ja.wikipedia.org/wiki/%E3%82%A4%E3%83%B3%E3%83%89%E3%81%AE%E6%AD%B4%E4%BB%A3%E9%A6%96%E7%9B%B8" title="インドの歴代首相 – Japanese" lang="ja" hreflang="ja" class="interlanguage-link-target"><span>日本語</span></a></li><li class="interlanguage-link interwiki-no mw-list-item"><a href="https://no.wikipedia.org/wiki/Liste_over_Indias_statsministre" title="Liste over Indias statsministre – Norwegian Bokmål" lang="nb" hreflang="nb" class="interlanguage-link-target"><span>Norsk bokmål</span></a></li><li class="interlanguage-link interwiki-oc mw-list-item"><a href="https://oc.wikipedia.org/wiki/Lista_dels_primi%C3%A8rs_ministres_d%27%C3%8Dndia" title="Lista dels primièrs ministres d&#039;Índia – Occitan" lang="oc" hreflang="oc" class="interlanguage-link-target"><span>Occitan</span></a></li><li class="interlanguage-link interwiki-or mw-list-item"><a href="https://or.wikipedia.org/wiki/%E0%AC%AD%E0%AC%BE%E0%AC%B0%E0%AC%A4%E0%AC%B0_%E0%AC%AA%E0%AD%8D%E0%AC%B0%E0%AC%A7%E0%AC%BE%E0%AC%A8%E0%AC%AE%E0%AC%A8%E0%AD%8D%E0%AC%A4%E0%AD%8D%E0%AC%B0%E0%AD%80%E0%AC%AE%E0%AC%BE%E0%AC%A8%E0%AC%99%E0%AD%8D%E0%AC%95%E0%AC%B0_%E0%AC%A4%E0%AC%BE%E0%AC%B2%E0%AC%BF%E0%AC%95%E0%AC%BE" title="ଭାରତର ପ୍ରଧାନମନ୍ତ୍ରୀମାନଙ୍କର ତାଲିକା – Odia" lang="or" hreflang="or" class="interlanguage-link-target"><span>ଓଡ଼ିଆ</span></a></li><li class="interlanguage-link interwiki-pa mw-list-item"><a href="https://pa.wikipedia.org/wiki/%E0%A8%AD%E0%A8%BE%E0%A8%B0%E0%A8%A4_%E0%A8%A6%E0%A9%87_%E0%A8%AA%E0%A9%8D%E0%A8%B0%E0%A8%A7%E0%A8%BE%E0%A8%A8_%E0%A8%AE%E0%A9%B0%E0%A8%A4%E0%A8%B0%E0%A9%80%E0%A8%86%E0%A8%82_%E0%A8%A6%E0%A9%80_%E0%A8%B8%E0%A9%82%E0%A8%9A%E0%A9%80" title="ਭਾਰਤ ਦੇ ਪ੍ਰਧਾਨ ਮੰਤਰੀਆਂ ਦੀ ਸੂਚੀ – Punjabi" lang="pa" hreflang="pa" class="interlanguage-link-target"><span>ਪੰਜਾਬੀ</span></a></li><li class="interlanguage-link interwiki-pnb mw-list-item"><a href="https://pnb.wikipedia.org/wiki/%D8%A8%DA%BE%D8%A7%D8%B1%D8%AA_%D8%AF%DB%92_%D9%88%D8%B2%DB%8C%D8%B1%D8%A7%D8%B9%D8%B8%D9%85" title="بھارت دے وزیراعظم – Western Punjabi" lang="pnb" hreflang="pnb" class="interlanguage-link-target"><span>پنجابی</span></a></li><li class="interlanguage-link interwiki-pt mw-list-item"><a href="https://pt.wikipedia.org/wiki/Lista_de_primeiros-ministros_da_%C3%8Dndia" title="Lista de primeiros-ministros da Índia – Portuguese" lang="pt" hreflang="pt" class="interlanguage-link-target"><span>Português</span></a></li><li class="interlanguage-link interwiki-ru badge-Q17506997 badge-featuredlist mw-list-item" title="featured list badge"><a href="https://ru.wikipedia.org/wiki/%D0%A1%D0%BF%D0%B8%D1%81%D0%BE%D0%BA_%D0%BF%D1%80%D0%B5%D0%BC%D1%8C%D0%B5%D1%80-%D0%BC%D0%B8%D0%BD%D0%B8%D1%81%D1%82%D1%80%D0%BE%D0%B2_%D0%98%D0%BD%D0%B4%D0%B8%D0%B8" title="Список премьер-министров Индии – Russian" lang="ru" hreflang="ru" class="interlanguage-link-target"><span>Русский</span></a></li><li class="interlanguage-link interwiki-sa mw-list-item"><a href="https://sa.wikipedia.org/wiki/%E0%A4%AD%E0%A4%BE%E0%A4%B0%E0%A4%A4%E0%A4%B8%E0%A5%8D%E0%A4%AF_%E0%A4%AA%E0%A5%8D%E0%A4%B0%E0%A4%A7%E0%A4%BE%E0%A4%A8%E0%A4%AE%E0%A4%A8%E0%A5%8D%E0%A4%A4%E0%A5%8D%E0%A4%B0%E0%A4%BF%E0%A4%A3%E0%A4%83" title="भारतस्य प्रधानमन्त्रिणः – Sanskrit" lang="sa" hreflang="sa" class="interlanguage-link-target"><span>संस्कृतम्</span></a></li><li class="interlanguage-link interwiki-sat mw-list-item"><a href="https://sat.wikipedia.org/wiki/%E1%B1%A5%E1%B1%A4%E1%B1%A7%E1%B1%9A%E1%B1%9B_%E1%B1%A8%E1%B1%A4%E1%B1%B1_%E1%B1%A2%E1%B1%9F%E1%B1%A8%E1%B1%9F%E1%B1%9D_%E1%B1%A2%E1%B1%9A%E1%B1%B1%E1%B1%9B%E1%B1%A8%E1%B1%A4_%E1%B1%A0%E1%B1%9A%E1%B1%A3%E1%B1%9F%E1%B1%9C_%E1%B1%9E%E1%B1%A4%E1%B1%A5%E1%B1%B4%E1%B1%A4" title="ᱥᱤᱧᱚᱛ ᱨᱤᱱ ᱢᱟᱨᱟᱝ ᱢᱚᱱᱛᱨᱤ ᱠᱚᱣᱟᱜ ᱞᱤᱥᱴᱤ – Santali" lang="sat" hreflang="sat" class="interlanguage-link-target"><span>ᱥᱟᱱᱛᱟᱲᱤ</span></a></li><li class="interlanguage-link interwiki-simple mw-list-item"><a href="https://simple.wikipedia.org/wiki/List_of_prime_ministers_of_India" title="List of prime ministers of India – Simple English" lang="en-simple" hreflang="en-simple" class="interlanguage-link-target"><span>Simple English</span></a></li><li class="interlanguage-link interwiki-sr mw-list-item"><a href="https://sr.wikipedia.org/wiki/%D0%A1%D0%BF%D0%B8%D1%81%D0%B0%D0%BA_%D0%BF%D1%80%D0%B5%D0%BC%D0%B8%D1%98%D0%B5%D1%80%D0%B0_%D0%98%D0%BD%D0%B4%D0%B8%D1%98%D0%B5" title="Списак премијера Индије – Serbian" lang="sr" hreflang="sr" class="interlanguage-link-target"><span>Српски / srpski</span></a></li><li class="interlanguage-link interwiki-fi mw-list-item"><a href="https://fi.wikipedia.org/wiki/Luettelo_Intian_p%C3%A4%C3%A4ministereist%C3%A4" title="Luettelo Intian pääministereistä – Finnish" lang="fi" hreflang="fi" class="interlanguage-link-target"><span>Suomi</span></a></li><li class="interlanguage-link interwiki-ta mw-list-item"><a href="https://ta.wikipedia.org/wiki/%E0%AE%87%E0%AE%A8%E0%AF%8D%E0%AE%A4%E0%AE%BF%E0%AE%AF%E0%AE%AA%E0%AF%8D_%E0%AE%AA%E0%AE%BF%E0%AE%B0%E0%AE%A4%E0%AE%AE_%E0%AE%AE%E0%AE%A8%E0%AF%8D%E0%AE%A4%E0%AE%BF%E0%AE%B0%E0%AE%BF%E0%AE%95%E0%AE%B3%E0%AE%BF%E0%AE%A9%E0%AF%8D_%E0%AE%AA%E0%AE%9F%E0%AF%8D%E0%AE%9F%E0%AE%BF%E0%AE%AF%E0%AE%B2%E0%AF%8D" title="இந்தியப் பிரதம மந்திரிகளின் பட்டியல் – Tamil" lang="ta" hreflang="ta" class="interlanguage-link-target"><span>தமிழ்</span></a></li><li class="interlanguage-link interwiki-te mw-list-item"><a href="https://te.wikipedia.org/wiki/%E0%B0%AD%E0%B0%BE%E0%B0%B0%E0%B0%A4_%E0%B0%AA%E0%B1%8D%E0%B0%B0%E0%B0%A7%E0%B0%BE%E0%B0%A8%E0%B0%AE%E0%B0%82%E0%B0%A4%E0%B1%8D%E0%B0%B0%E0%B1%81%E0%B0%B2%E0%B1%81" title="భారత ప్రధానమంత్రులు – Telugu" lang="te" hreflang="te" class="interlanguage-link-target"><span>తెలుగు</span></a></li><li class="interlanguage-link interwiki-tg mw-list-item"><a href="https://tg.wikipedia.org/wiki/%D0%A4%D0%B5%D2%B3%D1%80%D0%B8%D1%81%D1%82%D0%B8_%D1%81%D0%B0%D1%80%D0%B2%D0%B0%D0%B7%D0%B8%D1%80%D0%BE%D0%BD%D0%B8_%D2%B2%D0%B8%D0%BD%D0%B4%D1%83%D1%81%D1%82%D0%BE%D0%BD" title="Феҳристи сарвазирони Ҳиндустон – Tajik" lang="tg" hreflang="tg" class="interlanguage-link-target"><span>Тоҷикӣ</span></a></li><li class="interlanguage-link interwiki-tcy mw-list-item"><a href="https://tcy.wikipedia.org/wiki/%E0%B2%AD%E0%B2%BE%E0%B2%B0%E0%B2%A4%E0%B3%8A%E0%B2%A6_%E0%B2%AA%E0%B3%8D%E0%B2%B0%E0%B2%A7%E0%B2%BE%E0%B2%A8_%E0%B2%AE%E0%B2%82%E0%B2%A4%E0%B3%8D%E0%B2%B0%E0%B2%BF%E0%B2%A8%E0%B2%95%E0%B3%81%E0%B2%B2%E0%B3%81" title="ಭಾರತೊದ ಪ್ರಧಾನ ಮಂತ್ರಿನಕುಲು – Tulu" lang="tcy" hreflang="tcy" class="interlanguage-link-target"><span>ತುಳು</span></a></li><li class="interlanguage-link interwiki-tr mw-list-item"><a href="https://tr.wikipedia.org/wiki/Hindistan_ba%C5%9Fbakanlar%C4%B1_listesi" title="Hindistan başbakanları listesi – Turkish" lang="tr" hreflang="tr" class="interlanguage-link-target"><span>Türkçe</span></a></li><li class="interlanguage-link interwiki-uk mw-list-item"><a href="https://uk.wikipedia.org/wiki/%D0%A1%D0%BF%D0%B8%D1%81%D0%BE%D0%BA_%D0%BF%D1%80%D0%B5%D0%BC%27%D1%94%D1%80-%D0%BC%D1%96%D0%BD%D1%96%D1%81%D1%82%D1%80%D1%96%D0%B2_%D0%86%D0%BD%D0%B4%D1%96%D1%97" title="Список прем&#039;єр-міністрів Індії – Ukrainian" lang="uk" hreflang="uk" class="interlanguage-link-target"><span>Українська</span></a></li><li class="interlanguage-link interwiki-ur mw-list-item"><a href="https://ur.wikipedia.org/wiki/%D9%81%DB%81%D8%B1%D8%B3%D8%AA_%D9%88%D8%B2%D8%B1%D8%A7%D8%A6%DB%92_%D8%A7%D8%B9%D8%B8%D9%85_%D8%A8%DA%BE%D8%A7%D8%B1%D8%AA" title="فہرست وزرائے اعظم بھارت – Urdu" lang="ur" hreflang="ur" class="interlanguage-link-target"><span>اردو</span></a></li><li class="interlanguage-link interwiki-vi mw-list-item"><a href="https://vi.wikipedia.org/wiki/Danh_s%C3%A1ch_Th%E1%BB%A7_t%C6%B0%E1%BB%9Bng_%E1%BA%A4n_%C4%90%E1%BB%99" title="Danh sách Thủ tướng Ấn Độ – Vietnamese" lang="vi" hreflang="vi" class="interlanguage-link-target"><span>Tiếng Việt</span></a></li><li class="interlanguage-link interwiki-zh-yue mw-list-item"><a href="https://zh-yue.wikipedia.org/wiki/%E5%8D%B0%E5%BA%A6%E7%B8%BD%E7%90%86%E4%B8%80%E8%A6%BD" title="印度總理一覽 – Cantonese" lang="yue" hreflang="yue" class="interlanguage-link-target"><span>粵語</span></a></li></ul>
		<div class="after-portlet after-portlet-lang"><span class="wb-langlinks-edit wb-langlinks-link"><a href="https://www.wikidata.org/wiki/Special:EntityPage/Q492746#sitelinks-wikipedia" title="Edit interlanguage links" class="wbc-editpage">Edit links</a></span></div>
	</div>
</nav>

</div>

</div>

<footer id="footer" class="mw-footer" role="contentinfo" >
	<ul id="footer-info">
	<li id="footer-info-lastmod"> This page was last edited on 24 March 2022, at 11:33<span class="anonymous-show">&#160;(UTC)</span>.</li>
	<li id="footer-info-copyright">Text is available under the <a rel="license" href="//en.wikipedia.org/wiki/Wikipedia:Text_of_Creative_Commons_Attribution-ShareAlike_3.0_Unported_License">Creative Commons Attribution-ShareAlike License 3.0</a><a rel="license" href="//creativecommons.org/licenses/by-sa/3.0/" style="display:none;"></a>;
additional terms may apply.  By using this site, you agree to the <a href="//foundation.wikimedia.org/wiki/Terms_of_Use">Terms of Use</a> and <a href="//foundation.wikimedia.org/wiki/Privacy_policy">Privacy Policy</a>. Wikipedia® is a registered trademark of the <a href="//www.wikimediafoundation.org/">Wikimedia Foundation, Inc.</a>, a non-profit organization.</li>
</ul>

	<ul id="footer-places">
	<li id="footer-places-privacy"><a href="https://foundation.wikimedia.org/wiki/Privacy_policy" class="extiw" title="wmf:Privacy policy">Privacy policy</a></li>
	<li id="footer-places-about"><a href="/wiki/Wikipedia:About" title="Wikipedia:About">About Wikipedia</a></li>
	<li id="footer-places-disclaimer"><a href="/wiki/Wikipedia:General_disclaimer" title="Wikipedia:General disclaimer">Disclaimers</a></li>
	<li id="footer-places-contact"><a href="//en.wikipedia.org/wiki/Wikipedia:Contact_us">Contact Wikipedia</a></li>
	<li id="footer-places-mobileview"><a href="//en.m.wikipedia.org/w/index.php?title=List_of_prime_ministers_of_India&amp;mobileaction=toggle_view_mobile" class="noprint stopMobileRedirectToggle">Mobile view</a></li>
	<li id="footer-places-developers"><a href="https://www.mediawiki.org/wiki/Special:MyLanguage/How_to_contribute">Developers</a></li>
	<li id="footer-places-statslink"><a href="https://stats.wikimedia.org/#/en.wikipedia.org">Statistics</a></li>
	<li id="footer-places-cookiestatement"><a href="https://foundation.wikimedia.org/wiki/Cookie_statement">Cookie statement</a></li>
</ul>

	<ul id="footer-icons" class="noprint">
	<li id="footer-copyrightico"><a href="https://wikimediafoundation.org/"><img src="/static/images/footer/wikimedia-button.png" srcset="/static/images/footer/wikimedia-button-1.5x.png 1.5x, /static/images/footer/wikimedia-button-2x.png 2x" width="88" height="31" alt="Wikimedia Foundation" loading="lazy" /></a></li>
	<li id="footer-poweredbyico"><a href="https://www.mediawiki.org/"><img src="/static/images/footer/poweredby_mediawiki_88x31.png" alt="Powered by MediaWiki" srcset="/static/images/footer/poweredby_mediawiki_132x47.png 1.5x, /static/images/footer/poweredby_mediawiki_176x62.png 2x" width="88" height="31" loading="lazy"/></a></li>
</ul>

</footer>

<script>(RLQ=window.RLQ||[]).push(function(){mw.config.set({"wgPageParseReport":{"limitreport":{"cputime":"0.819","walltime":"1.011","ppvisitednodes":{"value":6145,"limit":1000000},"postexpandincludesize":{"value":161661,"limit":2097152},"templateargumentsize":{"value":8157,"limit":2097152},"expansiondepth":{"value":13,"limit":100},"expensivefunctioncount":{"value":3,"limit":500},"unstrip-depth":{"value":1,"limit":20},"unstrip-size":{"value":65474,"limit":5000000},"entityaccesscount":{"value":0,"limit":400},"timingprofile":["100.00%  777.018      1 -total"," 23.80%  184.905      2 Template:Reflist"," 16.89%  131.275      9 Template:Cite_news","  9.98%   77.521      1 Template:Politics_of_India","  9.70%   75.350      1 Template:Sidebar_with_collapsible_lists","  7.42%   57.627      1 Template:Short_description","  7.23%   56.174      6 Template:Party_index_link","  7.06%   54.850     45 Template:Sortname","  6.24%   48.505      6 Template:Color_box","  5.62%   43.685      9 Template:Main_other"]},"scribunto":{"limitreport-timeusage":{"value":"0.374","limit":"10.000"},"limitreport-memusage":{"value":8649562,"limit":52428800},"limitreport-logs":"table#1 {\n}\n"},"cachereport":{"origin":"mw1420","timestamp":"20220327190844","ttl":1814400,"transientcontent":false}}});});</script>
<script type="application/ld+json">{"@context":"https:\/\/schema.org","@type":"Article","name":"List of prime ministers of India","url":"https:\/\/en.wikipedia.org\/wiki\/List_of_prime_ministers_of_India","sameAs":"http:\/\/www.wikidata.org\/entity\/Q492746","mainEntity":"http:\/\/www.wikidata.org\/entity\/Q492746","author":{"@type":"Organization","name":"Contributors to Wikimedia projects"},"publisher":{"@type":"Organization","name":"Wikimedia Foundation, Inc.","logo":{"@type":"ImageObject","url":"https:\/\/www.wikimedia.org\/static\/images\/wmf-hor-googpub.png"}},"datePublished":"2004-07-20T13:01:24Z","dateModified":"2022-03-24T11:33:02Z","headline":"Wikimedia list article"}</script>
<script>(RLQ=window.RLQ||[]).push(function(){mw.config.set({"wgBackendResponseTime":138,"wgHostname":"mw1411"});});</script>
</body>
</html>"""
f.write(value)
f.close()
```
